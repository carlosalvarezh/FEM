## Elementos Unidimensionales Avanzados

### Introducción

El sencillo problema de modelar el comportamiento de una barra bajo fuerzas axiles mediante elementos unidimensionales de dos nodos estudiado en el capítulo anterior es sumamente instructivo porque, en esencia, resume las etapas básicas a seguir en el análisis de toda estructura por el método de los elementos finitos. Queda no obstante en el aire una serie de cuestiones tales como: ¿Es factible la utilización de elementos unidimensionales con funciones de forma de mayor grado? ¿Cuáles son las ventajas de estos elementos con respecto a la utilización del sencillo elemento de dos nodos estudiado en el capítulo anterior? Es posible garantizar que refinando progresivamente la malla, la solución converja al valor exacto? o ¿Qué factores afectan al error de la solución?. Estas y otras preguntas similares surgirán de forma casi inmediata al lector que se inicia en la utilización del método de elementos finitos. En este capítulo abordaremos dichas cuestiones aunque, como veremos, no existen respuestas definitivas para algunas de ellas y tendremos que limitarnos a ofrecer criterios orientativos que faciliten la utilización práctica del método. De nuevo para mayor sencillez nos referiremos siempre que sea posible al problema de Poisson unidimensional que permitirá presentar de forma clara conclusiones de validez general para otros casos más complejos.

La organización del capítulo esquemáticamente es la siguiente: En primer lugar, se presenta la obtención general de funciones de forma de los elementos unidimensionales válidos para el problema Poisson unidimensional. Dichas funciones de forma serán de gran utilidad para obtener las de elementos más complejos en dos y tres dimensiones. A continuación se tratan los conceptos de elemento isoparamétrico e integración numérica, ambos esenciales para el desarrollo de elementos finitos de órdenes superiores. Finalmente se estudian los requisitos para la convergencia de la solución y los tipos de errores más usuales que afectan a la misma.

### Elementos unidimensionales de clase $C^{0}$. Elementos Lagrangianos

#### Introducción

En un elemento unidimensional, la aproximación polinómica de una variable $u(x)$ puede escribirse en forma general como

$$
\phi(x)=\alpha_{0}+\alpha_{1} x+\alpha_{2} x^{2}+s
$$

donde $\alpha_{0}, \alpha_{1}$, etc., son constantes.

Tomando un polinomio de primer grado, se tiene

$$
\phi(x)=\alpha_{0}+\alpha_{1} x
$$

Para evitar tener que resolver un sistema de ecuaciones para obtener las funciones de forma de elementos unidimensionales de clase $C^{0}$ se puede hacer uso de las propiedades de los [polinomios de Lagrange](https://en.wikipedia.org/wiki/Lagrange_polynomial). Dichos polinomios toman un determinado valor en un punto y cero en un conjunto de puntos prefijados. Por consiguiente, normalizando dicho valor a la unidad y haciendo coincidir los puntos con la posición de los nodos, las funciones de forma coinciden precisamente con los $polinomios$ $de$ $Lagrange$, por lo que a estos elementos se les denomina *lagrangianos*.

La función de forma del nodo $i$ de un elemento lagrangiano unidimensional de $n$ nodos se obtiene directamente por la expresión general

<a id='Ec5_1'></a>
\begin{equation*}
N_{i}(x)^{(e)}=\frac{\left(x-x_{1}\right)\left(x-x_{2}\right) \cdots\left(x-x_{i-1}\right)\left(x-x_{i+1}\right) \cdots\left(x-x_{n}\right)}{\left(x_{i}-x_{1}\right)\left(x_{i}-x_{2}\right) \cdots\left(x_{i}-x_{i-1}\right)\left(x_{i}-x_{i+1}\right) \cdots\left(x_{i}-x_{n}\right)}
\label{eq:Ec5_1} \tag{5.1}
\end{equation*}

o, más sencillamente, por

<a id='Ec5_2'></a>
\begin{equation*}
N_{i}(x)^{(e)}=\prod_{j=1(j \neq i)}^{n}\left(\frac{x-x_{j}}{x_{i}-x_{j}}\right)
\label{eq:Ec5_2} \tag{5.2}
\end{equation*}

Así, por ejemplo, para un elemento de dos nodos se encuentra de nuevo

$$
\begin{aligned}
&N_{1}^{(e)}=\frac{x-x_{2}}{x_{1}-x_{2}}=\frac{x_{2}-x}{l^{(e)}} \\
&N_{2}^{(e)}=\frac{x-x_{1}}{x_{2}-x_{1}}=\frac{x-x_{1}}{l^{(e)}}
\end{aligned}
$$

Por conveniencia introduciremos el sistema de coordenadas natural o normalizado basado en la variable $\xi$ que se define como (Figura 5.1)

<a id='Fig5_1'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C05_F01_NaturalCoordinatesSystem.PNG?raw=true" width="350" />
</p>
<div style="text-align: center"><strong>Figura 5.1.</strong>Sistema de coordenadas natural, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

<a id='Ec5_3'></a>
\begin{equation*}
\xi=2 \frac{x-x_{c}}{l^{(e)}}
\label{eq:Ec5_3} \tag{5.3}
\end{equation*}

siendo $x_{c}$ la coordenada del centro del elemento, de manera que

- $\xi=-1$ en el extremo izquierdo del elemento

- $\xi=0$ en el punto central del elemento

- $\xi=1$ en el extremo derecho del elemento

La ecuación anterior transforma la geometría real del elemento en una geometría normalizada en la que la longitud del elemento es $2$. Podemos ahora expresar las funciones de forma en esta nueva geometría e independizar así la obtención de las mismas de la geometría real del elemento, lo cual es de gran interés práctico. La expresión general de $N_{i}(\xi)$ puede escribirse como

<a id='Ec5_4'></a>
\begin{equation*}
N_{i}(x)^{(e)}=\prod_{j=1(j \neq i)}^{n}\left(\frac{x-x_{j}}{x_{i}-x_{j}}\right)
\label{eq:Ec5_4} \tag{5.4}
\end{equation*}

Por tanto, para un elemento lagrangiano de dos nodos con $\xi_{1}=-1$ y $\xi_{2}=+1$, se deduce

<a id='Ec5_5'></a>
\begin{equation*}
\begin{aligned}
&N_{1}^{(e)}=\frac{\xi-\xi_{2}}{\xi_{1}-\xi_{2}}=\frac{1}{2}\left(1-\xi\right) \\
&N_{2}^{(e)}=\frac{\xi-\xi_{1}}{\xi_{2}-\xi_{1}}=\frac{1}{2}\left(1+\xi\right)
\end{aligned}
\label{eq:Ec5_5} \tag{5.5}
\end{equation*}

Es inmediato ver que sustituyendo el valor de $\xi$ se recupera la expresión cartesiana de las funciones de forma. 

Para un elemento cuadrático de tres nodos con $\xi_{1}=-1, \xi_{2}=0, \xi_{3}=+1$ (*Figura 5.2*) las funciones de forma se obtienen por

<a id='Fig5_2'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C05_F02_QuadraticCubic_1D_Element_C0_continuity.PNG?raw=true" width="350" />
</p>
<div style="text-align: center"><strong>Figura 5.2.</strong>Elemento 1D cuadrático y cúbico de clase C0, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

<a id='Ec5_6'></a>
\begin{equation*}
\begin{aligned}
&N_{1}^{(e)}=\frac{\left(\xi-\xi_{2}\right)\left(\xi-\xi_{3}\right)}{\left(\xi_{1}-\xi_{2}\right)\left(\xi_{2}-\xi_{3}\right)}=\frac{1}{2} \xi(\xi-1) \\
&N_{2}^{(e)}=\frac{\left(\xi-\xi_{1}\right)\left(\xi-\xi_{3}\right)}{\left(\xi_{2}-\xi_{1}\right)\left(\xi_{2}-\xi_{3}\right)}=(1+\xi)(1-\xi) \\
&N_{3}^{(e)}=\frac{\left(\xi-\xi_{1}\right)\left(\xi-\xi_{2}\right)}{\left(\xi_{3}-\xi_{1}\right)\left(\xi_{3}-\xi_{2}\right)}=\frac{1}{2} \xi(1+\xi)
\end{aligned}
\label{eq:Ec5_6} \tag{5.6}
\end{equation*}

Finalmente, para ur elemento cúbico de cuatro nodos $1,2,3$ y 4 con $\xi_{1}=-1, \xi_{2}=-1 / 3, \xi_{3}=1 / 3$ y $\xi_{4}=1$ (*Figura 5.2*), las funciones de forma son

<a id='Ec5_7'></a>
\begin{equation*}
\begin{aligned}
&N_{1}^{(e)}=\frac{\left(\xi-\xi_{2}\right)\left(\xi-\xi_{3}\right)\left(\xi-\xi_{4}\right)}{\left(\xi_{1}-\xi_{2}\right)\left(\xi_{1}-\xi_{3}\right)\left(\xi_{1}-\xi_{4}\right)}=-\frac{9}{16}\left(\xi+\frac{1}{3}\right)\left(\xi-\frac{1}{3}\right)(\xi-1) \\
&N_{2}^{(e)}=\frac{\left(\xi-\xi_{1}\right)\left(\xi-\xi_{3}\right)\left(\xi-\xi_{4}\right)}{\left(\xi_{2}-\xi_{1}\right)\left(\xi_{2}-\xi_{3}\right)\left(\xi_{2}-\xi_{4}\right)}=\quad \frac{27}{16}(\xi+1)\left(\xi-\frac{1}{3}\right)(\xi-1) \\
&N_{3}^{(e)}=\frac{\left(\xi-\xi_{1}\right)\left(\xi-\xi_{2}\right)\left(\xi-\xi_{1}\right)}{\left(\xi_{3}-\xi_{1}\right)\left(\xi_{3}-\xi_{2}\right)\left(\xi_{3}-\xi_{4}\right)}=-\frac{27}{16}(\xi+1)\left(\xi+\frac{1}{3}\right)(\xi-1) \\
&N_{4}^{(e)}=\frac{\left(\xi-\xi_{1}\right)\left(\xi-\xi_{2}\right)\left(\xi-\xi_{3}\right)}{\left(\xi_{4}-\xi_{1}\right)\left(\xi_{4}-\xi_{2}\right)\left(\xi_{4}-\xi_{3}\right)}=\quad \frac{9}{16}(\xi+1)\left(\xi+\frac{1}{3}\right)\left(\xi-\frac{1}{3}\right)
\end{aligned}
\label{eq:Ec5_7} \tag{5.7}
\end{equation*}

La expresión cartesiana de las funciones de forma anteriores se obtiene inmediatamente haciendo uso de la transformación entre $\xi$ y $x$. No obstante, esta operación no es necesaria y de hecho es de poco interés práctico.

Proponemos al lector que obtenga por sí mismo la expresión de las funciones de forma de otros elementos unidimensionales lagrangianos de órdenes superiores.

La aproximación de la incógnita dentro de cada elemento se escribe como

$$
\hat{\phi}=\sum_{i=1}^{3} N_{i}^{(e)} \phi_{i}=N_{1}^{(e)} \phi_{1}^{(e)}+N_{2}^{(e)} \phi_{2}^{(e)}+N_{3}^{(e)} \phi_{3}^{(e)}
$$

Obsérvese que la relación entre numeración nodal local y global se deduce

$$
\begin{aligned}
&\phi_{1}^{(1)} \equiv \phi_{1}, \phi_{2}^{(1)} \equiv \phi_{2}, \phi_{3}^{(1)} \equiv \phi_{3} \\
&\phi_{1}^{(2)} \equiv \phi_{3}, \phi_{2}^{(2)} \equiv \phi_{4}, \phi_{3}^{(2)} \equiv \phi_{5}
\end{aligned}
$$

Aplicando el *MRP de Galerkin* se llega a la ecuación siguiente

$$
\int_{\Omega} \frac{\mathrm{d} N_{i}}{\mathrm{d} x} k\left[\sum_{j=1}^{5} \frac{\mathrm{d} N_{j}}{\mathrm{d} x} \phi_{j}\right] \mathrm{d} x=\int_{\Omega} N_{i} Q \mathrm{d} x+\left[N_{i} q\right]_{0}-\left[N_{i} \bar{q}\right]_{i} ; \quad i=1,2, \ldots, 5
$$

Desarrollaremos dicha ecuación para los diferentes valores de

$$
\int_{0}^{l / 2} \frac{\mathrm{d} N_{1}^{(1)}}{\mathrm{d} x} k\left[\frac{\mathrm{d} N_{1}^{(1)}}{\mathrm{d} x} \phi_{1}+\frac{\mathrm{d} N_{2}^{(1)}}{\mathrm{d} x} \phi_{2}+\frac{\mathrm{d} N_{3}^{(1)}}{\mathrm{d} x} \phi_{3}\right] \mathrm{d} x=\int_{0}^{l / 2} N_{1}^{(1)} Q \mathrm{d} x+q_{0}
$$

$i=2$

$$
\begin{aligned}
&\int_{0}^{l / 2} \frac{\mathrm{d} N_{2}^{(1)}}{\mathrm{d} x} k\left[\frac{\mathrm{d} N_{1}^{(1)}}{\mathrm{d} x} \phi_{1}+\frac{\mathrm{d} N_{2}^{(1)}}{\mathrm{d} x} \phi_{2}+\frac{\mathrm{d} N_{3}^{(1)}}{\mathrm{d} x} \phi_{3}\right] \mathrm{d} x=\int_{0}^{l / 2} N_{2}^{(1)} Q \mathrm{d} x \\
\end{aligned}
$$

$i=3$

$$
\begin{gathered}
\int_{0}^{l / 2} \frac{\mathrm{d} N_{3}^{(1)}}{\mathrm{d} x} k\left[\frac{\mathrm{d} N_{1}^{(1)}}{\mathrm{d} x} \phi_{1}+\frac{\mathrm{d} N_{2}^{(1)}}{\mathrm{d} x} \phi_{2}+\frac{\mathrm{d} N_{3}^{(1)}}{\mathrm{d} x} \phi_{3}\right] \mathrm{d} x+ \\
\int_{l / 2}^{l} \frac{\mathrm{d} N_{1}^{(2)}}{\mathrm{d} x} k\left[\frac{\mathrm{d} N_{1}^{(2)}}{\mathrm{d} x} \phi_{3}+\frac{\mathrm{d} N_{2}^{(2)}}{\mathrm{d} x} \phi_{4}+\frac{\mathrm{d} N_{3}^{(2)}}{\mathrm{d} x} \phi_{5}\right] \mathrm{d} x= \\
\int_{0}^{l / 2} N_{3}^{(1)} Q \mathrm{d} x+\int_{l / 2}^{l} N_{1}^{(2)} Q \mathrm{d} x
\end{gathered}
$$

$i=4$

$$
\int_{l / 2}^{l} \frac{\mathrm{d} N_{2}^{(2)}}{\mathrm{d} x} k\left[\frac{\mathrm{d} N_{1}^{(2)}}{\mathrm{d} x} \phi_{3}+\frac{\mathrm{d} N_{2}^{(2)}}{\mathrm{d} x} \phi_{4}+\frac{\mathrm{d} N_{3}^{(2)}}{\mathrm{d} x} \phi_{5}\right] \mathrm{d} x=\int_{l / 2}^{l} N_{2}^{(2)} Q \mathrm{d} x
$$

$i=5$

$$
\begin{aligned}
&\int_{l / 2}^{l} \frac{\mathrm{d} N_{3}^{(2)}}{\mathrm{d} x} k\left[\frac{\mathrm{d} N_{1}^{(2)}}{\mathrm{d} x} \phi_{3}+\frac{\mathrm{d} N_{2}^{(2)}}{\mathrm{d} x} \phi_{4}+\frac{\mathrm{d} N_{3}^{(2)}}{\mathrm{d} x} \phi_{5}\right] \mathrm{d} x=\int_{l / 2}^{l} N_{3}^{(2)} Q \mathrm{d} x-\bar{q}
\end{aligned}
$$

$$
\underbrace{\begin{bmatrix}

\underbrace{\begin{Bmatrix}
\phi_1 \\
\phi_2 \\
\phi_3 \\
\phi_4 \\
\phi_5
\end{Bmatrix}}_{\mathbf{a}}
\underbrace{\begin{Bmatrix}
f_1^{(1)}+q_0 \\
f_2^{(1)} \\
f_3^{(1)} + f_1^{(2)} \\
f_2^{(2)}\\
f_3^{(2)}-\bar{q}
\end{Bmatrix}}_{\mathbf{f}}
$$

Se observa que las expresiones de $\mathbf{K}$ y $\mathbf{f}$ pueden obtenerse ensamblando en la forma usual las contribuciones elementales dadas por

$$
\mathbf{K}^{(e)}=\left[\begin{array}{lll}
K_{11}^{(e)} & K_{12}^{(e)} & K_{13}^{(e)} \\
K_{21}^{(e)} & K_{22}^{(e)} & K_{23}^{(e)} \\
K_{31}^{(e)} & K_{32}^{(e)} & K_{33}^{(e)}
\end{array}\right] ; \quad \mathbf{f}^{(e)}=\left\{\begin{array}{l}
f_{1}^{(e)} \\
f_{2}^{(e)} \\
f_{3}^{(e)}
\end{array}\right\}
$$

Por otra parte, las expresiones de $K_{i j}^{(e)}$ y $f_{i}^{(e)}$ coinciden con las obtenidas en un apartado anterior, asi

$$
K_{ij}^{(e)}=\int_{l^{(e)}} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} x} k \frac{\mathrm{d} N_{j}}{\mathrm{d} x} \mathrm{d} x \quad \text { y } f_{i}^{(e)}=\int_{l^{(e)}} N_{i} Q \mathrm{d} x + \left( \left(N_{i} q\right)_{0}-\left(N_{i} q\right)_{l}\right)
$$

Se deja al lector la solución final de este problema. Se comprobará que la solución coincide con la exacta en los nodos y dentro de cada elemento, dado que se utilizan funciones de aproximación cuadráticas que coinciden con la forma de la solución exacta.

### Formulación isoparamétrica e integración numérica

#### Introducción

En este apartado y en el siguiente se introducirán dos importantes conceptos sin los cuales es prácticamente imposible que el método de los elementos finitos se hubiese desarrollado hasta los niveles en que hoy se encuentra.

El primer concepto es el de formulación paramétrica. La idea es interpolar la geometría del elemento a partir de las coordenadas de una serie de puntos conocidos. Dicha interpolación es esencial para poder encontrar una relación general entre las coordenadas naturales y las cartesianas.

El segundo concepto es el de integración numérica. En la mayor parte de los cusos prácticos el cálculo analítico de las integrales del elemento es inabordable y la integración numérica es la única opción para evaluarlas de forma precisa y sencilla.

En este apartado y en el siguiente desarrollaremos las técnicas anteriores en su aplicación a elemeitos unidimensionales de clase $C^{0}$. Advertimos que sus ventajas no se apreciarán totalmente hasta que no estudiemos problemas más complicados en los que intervengan elementos bi o tridimensionales.


#### Concepto de interpolación paramétrica

Recordemos que para un elemento lineal de dos nodos la variable $\phi$ en un punto del elemento se expresa dentro del elemento por

<a id='Ec5_8'></a>
\begin{equation*}
\begin{aligned}
\hat{\phi}(\xi)=N_{1}^{(e)}(\xi) \phi_{1}^{(e)}+N_{2}^{(e)}(\xi) \phi_{2}^{(e)}
\end{aligned}
\label{eq:Ec5_8} \tag{5.8}
\end{equation*}

Obsérvese que hemos adoptado la expresión normalizada de las funciones de forma. Salvo en contadas excepciones ésta será la manera usual de proceder. Por otra parte, el gradiente $g$ se obtiene dentro de un elemento por

<a id='Ec5_9'></a>
\begin{equation*}
\begin{aligned}
g=\frac{\mathrm{d} \hat{\phi}}{\mathrm{d} x}=\frac{\mathrm{d} N_{1}^{(e)}(\xi)}{\mathrm{d} x} \phi_{1}^{(e)}+\frac{\mathrm{d} N_{2}^{(e)}(\xi)}{\mathrm{d} x} \phi_{2}^{(e)}
\end{aligned}
\label{eq:Ec5_9} \tag{5.9}
\end{equation*}

Por tanto, para el cálculo del gradiente necesitamos conocer la derivada de $N_{i}^{(e)}$ con respecto a la coordenada cartesiana $x$. Este cálculo es inmediato si las funciones de forma se expresan en función de $x$. No obstante, como hemos indicado, en general esto no será así debido a la utilización del sistema de coordenadas naturales.

Así, pues, la evaluación de dichas derivadas implica las operaciones siguientes:

<a id='Ec5_10'></a>
\begin{equation*}
\begin{aligned}
&\frac{\mathrm{d} N_{1}^{(e)}(\xi)}{\mathrm{d} x}=\frac{\mathrm{d} N_{1}^{(e)}(\xi)}{\mathrm{d} \xi} \frac{\mathrm{d} \xi}{\mathrm{d} x}=\frac{\mathrm{d}}{\mathrm{d} \xi}\left(\frac{1-\xi}{2}\right) \frac{\mathrm{d} \xi}{\mathrm{d} x}=-\frac{1}{2} \frac{\mathrm{d} \xi}{\mathrm{d} x} \\
&\frac{d N_{2}^{(e)}(\xi)}{\mathrm{d} x}=\frac{\mathrm{d} N_{2}^{(e)}(\xi)}{\mathrm{d} \xi} \frac{\mathrm{d} \xi}{\mathrm{d} x}=\frac{\mathrm{d}}{\mathrm{d} \xi}\left(\frac{1+\xi}{2}\right) \frac{\mathrm{d} \xi}{\mathrm{d} x}=\quad \frac{1}{2} \frac{\mathrm{d} \xi}{\mathrm{d} x}
\end{aligned}
\label{eq:Ec5_10} \tag{5.10}
\end{equation*}

con lo que la expresión del gradiente se escribe como

<a id='Ec5_11'></a>
\begin{equation*}
\begin{aligned}
g=-\frac{1}{2}\left(\frac{\mathrm{d} \xi}{\mathrm{d} x}\right) \phi_{1}^{(e)}+\frac{1}{2}\left(\frac{\mathrm{d} \xi}{\mathrm{d} x}\right) \phi_{2}^{(e)}
\end{aligned}
\label{eq:Ec5_11} \tag{5.11}
\end{equation*}

Para completar el cálculo de $g$ hay que evaluar $\dfrac{\mathrm{d} \xi}{\mathrm{d} x}$, lo que exige conocer una relación explícita entre $x$ y $\xi$.

Dicha relación puede obtenerse mediante una interpolación paramétrica de la geometría del elemento. Asi, si se conocen las coordenadas $x_{1}, x_{2}, \ldots$, $x_{m}$ de $m$ puntos cualesquiera del elemento, se puede calcular la coordenada de cualquier punto del mismo interpolando los valores de las coordenadas conocidas. Dicha interpolación puede escribirse en la forma

<a id='Ec5_12'></a>
\begin{equation*}
\begin{aligned}
x=\hat{N}_{1}^{(e)}(\xi) x_{1}+\hat{N}_{2}^{(e)}(\xi) x_{2}+\cdots+\hat{N}_{m}^{(e)}(\xi) x_{m}
\end{aligned}
\label{eq:Ec5_12} \tag{5.12}
\end{equation*}

Se deduce de lo anterior que $\hat{N}_{i}(\xi)$ son funciones de interpolación de geometrí que satisfacen los mismos requisitos que las funciones de forma utilizadas para interpolar el campo de desplazamientos, es decir, $\hat{N}_{i}(\xi)$ debe tomar el valor unidad en el punto $i$ y cero en el resto de los $m-1$ puntos. Por tanto, las funciones $\hat{N}_{i}(\xi)$ se obtienen en base al número de puntos escogido para la interpolación de la geometría.

Obsérvese que la ecuación anterior proporciona directamente la relación entre las coordenadas $x$ y $\xi$ buscada. Dicha ecuación puede también interpretarse como la transformación de coordenadas $\xi \rightarrow x$ de manera que a cada punto del espacio normalizado $[-1,1]$ le corresponde otro del espacio cartesiano $\left[x_{1}, x_{2}\right]$. Es fundamental que dicha transformación sea biunf́voca, lo que en general depende de la geometría del elemento. Más adelante volveremos a tratar este tema.

\end{gathered}
$$

donde las $N_{i}$ son las funciones de forma cúbicas. Es fácil comprobar que en este caso se obtiene, tras operar

Así, pues, como era de esperar, la aproximación cúbica escogida reproduce exactamente la función original, siendo el error de aproximación nulo en este caso.

#### Formulación isoparamétrica del elemento de dos nodos 

Expresaremos la geometría del elemento en función de las coordenadas de los dos nodos como

<a id='Ec5_13'></a>
\begin{equation*}
\begin{aligned}
x(\xi)=N_{1}(\xi) x_{1}+N_{2}(\xi) x_{2}
\end{aligned}
\label{eq:Ec5_13} \tag{5.13}
\end{equation*}

donde $N_{1}$ y $N_{2}$ son precisamente las mismas funciones de forma lineales utilizadas para describir la aproximación de la incógnita.

La ecuación anterior proporciona automáticamente $\dfrac{\mathrm{d} x}{\mathrm{d} \xi}$ como

$$
\frac{\mathrm{d} x}{\mathrm{d}\xi}=\frac{\mathrm{d} N_{1}^{(e)}}{\mathrm{d} \xi} x_{1}+\frac{\mathrm{d} N_{2}^{(e)}}{\mathrm{d} \xi} x_{2}=-\frac{1}{2} x_{1}+\frac{1}{2} x_{2}=\frac{l^{(e)}}{2}
$$

con lo que

<a id='Ec5_14'></a>
\begin{equation*}
\begin{aligned}
\mathrm{d} x=\frac{l^{(e)}}{2} \mathrm{d} \xi \quad \text { y } \quad \frac{\mathrm{d} \xi}{\mathrm{d} x}=\frac{2}{l^{(e)}}
\end{aligned}
\label{eq:Ec5_14} \tag{5.14}
\end{equation*}

Sustituyendo estas expresiones en la ecuación anterior se deduce

$$
\begin{aligned}
&\frac{\mathrm{d} N_{1}^{(e)}}{\mathrm{d} x}=\frac{2}{l^{(e)}} \frac{\mathrm{d} N_{1}^{(e)}}{\mathrm{d} \xi}=-\frac{1}{l^{(e)}} \\
&\frac{\mathrm{d} N_{2}^{(e)}}{\mathrm{d} x}=\frac{2}{l^{(e)}} \frac{\mathrm{d} N_{2}^{(e)}}{\mathrm{d} \xi}=\frac{1}{l^{(e)}}
\end{aligned}
$$

Resultado que, por otra parte, ya conocíamos. Hay que resaltar que en este caso particular pueden obtenerse estas expresiones de una manera más sencilla. No obstante, hemos prisorido seguir aquí un procedimiento más sistemático que facilítará la comprensión del desarrollo de elementos isoparamétricos más complejos.

La matriz de rigidez y el vector de fuerzas nodales equivalentes se pueden expresar ahora en el sistema normalizado como

<a id='Ec5_15'></a>
\begin{equation*}
\begin{aligned}
&K_{i j}^{(e)}=\int_{l^{(e)}} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} x} k \frac{\mathrm{d} N_{j}^{(e)}}{\mathrm{d} x} \mathrm{d} x=\int_{-1}^{+1} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} \xi} k \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} \xi} \frac{2}{l^{(e)}} \mathrm{d} \xi=(-1)^{i+j}\left(\frac{k}{l}\right)^{(e)} \\
&f_{i}^{(e)}=\int_{l^{(e)}} N_{i}^{(e)} Q \mathrm{d} x=\int_{-1}^{+1} N_{i}^{(e)} Q \frac{l^{(e)}}{2} \mathrm{d} \xi=\frac{Q l^{(e)}}{2}
\end{aligned}
\label{eq:Ec5_15} \tag{5.15}
\end{equation*}

Obviamente, para deducir la expresión de $f_{i}^{(e)}$ hemos supuesto que $Q$ es constante sobre el elemento. Las expresiones de $K_{i j}^{(e)}$ y $f_{i}^{(e)}$ coinciden como era de esperar con las obtenidas en el capitulo anterior.


#### Formulación isoparamétrica del elemento cuadrático de tres nodos

Estudiaremos ahora el elemento de tres nodos con funciones de forma polinómicas de segundo grado. La aproximación se expresa por

<a id='Ec5_16'></a>
\begin{equation*}
\begin{aligned}
\phi=N_{1}^{(e)}(\xi) \phi_{1}^{(e)}+N_{2}^{(e)}(\xi) \phi_{2}^{(e)}+N_{3}^{(e)}(\xi) \phi_{3}^{(e)}
\end{aligned}
\label{eq:Ec5_16} \tag{5.16}
\end{equation*}

En la formulación isoparamétrica la coordenada $x$ de un punto cualquiera del elemento se escribe como

<a id='Ec5_17'></a>
\begin{equation*}
\begin{aligned}
x=N_{1}^{(e)}(\xi) x_{1}+N_{2}^{(e)}(\xi) x_{2}+N_{3}^{(e)}(\xi) x_{3}
\end{aligned}
\label{eq:Ec5_17} \tag{5.17}
\end{equation*}

El gradiente dentro del elemento se obtiene por

<a id='Ec5_18'></a>
\begin{equation*}
\begin{aligned}
g=\frac{\mathrm{d} \phi}{\mathrm{d} x}=\sum_{i=1}^{3} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} x} \phi_{i}=\left[\frac{\mathrm{d} N_{1}^{(e)}}{\mathrm{d} \xi} \frac{\mathrm{d} \xi}{\mathrm{d} x}, \frac{\mathrm{d} N_{2}^{(e)}}{\mathrm{d} \xi} \frac{\mathrm{d} \xi}{\mathrm{d} x}, \frac{\mathrm{d} N_{3}^{(e)}}{\mathrm{d} \xi} \frac{\mathrm{d} \xi}{\mathrm{d} x}\right]\left\{\begin{array}{c}
\phi_{1} \\
\phi_{2} \\
\phi_{3}
\end{array}\right\}
\end{aligned}
\label{eq:Ec5_18} \tag{5.18}
\end{equation*}

Por otra parte, de las expresiones de $N_{i}$ se deduce

$$
\frac{\mathrm{d} N_{1}^{(e)}}{\mathrm{d} \xi}=\xi-\frac{1}{2} ; \frac{\mathrm{d} N_{2}^{(e)}}{\mathrm{d} \xi}=-2 \xi ; \frac{\mathrm{d} N_{3}^{(e)}}{\mathrm{d} \xi}=\xi+\frac{1}{2}
$$

Por tanto

<a id='Ec5_19'></a>
\begin{equation*}
\begin{aligned}
\frac{\mathrm{d} x}{\mathrm{d} \xi}=& \frac{\mathrm{d} N_1^{(e)}}{\mathrm{d} \xi} x_{1}+\frac{\mathrm{d} N_{2}^{(e)}}{\mathrm{d} \xi} x_{2}+\frac{\mathrm{d} N_{3}^{(e)}}{\mathrm{d} \xi} x_{3}=\left(\xi-\frac{1}{2}\right) x_{1}-2 \xi x_{2}+\left(\xi+\frac{1}{2}\right) x_{3}=\frac{l^{(e)}}{2}+\xi\left(x_{1}+x_{3}-2 x_{2}\right)=j^{(e)}
\end{aligned}
\label{eq:Ec5_19} \tag{5.19}
\end{equation*}

y

<a id='Ec5_20'></a>
\begin{equation*}
\begin{aligned}
\frac{d \xi}{d x}=\frac{1}{j^{(\epsilon)}}
\end{aligned}
\label{eq:Ec5_20} \tag{5.20}
\end{equation*}

La expresión anterior proporciona la relación entre $d x$ y $d \xi$ en función de las coordenadas de los tres nodos del elemento. Es interesante advertir que en el caso (por otra parte usual) de que el nodo intermedio esté situado en el centro del elemento, se tiene

<a id='Ec5_21'></a>
\begin{equation*}
\begin{aligned}
j^{(e)}=\frac{l^{(e)}}{2} \quad \text { y } \quad \frac{\mathrm{d} \xi}{\mathrm{d} x}=\frac{2}{l^{(e)}}
\end{aligned}
\label{eq:Ec5_21} \tag{5.21}
\end{equation*}

y, por consiguiente

<a id='Ec5_22'></a>
\begin{equation*}
\begin{aligned}
\frac{\mathrm{d} x}{\mathrm{d} \xi}=\frac{2}{l^{(e)}} \quad \text{ y } \quad \mathrm{d} x=\frac{l^{(e)}}{2} \mathrm{d} \xi
\end{aligned}
\label{eq:Ec5_22} \tag{5.22}
\end{equation*}

La matriz de rigidez y el vector de fuerzas nodales del elemento se obtienen de forma idéntica a lo explicado anteriormente para el elemento de dos nodos.

<a id='Ec5_23'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
K_{i j}^{(e)}=\int_{l^{(e)}} \frac{\mathrm{d} N_{i}}{\mathrm{d} x} k \frac{\mathrm{d} N_{j}}{\mathrm{d} x} \mathrm{d} x=\int_{-1}^{+1} \frac{\mathrm{d} N_{i}}{\mathrm{d} \xi} k \frac{\mathrm{d} N_{i}}{\mathrm{d} \xi} \frac{1}{j^{(e)}} \mathrm{d} \xi \\
f_{i}^{(e)}=\int_{l^{(e)}} N_{i} Q \mathrm{d} x=\int_{-1}^{+1} N_{i} Q j^{(e)} \mathrm{d} \xi
\end{gathered}
\end{aligned}
\label{eq:Ec5_23} \tag{5.23}
\end{equation*}

Suponiendo el nodo intermedio centrado en el elemento $\left(j^{(e)}=l^{(e)} / 2 \right)$ y aceptando que $k$ y $Q$ son constantes sobré el elemento se obtiene, tras operar

<a id='Ec5_24'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{K}^{(e)}=\left(\frac{k}{6l}\right)^{(e)}\left[\begin{array}{rrr}
14 & -16 & 2 \\
-16 & 32 & -16 \\
2 & -16 & 14
\end{array}\right]\\
\\
f^{(e)}=\frac{(Q)^{(e)}}{6}\left\{\begin{array}{l}
1 \\
4 \\
1
\end{array}\right\}
\end{gathered}
\end{aligned}
\label{eq:Ec5_24} \tag{5.24}
\end{equation*}

de donde se deduce que el nodo central absorbe cuatro veces más flujo que los nodos extremos, lo que no es un resultado evidente a priori y que, sin embargo, se obtiene de forma "natural" por aplicación de los sencillos conceptos de equilibrio en forma integral.

Evidentemente, para una posición arbitraria del nodo intermedio las expresiones de la matriz de rigidez y del vector de fuerzas nodales serían diferentes. De hecho, en dicho caso los términos de la matriz $K_{i j}$ contienen funciones racionales, como se deduce de la expresión general de $j^{(e)}$, y la integración directa de la matriz $\mathbf{K}$ no es tan inmediata.

La formulación isoparamétrica de elementos unidimensionales de órdenes superiores sigue exactamente los mismos pasos explicados aquí para los elementos lineal y cuadrático. No obstante, se resalta que a medida que el orden del elemento aumenta se incrementa también la dificultad del cálculo de las integrales del elemento. Dicha dificultad se puede sortear de una manera sencilla y elegante haciendo uso de la integración numérica como se explica en el apartado siguiente.


### Integración Numérica

#### Introducción

Hemos visto que pese a la gran simplicidad de los elementos unidimensionales, el cálculo analítico de las integrales del elemento puede resultar laborioso. En particular, sỉ se usa una formulación isoparamétrica que conduzca a expresiones racionales en los coeficientes de $K_{i j}^{(c)} \circ f_{i}^{(c)} .$ De hecho, en la mayor parte de los elementos bi o tridimensionales isoparamétricos el cálculo directo de dichas integrales es inabordable, salvo en raras excepciones, y es imprescindible hacer uso de la integración numérica.

Entrar en detalle en todos los fundamentos matemáticos relacionados con la integración numérica se sale de los objetivos de estos apuntes. A efectos de simplificar ideas presentaremos aquí únicamente la integnación numérica de [Gauss-Legendre](https://en.wikipedia.org/wiki/Gauss%E2%80%93Legendre_quadrature), por ser este el procedimiento más popular y utilizado en relación con el método de los elementos finitos. En este apartado estudiaremos únicamente las ideas básicas de dicha regla de integración en su aplicación unidimensional, dejando para capítulos posteriores la extensión de la misma a dos y tres dimensiones.

Supongamos una función $f(\xi)$ para la que se desea calcular la integral en el intervalo $[-\mathbf{1},+\mathbf{1}]$, es decir

<a id='Ec5_25'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
I=\int_{-1}^{+1} f(\xi) \mathrm{d} \xi
\end{gathered}
\end{aligned}
\label{eq:Ec5_25} \tag{5.25}
\end{equation*}

La *regla de integración* o *cuadratura de Gauss-Legendre* expresa el valor de dicha integral como suma de los productos de los valores del integrando en una serie de puntos conocidos en el interior del intervalo por unos coeficientes (pesos) determinados. Es decir, para una cuadratura de orden $p$ se tiene que

<a id='Ec5_26'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
I_{p}=\sum_{i=1}^{p} f\left(\xi_{i}\right) W_{i}
\end{gathered}
\end{aligned}
\label{eq:Ec5_26} \tag{5.26}
\end{equation*}

donde $W_{i}$ es el peso correspondiente al punto de integración $i$, y $p$ el número de dichos puntos. Es interesante destacar que la *cuadratura de Gauss-Legendre* de orden $n$ integra exactamente un polinomio de grado $2 n-1$ o menor. Por otra parte, el error en el cálculo aproximado de una integral es de orden $\mathcal{O}\left(\Delta^{2 n}\right)$, donde $\Delta$ es la distancia entre los puntos de integración.

$$
\begin{gathered}

### Cálculo de las matrices y vectores de un elemento isoparamétrico unidimensional para la solución de la ecuación de Poisson

#### Introducción

Presentaremos las etapas necesarias para el cálculo de la matriz de rigidez y del vector de fuerzas nodales equivalentes de un elemento unidimensional isoparamétrico de n nodos para la solución de la ecuación de Poisson. Las etapas se han ordenado de manera que se facilite al máximo la programación de todas las operaciones.

#### Interpolación de la incógnita

La variable $\phi$ en el interior del elemento se expresa por

<a id='Ec5_27'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\phi \simeq \hat{\phi} &= N_{1}^{(e)} \phi_{1}^{(e)}+N_{2}^{(e)} \phi_{2}^{(e)}+\ldots+N_{n}^{(e)} \phi_{n}^{(e)} \\
&=\sum_{i=1}^{n} N_{i}^{(e)} \phi_{i}^{(e)}=\left[N_{1}^{(e)}, N_{2}^{(e)}, \ldots, N_{n}^{(e)}\right]\left\{\begin{array}{c}
\phi_{1} \\
\phi_{2} \\
\vdots \\
\phi_{n}
\end{array}\right\}^{(e)}=\mathbf{N} \mathbf{a}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec5_27} \tag{5.27}
\end{equation*}


#### Interpolación de la geometría

En la formulación isoparamétrica la coordenada $x$ de un punto del elemento se calcula por

<a id='Ec5_28'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
x=N_{1}^{(e)} x_{1}+N_{2}^{(e)} x_{2}+\ldots+N_{n}^{(e)} x_{n}=\sum_{i=1}^{n} N_{i}^{(e)} x_{i}
\end{gathered}
\end{aligned}
\label{eq:Ec5_28} \tag{5.28}
\end{equation*}


#### Interpolación del gradiente

El gradiente de $\phi$ se expresa en función de las variables nodales por

<a id='Ec5_29'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
g=\frac{\mathrm{d} \phi}{\mathrm{d} x}=\frac{\mathrm{d} N_{1}^{(e)}}{\mathrm{d} x} \phi_{1}^{(\mathrm{e})}+\frac{\mathrm{d} N_{2}^{(e)}}{\mathrm{d} x} \phi_{2}^{(e)}+\ldots+\frac{\mathrm{d} N_{n}^{(e)}}{\mathrm{d} x} \phi_{n}^{(e)}=\sum_{i=1}^{n} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} x} \phi_{i}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec5_29} \tag{5.29}
\end{equation*}

Por otra parte, las derivades cartesianas de las funciones de forma se obtienen por

<a id='Ec5_30'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} x}=\frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} \xi} \frac{\mathrm{d} \xi}{\mathrm{d} x}
\end{gathered}
\end{aligned}
\label{eq:Ec5_30} \tag{5.30}
\end{equation*}

De la interpolación de la geometría se deduce

<a id='Ec5_31'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\frac{\mathrm{d} x}{\mathrm{d} \xi}=\sum_{i=1}^{n} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} \xi} x_{i}=j^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec5_31} \tag{5.31}
\end{equation*}

Por consiguiente

<a id='Ec5_32'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathrm{d} x=j^{(e)} \mathrm{d} \xi \quad ; \quad \frac{\mathrm{d} \xi}{\mathrm{d} x}=\frac{1}{j^{(e)}}
\end{gathered}
\end{aligned}
\label{eq:Ec5_32} \tag{5.32}
\end{equation*}

y

<a id='Ec5_33'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} x}=\frac{1}{j^{(e)}} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} \xi}
\end{gathered}
\end{aligned}
\label{eq:Ec5_33} \tag{5.33}
\end{equation*}

Es interesante advertir que $j^{(e)}$ es la relación entre los diferenciales de longitud de los dos sistemas de coordenadas $x$ y $\xi$ pudiendo interpretarse como el determinante del jacobiano de la transformación $x \rightarrow \xi$, y que precisamente por ser un problema unidimensional coincide con $\dfrac{\mathrm{d} z}{\mathrm{d} \xi}$. En capitulos posteriores veremos cómo en problemas bi y tridimensionales el jacobiano es una matriz $2 \times 2$ y $3 \times 3$, respectivamente, cuyo determinante expresa la relación entre los diferenciales de área (en $2D$) y volumen (en $3D$) en cada uno de los sistemas cartesiano y natural.

#### Cálculo del flujo

El flujo se obtiene en función de $\phi$ por

<a id='Ec5_34'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
q=-k g=-k \frac{\mathrm{d} \hat{\phi}}{\mathrm{d} x}=-k \sum_{i=1}^{n} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} x} \phi_{i}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec5_34} \tag{5.34}
\end{equation*}


#### Matriz de rigidez del elemento

Siguiendo el mismo proceso a partir del MRP que para los elementos de dos y tres nodos se obtiene

<a id='Ec5_35'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
K_{i j}^{(e)}=\int_{l^{(e)}} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} x} k \frac{\mathrm{d} N_{j}^{(e)}}{\mathrm{d} x} d x=\int_{-1}^{+1} \frac{1}{j^{(e)}} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} \xi} k \frac{\mathrm{d} N_{j}^{(e)}}{\mathrm{d} \xi} \mathrm{d} \xi
\end{gathered}
\end{aligned}
\label{eq:Ec5_35} \tag{5.35}
\end{equation*}

La mayor o menor sencillez de dicha integral depende de la expresión de las funciones de forma y de $j^{(e)}$. En general, el cálculo de $K_{i j}^{(e)}$ se efectúa por integración numérica con la cuadratura de Gauss-Legendre adecuada al grado de los polinomios que intervienen en el integrado. Así, utilizando una cuadratura de orden $p$

<a id='Ec5_36'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
K_{i j}^{(e)}=\sum_{m=1}^{p}\left[\frac{1}{j(e)} \frac{\mathrm{d} N_{i}^{(e)}}{\mathrm{d} \xi} k \frac{\mathrm{d} N_{j}^{(e)}}{\mathrm{d} \xi}\right]_{\xi_{m}} W_{m}
\end{gathered}
\end{aligned}
\label{eq:Ec5_36} \tag{5.36}
\end{equation*}


#### Vector de fuerzas nodales equivalentes

Para un flujo repartido de intensidad $Q(x)$ se obtiene

<a id='Ec5_37'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
f_{i}^{(e)}=\int_{l^{(e)}} N_{i}^{(e)} Q \mathrm{d} x=\int_{-1}^{+1} N_{i}^{(e)} Q j^{(e)} \mathrm{d} \xi
\end{gathered}
\end{aligned}
\label{eq:Ec5_37} \tag{5.37}
\end{equation*}

De nuevo, para el cálrulo de la integral anterior es útil emplear integración numérica, pudiendo escribirse en general

<a id='Ec5_38'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
f_{i}^{(e)}=\sum_{m=1}^{p}\left[N_{i}^{(e)} Q j^{(e)}\right]_{\xi_{m}} W_{m}
\end{gathered}
\end{aligned}
\label{eq:Ec5_38} \tag{5.38}
\end{equation*}
