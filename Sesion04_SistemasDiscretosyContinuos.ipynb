## Introducción

En el capítulo anterior se realizó un repaso muy general acerca de lo que es el FEM y cómo trabaja. Recordando los pasos a seguir en la resolución de un problema empleando el FEM:

- ***Preprocesamiento:*** Subdivisión del dominio del problema en elementos finitos.


- ***Formulación de los elementos:*** Desarrollo de las ecuaciones a nivel de elementos.


- ***Ensamblaje:*** Obtención de las ecuaciones del sistema completo desde las ecuaciones a nivel de elementos.


- ***Solución de las ecuaciones:*** Empleo de esquemas eficientes para la solución del sistema de ecuaciones lineales resultante.


- ***Posprocesamiento:*** Determinación de las cantidades de interés, tales como tensiones / deformaciones, y obtención de la visualización de las respuestas.

## Conceptos básicos del análisis matricial de estructuras de barras

### Descripción del comportamiento de un único elemento tipo barra

<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C1_F01_BarTruss01.jpg?raw=true" width="500" />
</p>

<div style="text-align: right"> Fuente: <a href="https://tiberiomurciagodoy.blogspot.com/2020/06/puente-pearson-por-hugo-rubio-salas.html">tiberiomurciagody.blogspot</a> </div>

Supongamos se tienen una estructura tipo cercha (como la de la foto) de donde se aisla un elemento tipo barra, con sección transversal arbitraria, y colocado a lo largo de un eje $x$, como se muestra en la [Figura 1](#Fig_1). En este capítulo asumimos que la barra es recta, su material obedece a la [ley de Hooke](https://en.wikipedia.org/wiki/Hooke%27s_law) y que solo puede soportar cargas axiales, es decir, no transmite flexión, cortante o torsión. El [módulo de Young](https://en.wikipedia.org/wiki/Young%27s_modulus) del elemento $e$ se denota por $E^{(e)}$, el área de la sección transversal por $A^{(e)}$ y su longitud por $l^{(e)}$.

<a id='Fig_1'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C1_F02_BarElement03.PNG?raw=true" width="500" />
</p>

<div style="text-align: center"><strong>Figura 1.</strong> Elemento tipo barra sometida a fuerzas axiales. <strong>Fuente:</strong> <a href="https://www.springer.com/gp/book/9781402087424">Oñate, E. Structural Analysis with the FEM</a></div>

    
Debido a las suposiciones realizadas sobre las fuerzas en el elemento, la única fuerza interna distinta de cero es una fuerza interna axial, que es colineal con el eje a lo largo de la barra. La fuerza interna a través de cualquier sección transversal de la barra se denota por $\mathcal{N}^{(e)}$. Se supone que la tensión axial es constante en la sección transversal y está dada por la fuerza interna dividida por el área de la sección transversal:
    
<a id='Ec1_1'></a>
\begin{equation*}
\sigma^{(e)}=\frac{\mathcal{N}^{(e)}}{A^{(e)}}
\label{eq:Ec1_1} \tag{1.1}
\end{equation*}
    
Recuerde que la fuerza axial y la tensión son positivas en tensión y negativas en compresión.
  
Las siguientes ecuaciones gobiernan el comportamiento de la barra:

    
- ***Equilibrio del elemento:*** la sumatoria de las fuerzas nodales internas actuando en el elemento es igual a cero:

<a id='Ec1_2'></a>
\begin{equation*}
F_{x_1}^{(e)}+F_{x_2}^{(e)}=0
\label{eq:Ec1_2} \tag{1.2}
\end{equation*}

- ***Ley de Hooke:*** conocida también como ley [esfuerzo - deformación elástica](https://en.wikipedia.org/wiki/Stress%E2%80%93strain_curve), que establece que la tensión (*stress*) $\sigma^{(e)}$ es una función lineal de la deformación (*strain*) $\varepsilon^{(e)}$

<a id='Ec1_3'></a>
\begin{equation*}
\sigma^{(e)}=E^{(e)} \varepsilon^{(e)}
\label{eq:Ec1_3} \tag{1.3}
\end{equation*}


- ***Compatibilidad de la deformación:*** No pueden existir vacíos o superposiciones en la estructura después de la deformación.


También se requiere de una definición de deformación para aplicar la *ley de Hooke*. La única deformación distinta de cero es la deformación axial $\varepsilon^{(e)}$, que se define como la relación entre el alargamiento $\delta^{(e)}$ y la longitud original del elemento $l^{(e)}$:

<a id='Ec1_4'></a>
\begin{equation*}
\varepsilon^{(e)}=\frac{\delta^{(e)}}{l^{(e)}}=\frac{u_2^{(e)}-u_1^{(e)}}{l^{(e)}}
\label{eq:Ec1_4} \tag{1.4}
\end{equation*}

donde $u_1^{(e)}$ y $u_2^{(e)}$ son los desplazamientos de los extremos $1$ y $2$ de la barra.

Reemplazando la ecuación [(1.4)](#Ec1_4) en la ecuación [(1.3)](#Ec1_3) para relacionar la tensión axial $\sigma$ con la deformación $\varepsilon$ por la [*ley de Hooke*](https://en.wikipedia.org/wiki/Hooke%27s_law):

<a id='Ec1_5'></a>
\begin{equation*}
\sigma=E\varepsilon=E \frac{u_2^{(e)}-u_1^{(e)}}{l^{(e)}}
\label{eq:Ec1_5} \tag{1.5}
\end{equation*}

Despejando el esfuerzo axial $\mathcal{N}^{(e)}$ dado en la ecuación [(1.1)](#Ec1_1) para reemplazar en la ecuación [(1.5)](#Ec1_5), se llega a (suponiendo material homogeneo):

<a id='Ec1_6'></a>
\begin{equation*}
\mathcal{N}^{(e)}=A^{(e)}\sigma=(EA)^{(e)}\frac{u_2^{(e)}-u_1^{(e)}}{l^{(e)}}
\label{eq:Ec1_6} \tag{1.6}
\end{equation*}

El equilibrio de las fuerzas axiales $R_1^{(e)}$ y $R_2^{(e)}$ que actuan en los extremos de la barra es:

<a id='Ec1_7'></a>
\begin{equation*}
-R_1^{(e)}=R_2^{(e)}=\mathcal{N}=(EA)^{(e)}\frac{u_2^{(e)}-u_1^{(e)}}{l^{(e)}}=k^{(e)} \left( u_2^{(e)}-u_1^{(e)} \right)
\label{eq:Ec1_7} \tag{1.7}
\end{equation*}

Reescribiendo la ecuación [(1.7)](#Ec1_7) en forma matricial

<a id='Ec1_8'></a>
\begin{equation*}
\underbrace{\begin{Bmatrix}
R_1^{(e)}  \\
R_2^{(e)}
\end{Bmatrix}}_{\mathbf{F}^{(e)}}=
\underbrace{\begin{bmatrix}
k^{(e)} & -k^{(e)}  \\
-k^{(e)} & k^{(e)} \end{bmatrix}}_{\mathbf{K}^{(e)}}
\underbrace{\begin{Bmatrix}
u_1^{(e)}  \\
u_2^{(e)} 
\end{Bmatrix}}_{\mathbf{d}^{(e)}}
\label{eq:Ec1_8} \tag{1.8}
\end{equation*}

Podemos entonces, escribir la realación entre las fuerzas y los desplazamientos nodales  como

<a id='Ec1_9'></a>
\begin{equation*}
\mathbf{F}^{(e)}=
\mathbf{K}^{(e)}\mathbf{d}^{(e)}
\label{eq:Ec1_9} \tag{1.9}
\end{equation*}

donde $\mathbf{K}^{(e)}$ es la [matriz de rigidez](https://en.wikipedia.org/wiki/Stiffness_matrix) del elemento (barra), que es función de sus parámtros geométricos, y es dada por:

<a id='Ec1_10'></a>
\begin{equation*}
\mathbf{K}^{(e)}=
\begin{bmatrix}
k^{(e)} & -k^{(e)}  \\
-k^{(e)} & k^{(e)} 
\end{bmatrix}=
\frac{A^{(e)}E^{(e)}}{l^{(e)}}
\begin{bmatrix}
1 & -1  \\
-1 & 1 
\end{bmatrix}
\label{eq:Ec1_10} \tag{1.10}
\end{equation*}

Si suponemos que sobre la barra actúa una fuerza uniformemente distribuida por unidad de longitud de intensidad $b^{(e)}$, la ecuación anterior se modifica repartiendo el efecto total de dicha fuerza en partes iguales en cada nodo de la siguiente manera:

<a id='Ec1_11'></a>
\begin{equation*}
\begin{Bmatrix}
R_1^{(e)}  \\
R_2^{(e)}
\end{Bmatrix}=
\begin{bmatrix}
k^{(e)} & -k^{(e)}  \\
-k^{(e)} & k^{(e)} \end{bmatrix}
\begin{Bmatrix}
u_1^{(e)}  \\
u_2^{(e)} 
\end{Bmatrix}-
\frac{(bl)^{(e)}}{2}\begin{Bmatrix}
1  \\
1
\end{Bmatrix}
\label{eq:Ec1_11} \tag{1.11}
\end{equation*}

donde el último término corresponde al vector de fuerzas que actúa en los nodos de la barra debido a la carga distribuida. Esta ecuación se puede escribir de forma compacta como:

<a id='Ec1_12'></a>
\begin{equation*}
\mathbf{F}^{(e)}=
\mathbf{K}^{(e)}\mathbf{d}^{(e)}-\mathbf{f}^{(e)}
\label{eq:Ec1_12} \tag{1.12}
\end{equation*}

***Comentarios:***

- Se puede generalizar el uso de la rigidez de un elemento, dada por la ecuación [(1.12)](#Ec1_12), para cualquier elemento de barra de área constante en una dimensión. Esta universalidad de las matrices de rigidez de elementos es uno de los atributos de FEM que conduce a su versatilidad. 


- Más adelante desarrollaremos matrices de elementos que se aplican a cualquier elemento de mayor dimensión (en 2D: triangular o cuadrilátero, por ejemplo) basado en la solución débil de ecuaciones diferenciales en vez de argumentos físicos.


- La ecuación [(1.11)](#Ec1_11) describe la relación entre las fuerzas nodales y los desplazamientos para un único elemento, es decir, describe el comportamiento de un elemento aislado. 


- Las fuerzas nodales están relacionadas linealmente con los desplazamientos nodales. Esto resulta de la linealidad de todos los componentes que describen el comportamiento de este elemento: la ley de Hooke, la linealidad entre la fuerza axial y la tensión, y la linealidad de la expresión de la deformación. 


- La matriz de rigidez es simétrica, es decir, $\left[ \mathbf{K}^{(e)}\right]=\left[\mathbf{K}^{(e)}\right]^T$

### Etapas del análisis matricial de sistemas discretos

Resumiendo, en el análisis de un sistema discreto están involucradas las siguientes etapas:

1. Definición de la malla (elementos discretos tipo barra).


2. Cálculo de las matrices de rigidez $\mathbf{K}^{(e)}$ y los vectores de fuerzas nodales $\mathbf{f}^{(e)}$ en cada elemento.


3. Ensamblaje y resolución del sistema de equilibrio global $\mathbf{Ka=f}$, para determinar los valores de las incógnitas, que en este caso serían los desplazamientos nodales $\mathbf{a}$


4. A partir de los valores nodales se obtienen otros parámetros de interés (p.ej. tensiones y deformaciones en las barras)

### Ejemplo 1: Desplazamientos y esfuerzos en una estructura de tres barras

<a id='Fig_3'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C1_F06_BarStructure.PNG?raw=true" width="350" />
</p>

<div style="text-align: center"><strong>Figura 3.</strong> Estructura de tres barras. <strong>Fuente:</strong> <a href="https://www.springer.com/gp/book/9781402087424">Oñate, E. Structural Analysis with the FEM</a></div>

***Solución:***
La ecuación de equilibrio en cada barra estará dada por:

\begin{align*}
\text{Barra 1:} \quad
\begin{Bmatrix}
  R_{1}^{(1)}  \\
  R_{2}^{(1)}  \\
\end{Bmatrix}
= k^{(1)}\left[\begin{array}{cc}
  1 & -1 & \\
  -1 & 1 & \\
 \end{array}\right]\begin{Bmatrix}
  u_1^{(1)}  \\
  u_2^{(1)}  \\
\end{Bmatrix}
\end{align*}

\begin{align*}
\text{Barra 2:} \quad
\begin{Bmatrix}
  R_{1}^{(2)}  \\
  R_{2}^{(2)}  \\
\end{Bmatrix}
= k^{(2)}\left[\begin{array}{cc}
  1 & -1 & \\
  -1 & 1 & \\
 \end{array}\right]\begin{Bmatrix}
  u_1^{(2)}  \\
  u_2^{(2)}  \\
\end{Bmatrix}
\end{align*}

\begin{align*}
\text{Barra 3:} \quad
\begin{Bmatrix}
  R_{1}^{(3)}  \\
  R_{2}^{(3)}  \\
\end{Bmatrix}
= k^{(3)}\left[\begin{array}{cc}
  1 & -1 & \\
  -1 & 1 & \\
 \end{array}\right]\begin{Bmatrix}
  u_1^{(3)}  \\
  u_2^{(3)}  \\
\end{Bmatrix}
\end{align*}

con $k^{(1)}=k^{(2)}=\frac{EA}{l}$ y $k^{(3)}=\frac{2EA}{l}$. Las ecuaciones de compatibilidad entre desplazamientos locales y globales en cada nudo son:

$$u_1^{(1)}=u_1; \quad u_2^{(1)}=u_3; \quad u_1^{(2)}=u_2$$
$$u_2^{(2)}=u_3; \quad u_1^{(3)}=u_3; \quad u_2^{(3)}=u_4$$

aplicando la ecuación de ensamblaje a cada uno de los cuatro nodos de la estructura se tiene:

$$\text{nodo 1:} \quad \sum\limits_{e=1}^3 R_i^{(e)}=-R_1$$
$$\text{nodo 2:} \quad \sum\limits_{e=1}^3 R_i^{(e)}=-R_2$$
$$\text{nodo 3:} \quad \sum\limits_{e=1}^3 R_i^{(e)}=0$$
$$\text{nodo 4:} \quad \sum\limits_{e=1}^3 R_i^{(e)}=P$$

Sustituyendo los valores de $R_i^{(e)}$ obtenidos en las ecuaciones de equilibrio de cada barra se llega a las ecuaciones

$$\text{nodo 1:} \quad k^{(1)}\left( u_1^{(1)}-u_2^{(1)}\right)=-R_1$$
$$\text{nodo 2:} \quad k^{(2)}\left( u_1^{(2)}-u_2^{(2)}\right)=-R_2$$
$$\text{nodo 3:} \quad k^{(1)}\left(-u_1^{(1)}+u_2^{(1)}\right)+k^{(2)}\left(-u_1^{(2)}+u_2^{(2)}\right)+k^{(3)}\left(u_1^{(3)}-u_2^{(3)}\right)=0$$
$$\text{nodo 4:} \quad k^{(3)}\left(-u_1^{(3)}+u_2^{(1)}\right)=P$$

Estas ecuaciones pueden ser escritas en forma matricial, empleando las condicions de compatibilidad de desplazamientos como

\begin{equation*}
\begin{bmatrix}
k^{(1)}  &       0     & -k^{(1)} & 0 \\
0           & k^{(2)}  & -k^{(2)} & 0 \\
-k^{(1)} & -k^{(2)} & (k^{(1)}+k^{(2)}+k^{(3)}) & -k^{(3)} \\
0           &       0      & -k^{(3)} & k^{(3)}\end{bmatrix}
\begin{Bmatrix}
u_1 \\
u_2 \\
u_3 \\
u_4
\end{Bmatrix}=
\begin{Bmatrix}
-R_1  \\
-R_2  \\
0     \\
P
\end{Bmatrix}
\end{equation*}

resolviendo el sistema y sustituyendo los valores de las rigideces de cada barra e imponiendo las condiciones de contorno $u_1=u_2=0$ se tiene

$$u_3=\frac{Pl}{2EA}; \quad u_4=\frac{Pl}{EA}; \quad R_1=R_2=\frac{P}{2}$$

y los esfuerzos axiales en cada barra

$$\text{Barra 1:} \quad N^{(1)}=\frac{EA}{l}\left(u_3-u_1\right)=\frac{P}{2}$$
$$\text{Barra 2:} \quad N^{(2)}=\frac{EA}{l}\left(u_3-u_2\right)=\frac{P}{2}$$
$$\text{Barra 3:} \quad N^{(3)}=\frac{2EA}{l}\left(u_4-u_3\right)=P$$

## Matriz de rigidez y vector de fuerzas globales

El ensamblaje de la matriz de rigidez global de la estructura se puede realizar mediante un procedimiento que involucra las contribuciones de rigidez y fuerza de las barras individuales, de la siguiente manera

Sea la barra $(e)$ que conecta dos articulaciones con números globales $i$ y $m$, como se muestra en la figura. Cada término en la posición $(i,m)$ de la matriz de rigidez de la barra contribuye a la misma posición $(i,m)$ de la matriz de rigidez global.


<a id='Fig_4'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C1_F07_BarIsolated.PNG?raw=true" width="350" />
</p>

<div style="text-align: center"><strong>Figura 4.</strong> Barra aislada. <strong>Fuente:</strong> <a href="https://www.springer.com/gp/book/9781402087424">Oñate, E. Structural Analysis with the FEM</a></div>

- $\text{global: i} \quad m$
- $\text{local: 1} \quad 2$
\begin{equation*}
\mathbf{K}^{(e)}=k^{(e)}
\begin{bmatrix}
1  & -1 \\
-1 &  1 \\
\end{bmatrix}
\begin{matrix}
1 & i  \\
2 & m
\end{matrix}
, \quad k^{(e)} = \left( \frac{EA}{l} \right)^{(e)}
\end{equation*}

\begin{matrix}
 & i & &  & &  & & m & & & &  & &  & &
\end{matrix}

\begin{equation*}
\begin{matrix}
i  \\
   \\
m
\end{matrix}
\begin{bmatrix}
      &     |      &  & | &               &       \\
-- & \left[k^{(e)}\right] & -- & \left[-k^{(e)}\right]  & -- \\
      &     |      &  & | &               &       \\
-- & \left[-k^{(e)}\right] & -- & \left[k^{(e)}\right]  & -- \\
      &     |      &  & | &               &       
\end{bmatrix}
\begin{Bmatrix}
      \\
u_i  \\
      \\
u_m\\
      \\
\end{Bmatrix}
=
\begin{Bmatrix}
      \\
R_i  \\
      \\
R_m\\
      \\
\end{Bmatrix}
\end{equation*}

La matriz de rigidez global de puede deducir siguiendo una regla general aprtiendo del ensamblaje que se puede hacer empezando por la contribución de las rigideces de sus elementos individuales. Tomando una barra $(e)$ conectada a los nodos con numeración global $i$ y $m$, cada elemento $(i,m)$ de la matriz de rigidez del elemento ocupa la misma posición $(i,m)$ en la matriz de rigidez global. Para ensamblar la matriz de rigidez global se van añadiendo directamente los coeficientes de rigidez de cada elemento. Este mismo procedimiento se emplea para ensamblar el vector global de fuerzas nodales.

### Ejemplo: Obtención del ancho de banda de la matriz de rigidez global

Determinar el ancho de banda de las matrices de rigidez resultantes de las estructuras mostradas con las numeraciones de los nodos dadas.

<p float="center">

\begin{bmatrix}
k_{11}^{(1)} &                      k_{12}^{(1)}                        &       0      &      0       \\

      0      &                      k_{12}^{(2)}                        & k_{11}^{(2)} &      0       \\
      0      &                      k_{21}^{(3)}                        &       0      &  k_{22}^{(3)}
\end{bmatrix}
\end{equation*}

- ***Numeración b:***

<a id='Fig_6'></a>

\begin{bmatrix}

      0      &       0      & k_{11}^{(2)} & k_{12}^{(2)} \\
k_{21}^{(1)} &       0      & k_{21}^{(2)} & \left(k_{11}^{(1)} + k_{22}^{(2)} + k_{11}^{(3)} \right)
