## Problemas de elasticidad lineal

### Introducción

La [Teoría de la Elasticidad](https://en.wikipedia.org/wiki/Elasticity_(physics)) determina el comportamiento de una estructura ante las cargas mediante la resolución de sistemas de ecuaciones diferenciales, llamados ecuaciones de campo o de gobierno, complementados por las condiciones de contorno o de sustentación.

Los métodos tradicionales de cálculo de estructuras utilizan la [Resistencia de Materiales](https://en.wikipedia.org/wiki/Strength_of_materials) que añade hipótesis simplificatorias a la Teoría de la Elasticidad para resolver las ecuaciones de gobierno, obteniéndose unas soluciones que se consideran exactas.

Sin embargo, estos métodos tradicionales de cálculo normalmente solo se pueden aplicar a estructuras [articuladas y reticuladas](https://en.wikipedia.org/wiki/Truss). En estructuras mas complejas como las estructuras continuas es necesario aplicar métodos aproximados como el Método de los elementos finitos. Después de dividir la estructura en partes, llamadas elementos finitos, interconectadas entre sí, el método de los elementos finitos reduce el problema elástico a la resolución de un sistema de ecuaciones algebraicas. Este método se basa en una serie de conceptos que se agrupan en los apartados siguientes.

$I$. Conceptos de Elasticidad.

$II$. Conceptos energéticos.

$III$. Conceptos matemáticos.

### Ecuaciones de equilibrio

<a id='Fig11_1a'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C11_F01a_Equilibrio.png?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 11.1a.</strong>Equilibrio de un cuerpo, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Sea un cuerpo sometido a un sistema de fuerzas aplicadas cualesquiera [Fig. 11_1a](#Fig11_1a) y sea $O$ el punto del cuerpo cuyos estados de esfuerzos y de deformaciones estamos considerando. Eligiendo el punto $O$ como origen del triedro de ejes coordenados $x, y, z$, el estado de esfuerzos en el punto $O$ esta definido por el vector de esfuerzos

<a id='Ec11_1'></a>
\begin{equation*}
\begin{aligned}
\{\boldsymbol{\sigma}\}=\left[\begin{array}{llllll}
\sigma_{x} & \sigma_{y} & \sigma_{z} & \tau_{x y} & \tau_{y z} & \tau_{z x}
\end{array}\right]^{T}
\end{aligned}
\label{eq:Ec11_1} \tag{11.1}
\end{equation*}

formado por los esfuerzos normales $\sigma_{x}, \sigma_{y}, \sigma_{z} \mathrm{y}$ los esfuerzos cortantes $\tau_{x y}, \tau_{y z}, \tau_{z x}$ [Fig. 11_1b](#Fig11_1b). Estas seis componentes constituyen las componentes del estado de esfuerzos en el punto considerado.

<a id='Fig11_1b'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C11_F01b_EsfuerzosNormalesyCortantes.png?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 11.1b.</strong>Esfuerzos normales y cortantes, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

A continuación se aísla del cuerpo un elemento de volumen diferencial en forma de paralelepípedo recto con vértice en el punto $O$, origen de un sistema de ejes coordenados coincidentes con las aristas del paralelepípedo. Cuando se reducen las dimensiones del paralelepípedo, manteniéndolo semejante a sí mismo, el paralelepípedo tiende al punto $O$.

El paralelepípedo elemental [Fig. 11_2](#Fig11_2) estará en equilibrio bajo la acción de las fuerzas que actúan sobre cada cara y de las fuerzas de volumen q $d x d y d z$, tal como el peso del paralelepípedo. Estableciendo las ecuaciones de la Estática $\Sigma X=0, \Sigma Y=0, \Sigma Z=0$, se obtienen las *ecuaciones de equilibrio* del paralelepípedo

<a id='Fig11_2'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C11_F02_ParalelepipedoElemental.png?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 11.2.</strong>Paralelepípedo elemental, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

<a id='Ec11_2'></a>
\begin{equation*}
\begin{aligned}
&\frac{\partial \sigma_{x}}{\partial x}+\frac{\partial \tau_{y x}}{\partial y}+\frac{\partial \tau_{z x}}{\partial z}+q_{x}=0 \\
&\frac{\partial \tau_{x y}}{\partial x}+\frac{\partial \sigma_{y}}{\partial y}+\frac{\partial \tau_{z y}}{\partial z}+q_{y}=0 \\
&\frac{\partial \tau_{x z}}{\partial x}+\frac{\partial \tau_{y z}}{\partial y}+\frac{\partial \sigma_{z}}{\partial z}+q_{z}=0
\end{aligned}
\label{eq:Ec11_2} \tag{11.2}
\end{equation*}

Las ecuaciones Estática $\Sigma M_{x}=0, \Sigma M_{y}=0, \Sigma M_{z}=0$, permiten determinar las relaciones

<a id='Ec11_3'></a>
\begin{equation*}
\begin{aligned}
\tau_{y z}=\tau_{z y}, \quad \tau_{z x}=\tau_{x z}, \quad \tau_{x y}=\tau_{y x}
\end{aligned}
\label{eq:Ec11_3} \tag{11.3}
\end{equation*}

que son la expresión analítica del [teorema de Cauchy](https://en.wikipedia.org/wiki/Cauchy_stress_tensor). Este teorema reduce a 6 las componentes independientes de esfuerzos que son las componentes que intervienen en el vector de esfuerzos $\{\boldsymbol{\sigma}\}$.


### Relaciones esfuerzos-deformaciones

Asimismo, el estado de deformaciones en el punto $\mathrm{O}$ [Fig. 11_3](#Fig11_3) está definido por el vector de deformaciones

<a id='Ec11_4'></a>
\begin{equation*}
\begin{aligned}
\{\boldsymbol{\varepsilon}\}=\left[\begin{array}{llllll}
\varepsilon_{x} & \varepsilon_{y} & \varepsilon_{z} & \gamma_{x y} & \gamma_{y z} & \gamma_{z x}
\end{array}\right]^{T}
\end{aligned}
\label{eq:Ec11_4} \tag{11.4}
\end{equation*}

formado por las deformaciones lineales $\varepsilon_{x}, \varepsilon_{y}, \varepsilon_{z}$ y las deformaciones angulares $\gamma_{x y}, \gamma_{y z}, \gamma_{z x}$, que constituyen las *componentes del estado de deformaciones* en el punto.

Cuando el cuerpo es de un material isótropo y linealmente elástico satisface la [ley de Hooke](https://en.wikipedia.org/wiki/Hooke%27s_law) generalizada que determina las componentes del estado de deformaciones en función de las componentes del estado de esfuerzos

<a id='Fig11_3'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C11_F03_EstadoDeformaciones.png?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 11.3.</strong>Paralelepípedo elemental, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

<a id='Ec11_5'></a>
\begin{equation*}
\begin{aligned}
&\varepsilon_{x}=\frac{\sigma_{x}}{E}-\frac{\nu}{E}\left(\sigma_{y}+\sigma_{z}\right) \\
&\varepsilon_{y}=\frac{\sigma_{y}}{E}-\frac{\nu}{E}\left(\sigma_{z}+\sigma_{x}\right) \\
&\varepsilon_{z}=\frac{\sigma_{x}}{E}-\frac{\nu}{E}\left(\sigma_{y}+\sigma_{z}\right) \\
&\gamma_{x y}=\frac{\tau_{x y}}{G} \\
&\gamma_{y z}=\frac{\tau_{y z}}{G} \\
&\gamma_{z x}=\frac{\tau_{z x}}{G}
\end{aligned}
\label{eq:Ec11_5} \tag{11.5}
\end{equation*}

Al ser $E / G=2(1+\nu)$, las anteriores relaciones tienen solamente dos coeficientes independientes. De las [Ecs. 11.5](#Ec11_5) se deducen las [ecuaciones de Lamé](https://en.wikipedia.org/wiki/Lam%C3%A9_parameters) que determinan las componentes del estado de esfuerzos en función de las componentes del estado de deformaciones

<a id='Ec11_6'></a>
\begin{equation*}
\begin{aligned}
&\sigma_{x}=\lambda\left(\varepsilon_{x}+\varepsilon_{y}+\varepsilon_{z}\right)+2 \mu \varepsilon_{x} \\
&\sigma_{y}=\lambda\left(\varepsilon_{x}+\varepsilon_{y}+\varepsilon_{z}\right)+2 \mu \varepsilon_{y} \\
&\sigma_{z}=\lambda\left(\varepsilon_{x}+\varepsilon_{y}+\varepsilon_{z}\right)+2 \mu \varepsilon_{z} \\
&\tau_{x y}=G \gamma_{x y} \\
&\tau_{y z}=G \gamma_{y z} \\
&\tau_{z x}=G \gamma_{z x}
\end{aligned}
\label{eq:Ec11_6} \tag{11.6}
\end{equation*}

siendo $\lambda$ y $\mu$ los [coeficientes de Lamé](https://en.wikipedia.org/wiki/Lam%C3%A9_parameters)

$$
\lambda=\frac{\nu E}{(1+\nu)(1-2 \nu)}, \quad \mu=\frac{E}{2(1+\nu)}=G
$$

Las [Ecs. 11.6](#Ec11_6) pueden expresarse en la forma matricial

<a id='Ec11_7'></a>
\begin{equation*}
\begin{aligned}
\{\boldsymbol{\sigma}\}=[\mathbf{D}]\{\boldsymbol{\varepsilon}\}
\end{aligned}
\label{eq:Ec11_7} \tag{11.7}
\end{equation*}

donde $[\mathbf{D}]$ es la *matriz esfuerzos-deformaciones* o *matriz constitutiva*

<a id='Ec11_8'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{D}\right]=
\frac{E}{(1+\nu)(1-2\nu)}
\begin{bmatrix}
(1-\nu) &   \nu   &   \nu   &      0            &    0              &         0          \\
\nu     & (1-\nu) &   \nu   &      0            &    0              &         0          \\ 
\nu     &    \nu  & (1-\nu) &      0            &    0              &         0          \\ 
0       &    0    &    0    &\frac{(1-2\nu)}{2} &    0              &         0          \\ 
0       &    0    &    0    &      0            &\frac{(1-2\nu)}{2} &         0          \\
0       &    0    &    0    &      0            &    0              & \frac{(1-2\nu)}{2} \\
\end{bmatrix}
\end{aligned}
\label{eq:Ec11_8} \tag{11.8}
\end{equation*}

que es una matriz simétrica, función del [módulo de elasticidad](https://en.wikipedia.org/wiki/Young%27s_modulus) y del [coeficiente de Poissón](https://en.wikipedia.org/wiki/Poisson%27s_ratio) del material. En el caso mas general de [materiales anisótropos](https://en.wikipedia.org/wiki/Anisotropy#Materials_science_and_engineering), la matriz constitutiva $[\mathbf{D}]$ contiene 21 coeficientes independientes.

Cuando en un cuerpo existen inicialmente esfuerzos $\left\{\boldsymbol{\sigma}_{0}\right\}$ y deformaciones $\left\{\boldsymbol{\varepsilon}_{0}\right\}$ que son anteriores a la aplicación de las cargas, la [Ec. 11.7](#Ec11_7) se amplia a

<a id='Ec11_9'></a>
\begin{equation*}
\begin{aligned}
\{\boldsymbol{\sigma}\}=[\mathbf{D}]\left(\{\boldsymbol{\varepsilon}\}-\left\{\boldsymbol{\varepsilon}_{0}\right\}\right)+\left\{\boldsymbol{\sigma}_{0}\right\}
\end{aligned}
\label{eq:Ec11_9} \tag{11.9}
\end{equation*}

En un material isótropo, una variación uniforme de temperatura $\Delta t$ origina las deformaciones iniciales

<a id='Ec11_10'></a>
\begin{equation*}
\begin{aligned}
\left\{\boldsymbol{\varepsilon}_{0}\right\}=\left[\begin{array}{llllll}
\alpha \Delta t & \alpha \Delta t & \alpha \Delta t & 0 & 0 & 0
\end{array}\right]^{T}
\end{aligned}
\label{eq:Ec11_10} \tag{11.10}
\end{equation*}

siendo $\alpha$ el coeficiente de dilatación lineal del material.


### Relaciones deformaciones-desplazamientos

Entre las componentes del estado de deformaciones en un punto y las componentes del desplazamiento de ese punto, $u(x, y, z), v(x, y, z), w(x, y, z)$, existen las relaciones de compatibilidad de las deformaciones

<a id='Ec11_11'></a>
\begin{equation*}
\begin{aligned}
&\varepsilon_{x}=\frac{\partial u}{\partial x}, \quad \varepsilon_{y}=\frac{\partial v}{\partial y}, \quad \varepsilon_{z}=\frac{\partial w}{\partial z} \\
&\gamma_{x y}=\frac{\partial u}{\partial y}+\frac{\partial v}{\partial x}, \quad \gamma_{y z}=\frac{\partial v}{\partial z}+\frac{\partial w}{\partial y}, \quad \gamma_{z x}=\frac{\partial w}{\partial x}+\frac{\partial u}{\partial z} .
\end{aligned}
\label{eq:Ec11_11} \tag{11.11}
\end{equation*}

Estas relaciones se pueden expresar matricialmente en la forma

<a id='Ec11_12'></a>
\begin{equation*}
\begin{aligned}
\left|\begin{array}{c}
\varepsilon_{x} \\
\varepsilon_{y} \\
\varepsilon_{z} \\
\gamma_{x y} \\
\gamma_{y z} \\
\gamma_{z x}
\end{array}\right|=\left|\begin{array}{ccc}
\frac{\partial}{\partial x} & 0 & 0 \\
0 & \frac{\partial}{\partial y} & 0 \\
0 & 0 & \frac{\partial}{\partial z} \\
\frac{\partial}{\partial y} & \frac{\partial}{\partial x} & 0 \\
0 & \frac{\partial}{\partial z} & \frac{\partial}{\partial y} \\
\frac{\partial}{\partial z} & 0 & \frac{\partial}{\partial x}
\end{array}\right|\left|\begin{array}{c}
u \\
v \\
w
\end{array}\right|
\end{aligned}
\label{eq:Ec11_12} \tag{11.12}
\end{equation*}

es decir

<a id='Ec11_13'></a>
\begin{equation*}
\begin{aligned}
\{\varepsilon\}=[\partial]\{\mathbf{u}\}
\end{aligned}
\label{eq:Ec11_13} \tag{11.13}
\end{equation*}

siendo $[\partial]$ el operador matricial

<a id='Ec11_14'></a>
\begin{equation*}
\begin{aligned}
[\boldsymbol{\partial}]=\left|\begin{array}{ccc}
\frac{\partial}{\partial x} & 0 & 0 \\
0 & \frac{\partial}{\partial y} & 0 \\
0 & 0 & \frac{\partial}{\partial z} \\
\frac{\partial}{\partial y} & \frac{\partial}{\partial x} & 0 \\
0 & \frac{\partial}{\partial z} & \frac{\partial}{\partial y} \\
\frac{\partial}{\partial z} & 0 & \frac{\partial}{\partial x}
\end{array}\right|
\end{aligned}
\label{eq:Ec11_14} \tag{11.14}
\end{equation*}

formado por operadores en derivadas parciales y $\{\mathbf{u}\}$ el vector de desplazamientos

<a id='Ec11_15'></a>
\begin{equation*}
\begin{aligned}
\{\mathbf{u}\}=\left|\begin{array}{c}
u \\
v \\
w
\end{array}\right|
\end{aligned}
\label{eq:Ec11_15} \tag{11.15}
\end{equation*}


### Problema elástico

El análisis del comportamiento de una estructura frente a las fuerzas aplicadas lleva consigo la resolución del problema elástico que consiste en determinar en cada uno de los puntos de la estructura las 15 magnitudes siguientes:

- $6$ componentes del estado esfuerzos: $\sigma_{x}, \sigma_{y}, \sigma_{z}, \tau_{x y}, \tau_{y z}, \tau_{z x}$

- $6$ componentes del estado de deformaciones: $\varepsilon_{x}, \varepsilon_{y}, \varepsilon_{z}, \gamma_{x y}, \gamma_{y z}, \gamma_{z x}$

- $3$ componentes del desplazamiento: $u, v, w$

Estas 15 magnitudes están relacionadas entre sí mediante 15 ecuaciones que son el fundamento de la Teoría de la Elasticidad:

- $3$ ecuaciones de equilibrio [(11.2)](#Ec11_2)

- $6$ ecuaciones que relacionan esfuerzos y deformaciones [(11.6)](#Ec11_6)

- $6$ ecuaciones de compatibilidad de las deformaciones [(11.11)](#Ec11_11)

Al resolver este sistema de 15 ecuaciones en derivadas parciales hay que tener en cuenta las condiciones de contorno que comprenden dos tipos de prescripciones:

$a)$ desplazamientos conocidos debido a la actuación de los enlaces.

$b)$ esfuerzos conocidos originados en las proximidades de las fuerzas aplicadas.

A veces interesa expresar las ecuaciones de equilibrio en función de los desplazamientos y a las ecuaciones diferenciales resultantes se las denomina ecuaciones diferenciales de gobierno.

La solución del problema elástico es única y solamente puede hallarse de forma exacta en casos particulares mediante la Teoría de la Elasticidad. Por el contrario, la Resistencia de Materiales resuelve el problema elástico de forma más simple utilizando, usualmente, métodos aproximados a partir de hipótesis simplificadoras.

La unicidad de la solución es una consecuencia de que la energía de deformación es siempre positiva cuando existen fuerzas aplicadas sobre la estructura. En efecto, supongamos que a un sistema de fuerzas aplicadas $\{\mathbf{P}\}$ le corresponde en un punto $O$ de la estructura un estado de esfuerzos $\sigma_{x}, \sigma_{y}, \sigma_{z}, \tau_{x y}$, $\tau_{y z}, \tau_{z x}$ y que al mismo sistema de fuerzas $\{\mathbf{P}\}$ le corresponde, en el mismo punto $O$, otro estado de esfuerzos diferente $\sigma_{x}^{\prime}, \sigma_{y}^{\prime}, \sigma_{z}^{\prime}, \tau_{x y}^{\prime}, \tau_{y z}^{\prime}, \tau_{z x}^{\prime} .$ Aplicando el principio de superposición, a un estado de fuerzas aplicadas $\{\mathbf{P}\}-\{\mathbf{P}\}=0$ le correspondería en el punto $O$ un estado de esfuerzos no nulo

$$
\sigma_{x}-\sigma_{x}^{\prime}, \quad \sigma_{y}-\sigma_{y y}^{\prime}, \quad \sigma_{z}-\sigma_{z}^{\prime}, \quad \tau_{x y}-\tau_{x y}^{\prime}, \quad \tau_{y z}-\tau_{y z}^{\prime}, \quad \tau_{z x}-\tau_{z x}^{\prime}
$$

y, por tanto, una energía de deformación positiva. Tendríamos así un sistema de fuerzas nulo que aplicado sobre la estructura realizaría un trabajo nulo y en cambio la estructura almacenaría una energía de deformación positiva. Ello estaría en contradicción con el principio de conservación de la energía y, por consiguiente, la solución del problema elástico tiene que ser única.

Cuando no existen fuerzas aplicadas sobre la estructura su energía de deformación es nula y son también nulos los esfuerzos en todos los puntos de la estructura. Sin embargo, hay casos en que pueden existir esfuerzos iniciales sin que haya fuerzas aplicadas sobre la estructura. En estos casos, si el principio de superposición es aplicable, los esfuerzos finales en el cuerpo elástico se hallan sumando los esfuerzos iniciales y los esfuerzos debidos a las fuerzas aplicadas que se obtienen resolviendo el problema elástico.


# CONCEPTOS ENERGÉTICOS

### Teorema de los Trabajos Virtuales

El [teorema, o principio, de los trabajos virtuales](https://en.wikipedia.org/wiki/Virtual_work) establece que la condición necesaria y suficiente para que una estructura esté en equilibrio es que sea nula la suma de los trabajos que realizan todas las fuerzas que actúan en la estructura para cualquier conjunto de desplazamientos y deformaciones virtuales compatibles con sus enlaces (Fig. 1.4). Estos trabajos comprenden el trabajo que realizan las fuerzas externas a la estructura $W_{e}$ y el trabajo que realizan sus fuerzas internas $W_{i}$. Por consiguiente

<a id='Ec11_16'></a>
\begin{equation*}
\begin{aligned}
W_{e}+W_{i}=0
\end{aligned}
\label{eq:Ec11_16} \tag{11.16}
\end{equation*}

El trabajo que realizan las fuerzas externas $\left\{\mathbf{F}_{e}\right\}$ cuando se producen unos desplazamientos virtuales $\left\{\delta_{e}^{*}\right\}$ compatibles con los enlaces es

<a id='Ec11_17'></a>
\begin{equation*}
\begin{aligned}
W_{e}=\Sigma F_{j} \cdot \delta_{j}^{*}=\left\{\delta_{e}^{*}\right\}^{T}\left\{\mathbf{F}_{e}\right\}
\end{aligned}
\label{eq:Ec11_17} \tag{11.17}
\end{equation*}

<a id='Fig11_4'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C11_F04_VirtualWork.png?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 11.4.</strong>Principio de los trabajos virtuales, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Las fuerzas internas de un paralelepípedo con vértice en uno de sus puntos son iguales $y$ de sentido contrario a las fuerzas externas que actúan sobre ese paralelepípedo, que son las originadas por el vector esfuerzos $\{\boldsymbol{\sigma}\}$ en la [Fig. 11.4](#Fig11_4). ese punto. En consecuencia, el trabajo elemental $\mathrm{d} W_{i}$ que realizan las fuerzas internas de un paralelepípedo elemental cuando se producen unas deformaciones virtuales $\left\{\boldsymbol{\varepsilon}^{*}\right\}$ compatibles con las desplazamientos virtuales $\left\{\delta^{*}\right\}$ es igual y de signo contrario a la suma de los trabajos realizados por las fuerzas originadas por los esfuerzos normales

\begin{equation*}
\begin{aligned}
-\left(\sigma_{x} \mathrm{d} y \mathrm{d} z\right) \varepsilon_{x}^{*} \mathrm{d} x-\left(\sigma_{y} \mathrm{d} z \mathrm{d} x\right) \varepsilon_{y}^{*} \mathrm{d} y-\left(\sigma_{z} \mathrm{d} x \mathrm{d} y\right) \varepsilon_{z}^{*} \mathrm{d} z
\end{aligned}
\end{equation*}

y por las fuerzas originadas por los esfuerzos cortantes

\begin{equation*}
\begin{aligned}
-\left(\tau_{x y} \mathrm{d} x \mathrm{d} y\right) \gamma_{x y} \mathrm{d} z-\left(\tau_{y z} \mathrm{d} y \mathrm{d} z\right) \gamma_{y z} \mathrm{d} x-\left(\tau_{z x} \mathrm{d} z \mathrm{d} x\right) \gamma_{z x} \mathrm{d} y
\end{aligned}
\end{equation*}

Es decir

\begin{equation*}
\begin{aligned}
\mathrm{d} W_{i}=-\left\{\boldsymbol{\varepsilon}^{*}\right\}^{T}\{\boldsymbol{\sigma}\} \mathrm{d} V
\end{aligned}
\end{equation*}

y, por tanto

<a id='Ec11_18'></a>
\begin{equation*}
\begin{aligned}
W_{i}=-\int_{V}\left\{\boldsymbol{\varepsilon}^{*}\right\}^{T}\{\boldsymbol{\sigma}\} \mathrm{d} V
\end{aligned}
\label{eq:Ec11_18} \tag{11.18}
\end{equation*}

Sustituyendo las ecuaciones [(11.17)](#Ec11_17) y [(11.18)](#Ec11_18) en [(11.16)](#Ec11_16), se obtiene

<a id='Ec11_19'></a>
\begin{equation*}
\begin{aligned}
\left\{\delta_{e}^{*}\right\}^{T}\left\{\mathbf{F}_{e}\right\}-\int_{V}\left\{\boldsymbol{\varepsilon}^{*}\right\}^{T}\{\boldsymbol{\sigma}\} \mathrm{d} V=0
\end{aligned}
\label{eq:Ec11_19} \tag{11.19}
\end{equation*}

que es la expresión analítica del teorema de los trabajos virtuales.

Este teorema se puede utilizar para calcular la fuerza $\mathbf{F}_{j}$ a aplicar en un punto determinado cuando se conoce la distribución de esfuerzos reales $\{\boldsymbol{\sigma}\}$ de la estructura. Para ello consideramos un único desplazamiento virtual $\delta_{j}^{*}$ manteniendo sin modificar los desplazamientos de los puntos de aplicación de las otras fuerzas de la estructura. En este caso la ecuación [(11.19)](#Ec11_19) se reduce a

<a id='Ec11_20'></a>
\begin{equation*}
\begin{aligned}
F_{j} \cdot \delta_{j}^{*}=\int_{V}\left\{\varepsilon^{*}\right\}^{T}\{\boldsymbol{\sigma}\} \mathrm{d} V
\end{aligned}
\label{eq:Ec11_20} \tag{11.20}
\end{equation*}

Cuando la estructura es linealmente elástica, las deformaciones virtuales $\left\{\boldsymbol{\varepsilon}^{*}\right\}$ correspondientes al desplazamiento virtual único $\delta_{j}^{*}$ son proporcionales a las deformaciones virtuales $\left\{\varepsilon_{u}^{*}\right\}$ correspondientes al desplazamiento virtual único $\delta_{j}^{*}=1$, verificándose

$$
\left\{\boldsymbol{\varepsilon}^{*}\right\}=\left\{\boldsymbol{\varepsilon}_{u}^{*}\right\} \delta_{j}^{*}
$$

que sustituida en [(11.20)](#Ec11_20), determina la ecuación

<a id='Ec11_21'></a>
\begin{equation*}
\begin{aligned}
F_{j}=\int_{V}\left\{\boldsymbol{\varepsilon}_{u}^{*}\right\}^{T}\{\boldsymbol{\sigma}\} \mathrm{d} V,
\end{aligned}
\label{eq:Ec11_21} \tag{11.21}
\end{equation*}

que representa el teorema del desplazamiento virtual unitario que se utiliza para determinar los coeficientes de rigidez de elementos estructurales bidimensionales y tridimensionales.


### Principio de la Energía Potencial Total

La suma de la energía potencial $V$ de las fuerzas externas que actúan sobre una estructura y de su energía de deformación $U$ se denomina energía potencial total $\phi$ de la estructura. Es decir,

<a id='Ec11_22'></a>
\begin{equation*}
\begin{aligned}
\phi=V+U
\end{aligned}
\label{eq:Ec11_22} \tag{11.22}
\end{equation*}

Considerando que la energía potencial de las fuerzas externas es cero cuando la estructura no se ha deformado, la energía potencial de las fuerzas externas en la estructura deformada es

<a id='Ec11_23'></a>
\begin{equation*}
\begin{aligned}
V=-\Sigma F_{j} \cdot \delta_{j}=-\left\{\delta_{e}\right\}^{T}\left\{\mathbf{F}_{e}\right\}
\end{aligned}
\label{eq:Ec11_23} \tag{11.23}
\end{equation*}

que equivale a la capacidad que tienen las fuerzas externas de realizar trabajo durante los desplazamientos, manteniéndose las fuerzas externas constantes e iguales a sus valores finales.

Si $\{\boldsymbol{\sigma}\}$ es el vector de esfuerzos y $\{\boldsymbol{\varepsilon}\}$ el vector de deformaciones correspondientes a un punto interior de la estructura, la energía de deformación de la estructura es

$$
U=\frac{1}{2} \int_{V}\left(\sigma_{x} \varepsilon_{x}+\sigma_{y} \varepsilon_{y}+\sigma_{z} \varepsilon_{z}+\tau_{x y} \gamma_{x y}+\tau_{y z} \gamma_{y z}+\tau_{z x} \gamma_{z x}\right) \mathrm{d} V
$$

o sea

<a id='Ec11_24'></a>
\begin{equation*}
\begin{aligned}
U=\frac{1}{2} \int_{V}\{\varepsilon\}^{T}\{\boldsymbol{\sigma}\} \mathrm{d} V
\end{aligned}
\label{eq:Ec11_24} \tag{11.24}
\end{equation*}

En consecuencia, la energía potencial total $\phi$ de la estructura es

<a id='Ec11_25'></a>
\begin{equation*}
\begin{aligned}
\phi=-\left\{\delta_{e}\right\}^{T}\left\{\mathbf{F}_{e}\right\}+\frac{1}{2} \int_{V}\{\boldsymbol{\varepsilon}\}^{T}\{\boldsymbol{\sigma}\} \mathrm{d} V
\end{aligned}
\label{eq:Ec11_25} \tag{11.25}
\end{equation*}

Consideremos ahora que la estructura experimenta, a partir de la configuración de equilibrio, unos desplazamientos virtuales $\left\{\delta_{e}^{*}\right\}$ y unas deformaciones también virtuales $\left\{\varepsilon^{*}\right\}$ compatibles con los enlaces adquiriendo una configuración diferente a la del equilibrio inicial. Estos desplazamientos y deformaciones virtuales producen una variación de la energía potencial total, definida por

$$
\mathrm{d} \phi=-\left\{\delta_{e}^{*}\right\}^{T}\left\{\mathbf{F}_{e}\right\}+\int_{V}\{\boldsymbol{\sigma}\}^{T}\left\{\boldsymbol{\varepsilon}^{*}\right\} \mathrm{d} V
$$

ya que tanto las fuerzas $\left\{\mathbf{F}_{e}\right\}$ como los esfuerzos $\{\boldsymbol{\sigma}\}$ permanecen constantes durante este proceso de cambio en la configuración de la estructura.

El teorema de los trabajos virtuales, ecuación (1.19), exige que se verifique

<a id='Ec11_26'></a>
\begin{equation*}
\begin{aligned}
\mathrm{d} \phi=0
\end{aligned}
\label{eq:Ec11_26} \tag{11.26}
\end{equation*}

según la cual, cuando se somete a una estructura a unos desplazamientos y deformaciones virtuales compatibles con los enlaces, la variación de la energía potencial total es nula.

Este principio de la energía potencial total permite conocer los desplazamientos de una estructura en equilibrio y, por tanto, la configuración de la estructura deformada, cuando esta configuración depende únicamente de los desplazamientos $\{\delta\}$ de $n$ puntos determinados de la estructura. Para ello se le da a uno de esos puntos un desplazamiento diferencial virtual $d \delta_{i}^{*}$ sin modificar los $n-1$ desplazamientos restantes. Como la modificación de la energía potencial total ha de ser nula, resulta

<a id='Ec11_27'></a>
\begin{equation*}
\begin{aligned}
\frac{\partial \phi}{\partial \delta_{i}}=0
\end{aligned}
\label{eq:Ec11_27} \tag{11.27}
\end{equation*}

Haciendo $i=1,2, \ldots, n$ se obtiene un sistema de $n$ ecuaciones que permiten calcular los $n$ desplazamientos $\{\delta\}$ y con ello la configuración de la estructura deformada.

La ecuación [(11.27)](#Ec11_27) es la expresión analítica del principio de la energía potencial total, según el cual, de todas las posibles configuraciones de una estructura deformada compatibles con los enlaces, a la configuración de equilibrio le corresponde una energía potencial estacionaria. Esta energía potencial estacionaria es mínima o máxima según que la configuración sea de equilibrio estable o inestable, respectivamente. Este principio es válido tanto para estructuras lineales como no lineales.
