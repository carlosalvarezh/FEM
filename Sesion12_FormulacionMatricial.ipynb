## Formulación Matricial

### Introducción

En capítulos anteriores hemos estudiado la solución de la ecuación de Poisson en 1D, 2D y 3D por el MEF. Es muy interesante e instructivo plantear la solución general de dicha ecuación por el MEF utilizando una formulación matricial. El planteamiento matricial presenta además la ventaja de que es generalizable a la solución de otros muchos problemas por ì MEF.


### Planteamiento matricial de la ecuación de Poisson

La ecuación de balance de flujos en un dominio infinitesimal en una, dos o tres dimensiones puede expresarse matricialmente por

<a id='Ec9_1'></a>
\begin{equation*}
\begin{aligned}
\nabla^{T} \mathbf{q}-Q=0
\end{aligned}
\label{eq:Ec9_1} \tag{9.1}
\end{equation*}

siendo $\nabla$ el [vector gradiente](https://en.wikipedia.org/wiki/Gradient) y $q$ el vector de flujos que se escriben como

<a id='Ec9_2'></a>
\begin{equation*}
\begin{aligned}
\begin{array}{ll}
\text { 1D: } \quad \nabla & =\left[\frac{\partial}{\partial x}\right], \quad q=[q] \\
\text { 2D: } \quad \nabla & =\left[\frac{\partial}{\partial x}, \frac{\partial}{\partial y}\right]^{T}, \quad q=\left[q_{x}, q_{y}\right]^{T} \\
\text { 3D: } \quad \nabla & =\left[\frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z}\right]^{T}, \quad q=\left[q_{x}, q_{y}, q_{z}\right]^{T}
\end{array}
\end{aligned}
\label{eq:Ec9_2} \tag{9.2}
\end{equation*}

Puede comprobarse que la [Ec. 9.1](#Ec9_1) en forma desarrollada coincide con la forma del balance de flujos obtenida en los capítulos anteriores. Así, en $2D$

<a id='Ec9_3'></a>
\begin{equation*}
\begin{aligned}
\left[\frac{\partial}{\partial x}, \frac{\partial}{\partial y}\right]\left\{\begin{array}{l}
q_{x} \\
q_{y}
\end{array}\right\}-Q=\frac{\partial q_{x}}{\partial x}+\frac{\partial q_{y}}{\partial y}-Q=0
\end{aligned}
\label{eq:Ec9_3} \tag{9.3}
\end{equation*}

El vector de gradientes se define en forma general, por

<a id='Ec9_4'></a>
\begin{equation*}
\begin{aligned}
\mathbf{g}=\nabla \phi
\end{aligned}
\label{eq:Ec9_4} \tag{9.4}
\end{equation*}

siendo

<a id='Ec9_5'></a>
\begin{equation*}
\begin{aligned}
\begin{array}{ll}
\text { 1D: } & \mathbf{g}=[g]=\left[\frac{\partial \phi}{\partial x}\right] \\
\text { 2D: } & \mathbf{g}=\left[g_{x}, g_{y}\right]^{T}=\left[\frac{\partial \phi}{\partial x}, \frac{\partial \phi}{\partial y}\right]^{T} \\
\text { 3D: } & \mathbf{g}=\left[g_{x}, g_{y}, g_{z}\right]^{T}=\left[\frac{\partial \phi}{\partial x}, \frac{\partial \phi}{\partial y}, \frac{\partial \phi}{\partial z}\right]^{T}
\end{array}
\end{aligned}
\label{eq:Ec9_5} \tag{9.5}
\end{equation*}

La relación entre flujo y gradiente se expresa por la ley de Fourier que en forma matricial se escribe por

<a id='Ec9_6'></a>
\begin{equation*}
\begin{aligned}
\mathbf{q}=-\mathbf{D} \mathbf{g}=-\mathbf{D} \nabla \phi
\end{aligned}
\label{eq:Ec9_6} \tag{9.6}
\end{equation*}

siendo $\mathbf{D}$ la matriz de conductividad dada por

<a id='Ec9_7'></a>
\begin{equation*}
\begin{aligned}
\begin{array}{ll}
\text { 1D: } & \mathbf{D}=[k]\\
\text { 2D: } & \mathbf{D}=\left[\begin{array}{cc}k_{x} & 0 \\ 0 & k_{y}\end{array}\right] \\ 
\text { 3D: } & \mathbf{D}=\left[\begin{array}{ccc}k_{x} & 0 & 0 \\ 0 & k_{y} & 0 \\ 0 & 0 & k_{z}\end{array}\right]
\end{array}
\end{aligned}
\label{eq:Ec9_7} \tag{9.7}
\end{equation*}

Sustituyendo la [Ec. 9.6](#Ec9_6) en la [Ec. 9.1](#Ec9_1) se obtiene la forma matricial de las ecuaciones de balance de flujos por

<a id='Ec9_8'></a>
\begin{equation*}
\begin{aligned}
\nabla^{T}(\mathrm{D} \nabla \phi)+Q=0 \quad \text { en } \Omega
\end{aligned}
\label{eq:Ec9_8} \tag{9.8}
\end{equation*}

Desarrollando dicha ecuación matricial se obiene la forma usual de la ecuación de Poisson. Así en $2D$

<a id='Ec9_9'></a>
\begin{equation*}
\begin{aligned}
&{\left[\frac{\partial}{\partial x}, \frac{\partial}{\partial y}\right]\left[\begin{array}{cc}
k_{x} & 0 \\
0 & k_{y}
\end{array}\right]\left\{\begin{array}{l}
\frac{\partial \phi}{\partial x} \\
\frac{\partial \phi}{\partial y}
\end{array}\right\}+Q=} \\
&\quad=\frac{\partial}{\partial x}\left(k_{x} \frac{\partial \phi}{\partial x}\right)+\frac{\partial}{\partial y}\left(k_{y} \frac{\partial \phi}{\partial y}\right)+Q=0
\end{aligned}
\label{eq:Ec9_9} \tag{9.9}
\end{equation*}

Las condiciones de contorno se escriben de fornia general por:

- ***Condición de variable prescrita***

<a id='Ec9_10'></a>
\begin{equation*}
\begin{aligned}
\phi-\bar{\phi}=0 \quad \text { en } \Gamma_{\phi}
\end{aligned}
\label{eq:Ec9_10} \tag{9.10}
\end{equation*}

- ***Condición de flujo prescrito***

<a id='Ec9_11'></a>
\begin{equation*}
\begin{aligned}
\mathbf{n}^{T} \mathbf{D} \nabla \phi+\bar{q}_{n}+\alpha\left(\phi-\phi_{q}\right)=0 \quad \text { en } \Gamma_{q}
\end{aligned}
\label{eq:Ec9_11} \tag{9.11}
\end{equation*}

donde $\mathbf{n}$ es la normal exterior al contorno $\Gamma_{q}$ y el resto de los términos tienen el significado usual.


### Forma integral débil

La forma integral equivalente a las ecuaciones de balance y las condiciones de contorno se obtiene a partir del *MRP* por

<a id='Ec9_12'></a>
\begin{equation*}
\begin{aligned}
\int_{\Omega} W\left[\nabla^{T} \mathbf{D} \nabla \phi+Q\right]+\oint_{\Gamma_{q}} W\left[\mathbf{n}^{T} \mathbf{D} \nabla \phi+\bar{q}_{n}+\alpha\left(\phi-\phi_{a}\right)\right] \mathrm{d} \boldsymbol{\Gamma}=0
\end{aligned}
\label{eq:Ec9_12} \tag{9.12}
\end{equation*}

Integrando por partes el término en derivadas segundas en la primera integral y escogiendo $\bar{W}=-W$ se obtiene la forma débil por

<a id='Ec9_13'></a>
\begin{equation*}
\begin{aligned}
-\int_{\Omega}\left(\nabla^{T} W\right) \mathbf{D} \nabla \phi \mathrm{d} \Omega+\int_{\Omega} W Q \mathrm{d} \Omega-\int_{\Gamma_{q}} W\left[\bar{q}_{n}-\alpha\left(\phi-\phi_{a}\right)\right] \mathrm{d} \Omega+\int_{\Gamma_{\phi}} W \mathbf{n}^{T} \mathbf{D} \nabla \phi \mathrm{d} \Omega=0
\end{aligned}
\label{eq:Ec9_13} \tag{9.13}
\end{equation*}

Teniendo en cuenta la definición del vector de flujos y denominando $W=$ $\delta \phi$ la variación virtual de la incógnita y $\nabla=\delta \mathrm{g}$ el vector de gradientes virtuales, la forma débil se puede reescribir por

<a id='Ec9_14'></a>
\begin{equation*}
\begin{aligned}
\int_{\Omega} \delta \mathbf{g}^{T} \mathbf{q} \mathrm{d} \Omega+\int_{\Omega} \delta \phi Q \mathrm{d} \Omega-\int_{\Gamma_{q}} \delta \phi\left[\bar{q}_{n}-\alpha\left(\phi-\phi_{a}\right)\right] \mathrm{d} \mathbf{\Gamma}-\int_{\Gamma_{\phi}} \delta \phi q_{n} \mathrm{d} \mathbf{\Gamma}=0
\end{aligned}
\label{eq:Ec9_14} \tag{9.14}
\end{equation*}

La [Ec. 9.14](#Ec9_14) puede denominarse principio de variaciones virtuales y es la versión equivalente al principio de trabajos virtuales en mecánica de sólidos. En la práctica se supone que el campo virtual $\delta \phi$ satisface las condiciones de contorno en la variable. Asi si $\phi=0$ en $\Gamma_{\phi}$ se prescinde de la última integral en [Ec. 9.14](#Ec9_14).

La expresión integral dad por la [Ec. 9.13](#Ec9_13) (o la equivalente [(9.14)](#Ec9_14)) es el punto de partida para obtener las ecuaciones de la discretización por el MEF.


### Forma matricial de las ecuaciones del MEF

La aproximación de elementos finitos se escribe matricialmente para un elemento de $n$ nodos por

<a id='Ec9_15'></a>
\begin{equation*}
\begin{aligned}
\phi=\sum_{i=1}^{n} N_{i}^{(e)} \phi_{i}^{(e)}=\left[N_{1}^{(e)}, N_{2}^{(e)}, \ldots, N_{n}^{(e)}\right]\left\{\begin{array}{c}
\phi_{1}^{(e)} \\
\phi_{2}^{(e)} \\
\vdots \\
\phi_{n}^{(e)}
\end{array}\right\}=\mathbf{N}^{(e)} \mathbf{a}^{(e)}
\end{aligned}
\label{eq:Ec9_15} \tag{9.15}
\end{equation*}

con

<a id='Ec9_16'></a>
\begin{equation*}
\begin{aligned}
&\mathbf{N}^{(e)}=\left[N_{1}^{(e)}, N_{2}^{(e)}, \ldots, N_{n}^{(e)}\right] \\
&\mathbf{a}^{(e)}=\left[\phi_{1}^{(e)}, \phi_{2}^{(e)}, \ldots, \phi_{n}^{(e)}\right]^{T}
\end{aligned}
\label{eq:Ec9_16} \tag{9.16}
\end{equation*}

La matriz $\mathbf{N}^{(e)}$ recibe el nombre de matriz de funciones de forma del elemento y $\mathbf{a}^{(e)}$ es el vector de variables nodales del elemento.

El vector de gradientes se expresa en función de las variables nodales elementales por

<a id='Ec9_17'></a>
\begin{equation*}
\begin{aligned}
\mathrm{g}=\nabla \phi=\left[\nabla \mathrm{N}^{(e)} \mathbf{a}^{(e)}\right]=\left[\nabla \mathbf{N}^{(e)}\right] \mathbf{a}^{(e)}=\mathbf{B}_{i}^{(e)} \mathbf{a}^{(e)}
\end{aligned}
\label{eq:Ec9_17} \tag{9.17}
\end{equation*}

donde $\mathbf{B}^{(e)}$ es la matriz de gradientes del elemento definida por

<a id='Ec9_18'></a>
\begin{equation*}
\begin{aligned}
\mathbf{B}^{(e)} &=\nabla \mathbf{N}^{(e)}=\nabla\left[N_{1}^{(e)}, N_{2}^{(e)}, \ldots, N_{n}^{(e)}\right]=\\
&=\left[\nabla N_{1}^{(e)}, \nabla N_{2}^{(e)}, \ldots, \nabla N_{n}^{(e)}\right]=\left[\mathbf{B}_{1}^{(e)}, \mathbf{B}_{2}^{(e)}, \ldots, \mathbf{B}_{n}^{(e)}\right]
\end{aligned}
\label{eq:Ec9_18} \tag{9.18}
\end{equation*}

siendo $\mathbf{B}_{i}^{(e)}=\nabla N_{i}^{(e)}$ la matriz de gradientes del nodo $i$ del elemento $e$. Obsérvese que $\mathbf{B}^{(e)}$ está compuesta de tantas submatrices $\mathbf{B}_{i}^{(e)}$ como nodos tiene el elemento. Así, por ejemplo, para un elemento triangular de tres nodos

<a id='Ec9_19'></a>
\begin{equation*}
\begin{aligned}
\mathbf{B}^{(e)}=\left[\begin{array}{c|c}
\frac{\partial N_{1}^{(e)}}{\partial x} & \frac{\partial N_{2}^{(e)}}{\partial x} & \frac{\partial N_{3}^{(e)}}{\partial x} \\
\frac{\partial N_{1}^{(e)}}{\partial y} & \frac{\partial N_{2}^{(e)}}{\partial y} & \frac{\partial N_{3}^{(e)}}{\partial y}\\
\end{array} \begin{array}{c}
\end{array}\right] ; \quad \mathbf{B}_{i}^{(e)}=\left[\begin{array}{c}
\frac{\partial N_{i}^{(0)}}{\partial x} \\
\frac{\partial N_{i}^{(e)}}{\partial y}
\end{array}\right]=\frac{1}{2 A^{(e)}}\left\{\begin{array}{l}
b_{i}^{(e)} \\
c_{i}^{(e)}
\end{array}\right\}
\end{aligned}
\label{eq:Ec9_19} \tag{9.19}
\end{equation*}

El vector de flujos se expresa en función de las variables nodales por

<a id='Ec9_20'></a>
\begin{equation*}
\begin{aligned}
\mathbf{q}=-\mathbf{D}_{\mathbf{g}}=-\mathbf{D} \mathbf{B}^{(e)} \mathbf{a}^{(e)}
\end{aligned}
\label{eq:Ec9_20} \tag{9.20}
\end{equation*}

Sustituyendo las ecs. [(9.17)](#Ec9_17) y [(9.18)](#Ec9_18) en la expresión integral [(9.13)](#Ec9_13) y escogiendo la forma de $Galerkin$ con $W_{i}=\delta \phi=N_{i}$ se obtiene el sistema de ecuaciones de la discretización en la forma $\mathbf{K} \mathbf{a}=\mathbf{f}$ en la que la matriz $\mathbf{K}$ y el vector $\mathbf{f}$ se obtienen ensamblando las contribuciones elementales en la forma usual.

La expresión de la matriz de rigidez del elemento puede escribirse en forma genérica por

<a id='Ec9_21'></a>
\begin{equation*}
\begin{aligned}
\mathbf{K}^{(e)}=\int_{\Omega^{(e)}} \mathbf{B}^{(e)^{T}} \mathbf{D B}^{(e)} \mathrm{d} \Omega+\int_{\Gamma_{9}^{(e)}} \alpha\left[\mathbf{N}^{(e)}\right]^{T} \mathbf{N}^{(e)} \mathrm{d} \Gamma
\end{aligned}
\label{eq:Ec9_21} \tag{9.21}
\end{equation*}

con

<a id='Ec9_22'></a>
\begin{equation*}
\begin{aligned}
K_{i j}^{(e)}=\int_{\Omega^{(e)}}\left[\mathbf{B}_{i}^{(e)}\right]^{T} \mathbf{D B}_{j}^{(e)} \mathrm{d} \Omega+\int_{\Gamma_{q}^{(e)}} \alpha N_{i}^{(e)} N_{j}^{(e)} \mathrm{d} \mathbf{\Gamma}
\end{aligned}
\label{eq:Ec9_22} \tag{9.22}
\end{equation*}

$\mathbf{L a}$ expresión de $\mathbf{K}^{(c)}$ anterior es válida para problemas uni, bi y tridimensionales. Obviamente, en cada caso las integrales se efectúan sobre el dominio elemental correspondiente y su parte del contorno donde se prescribe el flujo normal, en su caso.

El vector de flujos nodales equivalentes del elemento puede también escribirse matricialmente por

<a id='Ec9_23'></a>
\begin{equation*}
\begin{aligned}
\mathbf{f}^{(e)}=\int_{\Omega^{(e)}}\left[\mathbf{N}^{(e)}\right]^{T} Q \mathrm{d} \Omega-\oint_{\Gamma_{q}^{(e)}}\left[\mathbf{N}^{(e)}\right]^{T}\left[\bar{q}_{n}-\alpha \phi_{a}\right]-\mathbf{r}^{(e)}
\end{aligned}
\label{eq:Ec9_23} \tag{9.23}
\end{equation*}

con

<a id='Ec9_24'></a>
\begin{equation*}
\begin{aligned}
f_{i}^{(e)}=\int_{\Omega^{(e)}} i v_{i}^{(e)} Q \mathrm{d} \Omega-\oint_{\Gamma_{q}^{(e)}} N_{i}^{(e)}\left[\bar{q}_{n}-\alpha \phi_{a}\right] \mathrm{d} \Gamma-q_{n_{i}}^{(e)}
\end{aligned}
\label{eq:Ec9_24} \tag{9.24}
\end{equation*}

Como antes, las integrales se calculan sobre el dominio elemental correspondiente.

Se recuerda, asimismo, que $\mathbf{r}^{(e)}$ es el vector de flujo "reacción " en los nodos del elemento que salen del contorno $\Gamma_{\phi}$ donde la variable $\phi$ está prescrita.
Es interesante remarcar de nuevo que la formulación matricial anterior proporciona un marco general para la solución de cualquier problema de Poisson en 1D, 2D y 3D. El planteamiento matricial aquí utilizado se repite de forma casi idéntica al aplicar el MEF a otros problemas de mecánica en estructuras y fluidos.


### Obtención de las ecuaciones matriciales de equilibrio a partir de la suma de flujos nodales

Vimos en un capítulo anterior que la forma usual de obtener las ecuaciones matriciales globales en sistemas discretos era utilizar el concepto de equilibrio de flujos en todos los nudos del sistema. Así, en redes eléctricas se aplica la conocida ley de Kirchhoff que establece que la suma de corrientes que concurren en un nudo es nula. En redes de tuberías se aplica una ley análoga, mientras que la ley equivalente en problemas de estructura es que la suma de las fuerzas aplicadas en un nudo de la estructura es igual a cero. Expresando los "flujos" (corrientes eléctricas, fuerzas o flujos de agua) en los extremos de cada "elemento" en función de las variables nodales y las acciones exteriores a través de las ecuaciones de equilibrio elemental, y aplicando la ecuación de equilibrio nodal para todos los nudos del sistema, se obtiene el sistema de ecuaciones global $\mathbf{K a}=\mathbf{f}$ que relaciona las acciones exteriores y las incógnitas nodales. El proceso de "suma de flujos" en cada nodo permite obtener las expresiones globales de $\boldsymbol{K}^{(e)}$ y $\mathbf{f}^{(e)}$ para cada elemento discreto.

Se ha recordado aquí este proceso porque es posible reproducirlo exactamente en la solución de la ecuación de Poisson por el MEF. Para mayor generalidad utilizaremos la notación matricial estudiada en apartados anteriores.

#### Elemento unidimensional de dos nodos

Comencemos por el sencillo elemento unidimensional sobre el que actúan un flujo distribuido $Q$ y dos flujos $q_{1}^{(e)}$ y $q_{2}^{(e)}$ que salen por los nodos 1 y 2 del elemento respectivamente.

Consideremos ahora la expresión de balance global de flujos sobre el elemento escrita en forma integral. Esta ecuación puede obtenerse a través de la forma integral débil, o más directamente por la expresión equivalente del principio de variaciones virtuales que se escribe por 

<a id='Ec9_25'></a>
\begin{equation*}
\begin{aligned}
\int_{l(e)} \delta g q \mathrm{d} x+\int_{l^{(e)}} \delta \phi Q \mathrm{d} x-\delta \phi_{1}^{(e)} q_{1}^{(e)}-\delta \phi_{2}^{(e)} q_{2}^{(e)}=0
\end{aligned}
\label{eq:Ec9_25} \tag{9.25}
\end{equation*}

Los signos menos en la ecuación anterior surgen de la definición de los flujos $q_{1}^{(c)}$ y $q_{2}^{(e)}$ hacia el exterior del elemento (el calor "sale" del elemento).

Recordemos que en la integral anterior $\delta \phi$ son variaciones virtuales (arbitrarias) de $\phi$ y $\delta g$ es el campo de gradientes virtuales producido por $\delta \phi\left(\delta_{\mathbf{g}}=\right.$ $\nabla \delta \phi) .$

La [Ec. 9.25](#Ec9_25) permite expresar el "trabajo virtual" realizado por los flujos nodales en función del resto de variables elementales por

<a id='Ec9_26'></a>
\begin{equation*}
\begin{aligned}
\left[\delta \mathbf{a}^{(e)}\right]^{T} \mathbf{q}^{(e)}=\int_{I(e)} \delta g q \mathrm{d} \Omega+\int_{l(e)} \delta \phi Q \mathrm{d} \Omega
\end{aligned}
\label{eq:Ec9_26} \tag{9.26}
\end{equation*}

siendo

<a id='Ec9_27'></a>
\begin{equation*}
\begin{aligned}
\delta \mathbf{a}^{(e)}=\left[\delta \phi_{1}^{(e)}, \delta \phi_{2}^{(e)}\right]^{T}, \quad \mathbf{q}^{(e)}=\left[q_{1}^{(e)}, q_{2}^{(e)}\right]^{T}
\end{aligned}
\label{eq:Ec9_27} \tag{9.27}
\end{equation*}

El vector $\mathbf{q}^{(\boldsymbol{e})}$ recibe el nombre de vector de flujos nodales interelementales de equilibrio.

Sustituyamos ahora en la [Ec. 9.26](#Ec9_26) la aproximación del MEF. Los campos $\phi$ y $\delta \phi$ pueden expresarse en el interior de cada elemento en forma matricial por

<a id='Ec9_28'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\phi=\sum_{i=1}^{2} N_{i}^{(e)} \phi_{i}^{(e)}=\mathbf{N}^{(e)} \mathbf{a}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec9_28} \tag{9.28}
\end{equation*}

<a id='Ec9_29'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\delta \phi=\sum_{i=1}^{2} N_{i}^{(e)} \delta \phi_{i}^{(e)}=\mathbf{N}^{(e)} \delta \mathbf{a}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec9_29} \tag{9.29}
\end{equation*}

con

<a id='Ec9_30'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{N}^{(e)}=\left[N_{1}^{(e)}, N_{2}^{(e)}\right], \quad \mathbf{a}^{(e)}=\left[\phi_{1}^{(e)}, \phi_{2}^{(e)}\right]^{T}
\end{gathered}
\end{aligned}
\label{eq:Ec9_30} \tag{9.30}
\end{equation*}

y

<a id='Ec9_31'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\delta \mathrm{a}^{(e)}=\left[\delta \phi_{1}^{(e)}, \delta \phi_{2}^{(e)}\right]^{T}
\end{gathered}
\end{aligned}
\label{eq:Ec9_31} \tag{9.31}
\end{equation*}

El gradiente virtual se expresa en función de las variables virtuales nodales $\delta \mathrm{a}^{(e)}$ por

<a id='Ec9_32'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\delta g=\nabla \delta \phi=\left[\nabla \mathrm{N}^{(e)}\right] \delta \mathrm{a}^{(e)}=\mathrm{B}^{(e)} \delta \mathrm{a}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec9_32} \tag{9.32}
\end{equation*}

con

<a id='Ec9_33'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\nabla=\left[\frac{\partial}{\partial x}\right], \quad \mathbf{B}^{(e)}=\left[B_{1}^{(e)}, B_{2}^{(e)}\right], \quad B_{i}^{(e)}=\left[\frac{\partial N_{i}^{(e)}}{\partial x}\right]
\end{gathered}
\end{aligned}
\label{eq:Ec9_33} \tag{9.33}
\end{equation*}

Sustituyendo las expesiones anteriores en [Ec. 9.26](#Ec9_26) y sacando factor común $\delta \mathrm{a}^{(e)}$ en el segundo miembro se obtiene

<a id='Ec9_34'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\left.\left[\delta \mathbf{a}^{(e)}\right]^{T} \mathbf{q}^{(e)}=\left[\delta \mathbf{a}^{(e)}\right]^{T}\left\{-\left[\int_{l^{(e)}}\left[\mathbf{B}^{(e)}\right]^{T} \mathbf{D B}^{(e)}  \mathrm{d} \Omega\right] \mathbf{a}^{(e)}+\int_{l^{(e)}}\left[\mathbf{N}^{(e)}\right]^{T} Q  \mathrm{d} \Omega\right]\right\}
\end{gathered}
\end{aligned}
\label{eq:Ec9_34} \tag{9.34}
\end{equation*}

Teniendo en cuenta que las variables virtuales nodales $\delta \phi \int$ for arbitrarias, el cumplimiento de la ecuación anterior para cualquier valor de $\delta \mathbf{a}^{(e)}$ exige que

<a id='Ec9_35'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{q}^{(e)}=-\left[\int_{I^{(e)}}\left[\mathbf{B}^{(e)}\right]^{T} \mathbf{D B}^{(e)}  \mathrm{d} x\right] \mathbf{a}^{(e)}+\int_{l(e)}\left[\mathbf{N}^{(e)}\right]^{T} Q  \mathrm{d} x
\end{gathered}
\end{aligned}
\label{eq:Ec9_35} \tag{9.35}
\end{equation*}

o

<a id='Ec9_36'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{q}^{(e)}=-\mathbf{K}^{(e)} \mathbf{a}^{(e)}+\mathbf{f}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec9_36} \tag{9.36}
\end{equation*}

siendo

<a id='Ec9_37'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\begin{gathered}
\mathbf{K}^{(e)}=\int_{I(e)}\left[\mathbf{B}^{(e)}\right]^{T} \mathbf{D B}^{(e)} \mathrm{d} x
\end{gathered}
\end{gathered}
\end{aligned}
\label{eq:Ec9_37} \tag{9.37}
\end{equation*}

<a id='Ec9_38'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\begin{gathered}
\mathbf{f}^{(e)}=\int_{l(e)}\left[\mathbf{N}^{(e)}\right]^{T} Q  \mathrm{d} x
\end{gathered}
\end{gathered}
\end{aligned}
\label{eq:Ec9_38} \tag{9.38}
\end{equation*}

la matriz de rigidez del elemento y el vector de flujos nodales debidos a la fuente distribuida $Q$.

La expresión [(9.36)](#Ec9_36) es la ecuación de balance de flujos del elemento y expresa el equilibrio entre los flujos nodales interelementales, el flujo debido a la fuente $Q$ y el flujo debido al gradiente de $\phi$ en el interior del elemeno. Obsérvase la analogía entre la ecuación de equlibrios de flujos nodales elementales y la ecuación de equilibrio de fuerzas nodales en un elemento de un sistema discreto. Dicha analogía se cumple también para el proceso de ensamblaje, como se explica a continuación.

La [Ec. 9.36](#Ec9_36) es clave para obtener las ecuaciones de equilibrio global en función de las variables elementales. El proceso que se sigue es idéntico al explicado para los sistemas discretos. En cada nodo se establece el equilibrio entre todos los flujos interelementales contri\iotauidos por los elementos que concurren en el nodo.

Así, la ecuación de equilibrio nodal se escribe por

<a id='Ec9_39'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\begin{gathered}
\sum_{e} q_{i}^{(e)}=q_{j}^{\mathrm{c} r t} \quad j=1,2, \ldots N
\end{gathered}
\end{gathered}
\end{aligned}
\label{eq:Ec9_39} \tag{9.39}
\end{equation*}

donde la suma se extiende sobre todos los elementos nue concurren en el nodo con numeración global $j$ y $q_{j}^{\text {ext }}$ es el flujo que entra desde el exterior al nodo $j$. (En general $q_{j}^{\text {ext }}=0$ ). En [Ec. 9.39](#Ec9_39) $N$ es el número total de nodos en la malla.

Sustituyendo los valores de $q_{i}^{(e)}$ de las ecuaciones de balance elemental [Ec. 9.36](#Ec9_36) en función de las variables nodales a $^{(e)}$ y de los flujos nodales "exteriores" $f^{(e)}$, se obtiene el sistema de ecuaciones de balance global en la forma usual $\mathbf{K a}=\mathbf{f}$. Tanto la matriz de rigidez $\mathbf{K}$ como el vector de flujos nodales $\mathbf{f}$ se forman siguiendo exactamente las mismas reglas de ensamblaje que en el caso de sistemas discretos.

El proceso de ensamblaje anterior es idéntico al explicado en los capítulos anteriores utilizando la formulación integral y las propiedades aditivas de las integrales definidas sobre todo el dominio de análisis. Según sea la formación básica del lector, bien más matemática o ingenieril, preferirá uno u otro procedimiento. Lo realmente importante es no olvidar que ambos procedimientos conducen, en la práctica, a un proceso de ensamblaje idéntico y a las mismas ecuaciones matriciales globales.

Para aclarar todavía más este concepto repetiremos el proceso de ensamblaje anterior para el elemento triangular de tres nodos.


#### Ejemplo 9.1 Obtención de las ecuaciones matriciales globales a partir de los flujos nodales para el triángulo de tres nodos

***Solución***

Consideremos el elemento triangular de tres nodos de la [Figura 9.1b](#Fig9_1). El elemento está sometido a una fuente distribuida $Q$, a los flujos en dirección normal a través de los lados que forman parte de contornos exteriores $\Gamma_{q}$ y a los flujos nodales interelementales. La expresión de balance entre todos estos flujos y los gradientes internos se puede expresar por el principio de variaciones virtuales que se escribe en este caso por

<a id='Ec9_40'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\iint_{\Omega^{(\mathrm{e})}}\left[\delta \mathrm{g}^{(\mathrm{e})}\right]^{T} \mathbf{q} \mathrm{d} \Omega+\iint_{\Omega^{(\mathrm{e})}} \delta \phi Q \mathrm{d} \Omega-\int_{\Gamma_{q}^{(\mathrm{e})}} \delta \phi \bar{q}_{n} \mathrm{d} \Gamma-\left[\delta \mathbf{a}^{(\mathrm{e})}\right]^{T} \mathbf{q}^{(\mathrm{e})}=0
\end{gathered}
\end{aligned}
\label{eq:Ec9_40} \tag{9.40}
\end{equation*}

donde

<a id='Ec9_41'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\delta \mathbf{g}^{(e)}=\left[\delta g_{x}, \delta g_{y}\right]^{T}=\nabla(\delta \phi)
\end{gathered}
\end{aligned}
\label{eq:Ec9_41} \tag{9.41}
\end{equation*}

<a id='Ec9_42'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{q}=\left[q_{x}, q_{y}\right]^{T}=-\mathbf{D} \mathbf{g}=-\mathbf{D} \nabla \phi, \quad \nabla=\left\{\begin{array}{l}\frac{\partial}{\partial x} \\
\frac{\partial}{\partial y}
\end{array}\right\}
\end{gathered}
\end{aligned}
\label{eq:Ec9_42} \tag{9.42}
\end{equation*}

<a id='Ec9_43'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\delta \mathbf{a}^{(e)}=\left[\delta \phi_{1}^{(e)}, \partial \phi_{2}^{(e)}, \delta \phi_{3}^{(e)}\right]^{T}
\end{gathered}
\end{aligned}
\label{eq:Ec9_43} \tag{9.43}
\end{equation*}

<a id='Ec9_44'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{q}^{(e)}=\left[q_{1}^{(e)}, q_{2}^{(e)}, q_{3}^{(e)}\right]^{T}
\end{gathered}
\end{aligned}
\label{eq:Ec9_44} \tag{9.44}
\end{equation*}

En lo anterior $q_{i}^{(c)}$ es el flujo interelemental de equilibrio que sale del nodo $i$. Dicho flujo puede interpretarse como la suma de los flujos en direcciones $x$ e $y$, es decir $q_{i}^{(e)}=q_{x_{i}}^{(e)}+q_{y_{i}}^{(e)}$.

Es importante advartir que el vector $\mathbf{q}^{(e)}$ incorpora el efecto de los flujos que se transmiten a través de los contornos interelementales. Por otra parte, la tercera integral en la expresión anterior sólo aparece si el q̣lemento tiene algún lado sobre el contorno donde el flujo está prescrito a un valor $\bar{q}_{n}$. Asimismo, en el caso de existir fenómenos de convección con el exterior habrá que añadir el término $\alpha\left(\phi-\phi_{a}\right)$ al valor de $\bar{q}_{n}$. Para mayor sencillez prescindiremos aquí de este término.

La discretización del MEF se escribe en la forma matricial usual por

<a id='Ec9_45'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\phi=\sum_{i=1}^{3} N_{i}^{(e)} \phi_{i}^{(e)}=\mathbf{N}^{(e)} \varepsilon^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec9_45} \tag{9.45}
\end{equation*}

<a id='Ec9_46'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{g}=\mathbf{B}^{(e)} \mathbf{a}^{(e)}, \quad \delta \mathbf{g}=\mathbf{B}^{(e)} \delta \mathbf{a}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec9_46} \tag{9.46}
\end{equation*}

<a id='Ec9_47'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{q}=-\mathbf{D}_{\mathbf{g}}=-\mathbf{D} \mathbf{B}^{(e)} \mathbf{a}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec9_47} \tag{9.47}
\end{equation*}

con

<a id='Ec9_48'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathrm{N}^{(e)}=\left[N_{1}^{(e)}, N_{2}^{(e)}, N_{3}^{(e)}\right]
\end{gathered}
\end{aligned}
\label{eq:Ec9_48} \tag{9.48}
\end{equation*}

<a id='Ec9_49'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{B}^{(e)}=\left[\mathbf{B}_{1}^{(e)}, \mathbf{B}_{2}^{(e)}, \mathbf{B}_{3}^{(e)}\right], \quad \mathbf{B}_{i}^{(e)}=\nabla N_{i}^{(e)}=\left[\begin{array}{c}\frac{\partial N_{i}^{(e)}}{\partial x} \\
\frac{\partial N_{i}^{(e)}}{\partial y}
\end{array}\right]
\end{gathered}
\end{aligned}
\label{eq:Ec9_49} \tag{9.49}
\end{equation*}

<a id='Ec9_50'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{D}=\left[\begin{array}{cc}
k_{x} & 0 \\
0 & k_{y}
\end{array}\right]
\end{gathered}
\end{aligned}
\label{eq:Ec9_50} \tag{9.50}
\end{equation*}

Sustituyendo estas expresiones en la forma integral y sacando factor común $\delta \mathbf{a}^{(e)}$ se deduce

<a id='Ec9_51'></a>
\begin{equation*}
\begin{gathered}
\begin{aligned}
\left[\delta \mathbf{a}^{(e)}\right]^{T} \mathbf{q}^{(e)} &=\left[\delta \mathbf{a}^{(e)}\right]^{T}\left\{\left[-\iint_{\Omega^{(e)}}\left[\mathbf{B}^{(e)}\right]^{T} \mathbf{D B}^{(e)} \mathrm{d} \Omega\right] \mathbf{a}^{(e)}+\right.\\
&\left.\left.+\iint_{\Omega(e)}\left[\mathbf{N}^{(e)}\right]^{T} Q \mathrm{d} \Omega-\oint_{\Gamma_{q}^{(e)}}\left[\mathbf{N}^{(e)}\right]^{T} \bar{q}_{n} \mathrm{d} \Gamma\right]\right\}
\end{aligned}
\end{gathered}
\label{eq:Ec9_51} \tag{9.51}
\end{equation*}

de donde se obtiene

<a id='Ec9_52'></a>
\begin{equation*}
\begin{gathered}
\begin{aligned}
\mathbf{q}^{(e)}=-\mathbf{K}^{(e)} \mathbf{a}^{(e)}+\mathbf{f}^{(e)}
\end{aligned}
\end{gathered}
\label{eq:Ec9_52} \tag{9.52}
\end{equation*}

con

<a id='Ec9_53'></a>
\begin{equation*}
\begin{gathered}
\begin{aligned}
K_{i j}^{(e)} &=\iint_{\Omega^{(e)}}\left[\mathbf{B}_{i}^{(e)}\right]^{T} \mathbf{D B}_{j}^{(e)} \mathrm{d} \Omega \\
f_{i}^{(e)} &=\iint_{\Omega^{(e)}} N_{i}^{(e)} Q \mathrm{d} \Omega-\oint_{\Gamma_{q}^{(e)}} N_{i}^{(e)} \bar{q}_{n} \mathrm{d} \Gamma
\end{aligned}
\end{gathered}
\label{eq:Ec9_53} \tag{9.53}
\end{equation*}

El proceso de ensamblaje sigue ahora la etapa usual de establecer el equilibrio de flujos nodales interelementales como

$$
\sum_{e} q_{i}^{(e)}=q_{j}^{\text {ext }} \quad j=1,2, \ldots N
$$

Sustituyendo en esta suma las expresiones de $q_{i}^{(e)}$ de [Ec. 9.53](#Ec9_53) para cada elemento se obtiene el sistema de ecuaciones matriciales global $\mathbf{K a}=\mathbf{f}$ en el que la matriz $\mathbf{K}$ y el vector $\mathbf{f}$ se forman ensamblando las contribuciones elementales $\mathbf{K}^{(e)}$ y $\mathbf{f}^{(e)}$ en la forma estándar.
