## Cálculo Variacional

### Valores extremos de una función

El [criterio de la primera derivada](https://en.wikipedia.org/wiki/Derivative_test), visto en un curso de cálculo diferencial, establece que la condición necesaria para que una función $f(x)$, diferenciable en un intervalo $x_1 < x < x_2$, tenga un valor extremo en el punto $(x:0, f(x_0))$ es

<a id='Ec2_1'></a>
\begin{equation*}
f'(x_0)=\left(\frac{df(x)}{dx}\right)_{x=x_0}=0
\label{eq:Ec2_1} \tag{2.1}
\end{equation*}

Una interpretación geométrica de la [Ec. 2.1](#Ec2_1) es que la recta tangente a la curva $f(x)$ en el punto $x_0$ es horizontal. Se pueden presentar tres situaciones:

<a id='Fig_2'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C02_F01_MaxMinInflection.gif?raw=true" width="250" />
</p>

<div style="text-align: center"><strong>Figura 1.</strong> Valores extremos de una función. <strong>Fuente:</strong> <a href="https://co.pinterest.com/pin/338262621996687738/">pinterest</a></div>

- El punto $x_0$ es un máximo relativo de la función $f(x)$


- El punto $x_0$ es un mínimo relativo de la función $f(x)$


- El punto $x_0$ es un punto de inflexión de la función $f(x)$

El [criterio de la segunda derivada](https://en.wikipedia.org/wiki/Derivative_test#Second-derivative_test_(single_variable)) permitirá establecer cuál de esas condiciones corresponderá a un valor extremo de la función en el punto $x_0$:

- La función tendrá un mínimo relativo si $f''(x_0) > 0$


- La función tendrá un máximo relativo si $f''(x_0) < 0$

También puede suceder que en un cierto valor $x=a$, se tenga que $f'(a)=f''(a)=0$, por lo que sería imposible determinar en cuál de las condiciones vistas se estaría en el punto. Esta dificultad se resolverá escriviendo la función como una expansión en [*series de Taylor*](https://en.wikipedia.org/wiki/Taylor_series) en un punto cercano a $x=a$, de la siguiente manera:

<a id='Ec2_2'></a>
\begin{equation*}
f(x)=f(a)+f'(a)(x-a)+\frac{f''(a)}{2!}(x-a)^2+\frac{f'''(a)}{3!}(x-a)^3+\ldots
\label{eq:Ec2_2} \tag{2.2}
\end{equation*}

La [Ec. 2.2](#Ec2_2) puede ser escrita de forma más compacta de la siguiente forma

<a id='Ec2_3'></a>
\begin{equation*}
f(x)-f(a)=f'(a)h+\frac{f''(a)}{2!}h^2+\frac{f'''(a)}{3!}h^3+\ldots
\label{eq:Ec2_3} \tag{2.3}
\end{equation*}

con $h = x-a$.

La [Ec. 2.3](#Ec2_3) establece las condiciones necesarias y suficientes para la existencia de máximos y mínimos. Para esto, supóngase que $f(a)$ sea un valor extremo de la función en el intervalo $x_1<a<x_2$. Con esto, $f(x)-f(a)$ será un número positivo si en el valor cercano a $x=a$ el valor de la función es un mínimo. Ahora, como $h$ puede ser positivo o negativo para los valores posibles de $x$, entonces se tiene que $f'(a)=0$ por ser el primer término del lado derecho de la igualdad. La condición necesaria para que se tenga un valor extremo en el punto $x=a$ es que $f'(a)=0$.

Como el primer término del lado derecho de la igualdad es nulo, el término predominante de la serie es el que contiene a la segunda derivada. Como $h^2$ siempre es positivo, se tiene que $f(a)$ será un valor mínimo si $f''(a)>0$ y será un valor máximo si $f''(a)<0$. Con lo que se tendría la condición de suficiencia.

Ahora, supóngase que las derivadas hasta de orden $n-1$ se anulan en el punto $x=a$, es decir

$$f'(a)=f''(a)=\ldots=f^{(n-1)}(a)=0$$

pero con $f^{(n)} \neq 0$, se puede concluir que:

- Si $n$ es par, $f(a)$ será un valor máximo si $f^{(n)}(a)<0$ y $f(a)$ será un valor mínimo si $f^{(n)}(a)>0$


- si $n$ es impar, se tiene un punto de inflexión.

De forma general, la condición necesaria y suficiente para que una función bidimensioanl $f(x,y)$, diferenciable, tenga un valor extremo en un punto de coordenadas $x=x_0$ y $y=y_0$, es que su diferencial se anule, es decir

<a id='Ec2_4'></a>
\begin{equation*}
df = \frac{\partial f}{\partial x}dx+\frac{\partial f}{\partial y}dy = 0
\label{eq:Ec2_4} \tag{2.4}
\end{equation*}

en $(x_0,y_0)$.

### Cálculo variacional

En 1696 el matemático [*Johann Bernoulli(1667-1748)*](https://en.wikipedia.org/wiki/Johann_Bernoulli) planteó el siguiente problema:

<font face = "Times New Roman"><div style="text-align: center"><em>Dados dos puntos A y B en un plano vertical, ¿cuál es la curva trazada por un punto sobre el que actúa solo la gravedad, que comienza en A y llega a B en el menor tiempo?</em></div></font>

Fue [*Leonhard Euler (1707-1783)*](https://en.wikipedia.org/wiki/Leonhard_Euler) quién dio respuesta a dicha cuestión con la ecuación de la [braquistócrona](https://en.wikipedia.org/wiki/Brachistochrone_curve) y dando inicio al [Cálculo variacional](https://en.wikipedia.org/wiki/Calculus_of_variations), en el que no se trabaja con funciones sino con [funcionales](https://en.wikipedia.org/wiki/Functional_(mathematics)). Los funcionales son entidades que dependen de una función, es decir, la función de una función. 

Como ejemplo básico de un funcional se tiene la expresión matemática que representa la longitud de la curva dada por

$$l=\int_{x_1}^{x_2} \sqrt{1+(y')^2}$$

El cálculo variacional estudia los métodos que permiten encontrar los valores extremos de los funcionales. Las leyes de la mecánica que aplican a esos métodos son llamados *principios variacionales*. A ellas pertenecen, por ejemplo, los principios de [minimización de la energía potencial](https://en.wikipedia.org/wiki/Principle_of_minimum_energy), [ley de la conservación de la energía](https://en.wikipedia.org/wiki/Conservation_of_energy), entre otros.

### Operador variacional

<a id='Fig_2'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C02_F02_OperadorVariacional01.PNG?raw=true" width="250" />
</p>

<div style="text-align: center"><strong>Figura 2.</strong> funcionales. <strong>Fuente:</strong> <a href="https://editoraunicamp.com.br/produto/503/metodo-dos-elementos-finitos-primeiros-passos-ed--3">Assan, Método dos elementos finitos - Primeiros passos. </a></div>

Sea $F$ una función que depende de la función $y$ y de su primera derivada $y'=\frac{dy}{dx}$, denotada como:

$$F=F(x,y,y')$$

de la figura se tiene que en el intervalo $x_1<x<x_2$, la función $y=y(x)$ asume los valores de $y_1=y(x_1)$ y $y_2=y(x_2)$.

Podemos pensar que la función $y(x)$ es la representación analítica de las desviaciones de una viga en equilibrio sometida a cierto estado de carga.

Se dice que una función es *admisible* cuando, dentro de cierta tolerancia y satisfaciendo las condiciones de contorno, representa a la función exacta $y(x)$. La función admisible es de la forma:

<a id='Ec2_5'></a>
\begin{equation*}
\bar{y}(x)=y(x)+\lambda \eta(x)
\label{eq:Ec2_5} \tag{2.5}
\end{equation*}

donde $\eta(x)$ es una función arbitraria que satisface las condiciones de contorno: $\eta(x_1)=0$ para $x=x_1$ y $\eta(x_2)=0$ para $x=x_2$, y $\lambda$ es un número pequeño. Para una misma función $\eta(x)$, se tiene una familia de funciones $\bar{y}(x)$, variando los valores de $\lambda$. 

Para cada valor de $x$, el producto $\lambda \eta(x)$ se puede entender como una variación de la función exacta. Entonces, se puede generalizar como:

<a id='Ec2_6'></a>
\begin{equation*}
\delta y = \lambda \eta(x)
\label{eq:Ec2_6} \tag{2.6}
\end{equation*}

derivando la [Ec. 2.5](#Ec2_5) respecto a $x$.

$$\bar{y}'(x)-y'(x)=\lambda \eta '(x)$$

que, en forma similar a la [Ec. 2.6](#Ec2_6) puede representarse como la variación de la derivada de $y(x)$:

<a id='Ec2_7'></a>
\begin{equation*}
\delta \left( \frac{dy}{dx}\right)=\delta y'=\lambda \eta'(x)
\label{eq:Ec2_7} \tag{2.7}
\end{equation*}

con esto, la variación de la función $F(x,y,y')$ asociada a la variación de $y$ y $y'$, para un valor de $x$ se escribe como:

$$\Delta F=F(x,y+\lambda \eta, y'+\lambda \eta')-F(x,y,y')$$

desarrollando el primer término del lado derecho de esta igualdad en una sere de Taylor y sustituyendo en la igualdad arriba, se tiene

$$\Delta F=\left( F+\frac{\partial F}{\partial y}\lambda \eta +\frac{\partial F}{\partial y'}\lambda \eta' +\frac{1}{2!}\frac{\delta^2 F}{\delta y^2}(\lambda \eta)^2 + \ldots\right)-F=\frac{\partial F}{\partial y}\lambda \eta + \frac{\partial F}{\partial y'}\lambda \eta'+\ldots $$

substituyendo en esa igualdad $\lambda \eta$ y $\lambda \eta'$ por los valores correspondientes a las igualdadas en las [Ecs. 2.4 y 2.5](#Ec2_4), la variación de la función $F$ resulta siendo:

<a id='Ec2_8'></a>
\begin{equation*}
\Delta F=\frac{\partial F}{\partial y}\delta y +\frac{\partial F}{\partial y'}\delta y' + \ldots
\label{eq:Ec2_8} \tag{2.8}
\end{equation*}

Se sabe que la [derivada total](https://en.wikipedia.org/wiki/Differential_of_a_function#Differentials_in_several_variables) de una función $f(u,v)$ de variables reales $u$ y $v$ está dada por

<a id='Ec2_9'></a>
\begin{equation*}
df=\frac{\partial f}{\partial u}du +\frac{\partial f}{\partial v}dv
\label{eq:Ec2_9} \tag{2.9}
\end{equation*}

comparando los dos primeros términos de las ecuaciones [Ecs. 2.8 y 2.9](#Ec2_8) se observa que son similares al de la derivada total. Esto permite tratar el variacional de la función $F$ como una derivada total de esa función. Empleando la notación dada en la [Ec. 2.6](#Ec2_6) se tiene:

$$\delta F=\frac{\partial F}{\partial y}\delta y+\frac{\partial F}{\partial y'}\delta y'$$

al operador $\delta$ se le denomina *operador variacional* y tiene las mismas propiedades del operador diferencial relativo a la variable dependiente, por ejemplo

$$\frac{d}{dx}(\delta F)=\delta \left(\frac{dF}{dX}\right)$$

$$\delta \int_{x_1}^{x_2}Fdx=\int_{x_1}^{x_2}\delta F dx$$

La función integral del tipo

$$Y=\int_{x_1}^{x_2}F(x,y,y')dx$$

que asume valores reales para los valores de la función $y$ es un funcional.

De las propiedades vistas arriba, se puede decir que la primera variación es dada por:

<a id='Ec2_10'></a>
\begin{equation*}
\delta Y=\int_{x_1}^{x_2}\left(\frac{\partial F}{\partial y}\delta y+\frac{\partial F}{\partial y'}\delta y'\right)dx
\label{eq:Ec2_10} \tag{2.10}
\end{equation*}

### Extremos de un funcional

Así como en una función la primera derivada igual a cero dtermina una condición extrema de la función, en los funcionales la condición extrema es determinada por la primera variación del funcional. En las funciones presentadas, los extremos pueden ocurrir para varios valores de la variable $x$. 

En el caso de los funcionales, se quiere encontrar una función que minimiza/maximiza el funcional, o sea, que vuelve el funcional estacionario y satisface las condiciones de contorno. Al conjunto de funciones que satisfacen las condiciones de contorno son llamadas de admisibles, y una de ellas vuelve el funcional estacionario. Se mostrará esto considerando el siguiente funcional:

<a id='Ec2_11'></a>
\begin{equation*}
Y=\int_{x_1}^{x_2}F(x,y,y')dx
\label{eq:Ec2_11} \tag{2.11}
\end{equation*}

El problema que se quiere resolver consiste en encontrar la función $y(x)$ que satisfaga las condiciones $y_1=y(x_1)$ en $x=x_1$ y $y_2=y(x_2)$ en $x=x_2$, y vuelve el funcional $Y$ estacionario. Los valores $y_1$ y $y_2$ son conocidos.

El funcional también puede expresarse en función de $\bar{y}(x)$, así

<a id='Ec2_12'></a>
\begin{equation*}
\bar{Y} = \int_{x_1}^{x_2}F(x,y,y')dx
\label{eq:Ec2_12} \tag{2.12}
\end{equation*}

En este funcional, cuando $\lambda = 0$, coincide con el valor extremo del funcional $Y$, y expandiendo en series de Taylor para un valor de $x$, se llega a:

<a id='Ec2_13'></a>
\begin{equation*}
\bar{Y} = Y+\lambda \left( \frac{d\bar{Y}}{d\lambda} \right)_{\lambda=0}+\frac{1}{2!}\lambda^2 \left( \frac{d^2\bar{Y}}{d\lambda^2} \right)_{\lambda=0}+\ldots
\label{eq:Ec2_13} \tag{2.13}
\end{equation*}

De manera semejante a lo que se consideró para las funciones, la condición necesaria para un funcional $Y$ alcanzar el valor extremo es que su primera variación sea nula.

Como $\lambda$ es un parámetro que identifica cada una de las funciones admisibles, y para $\lambda=0$ se tiene que la función correspondiente coincide con la función exacta, la primera variación del funcional $\bar{Y}$ con relación a $\lambda$ vale:

$$\delta \bar{Y}=\frac{d\bar{Y}}{d\lambda}\delta \lambda$$

y de la condición de estacionariedad se obtiene:

<a id='Ec2_14'></a>
\begin{equation*}
\left( \frac{d\bar{Y}}{d\lambda}\right)_{\lambda=0}=0
\label{eq:Ec2_14} \tag{2.14}
\end{equation*}

o

<a id='Ec2_15'></a>
\begin{equation*}
\left( \frac{d\bar{Y}}{d\lambda}\right)_{\lambda=0}=\left(\int_{x_1}^{x_2}\frac{dF}{d\lambda}dx\right)_{\lambda=0}=\int_{x_1}^{x_2}\left(\frac{\partial F}{\partial \bar{y}}\frac{d\bar{y}}{d\lambda}+\frac{\partial F}{\partial \bar{y}'}\frac{d\bar{y}'}{d\lambda}\right)dx
\label{eq:Ec2_15} \tag{2.15}
\end{equation*}

de la [Ec. 2.5](#Ec2_5) se puede escribir:

$$\frac{d\bar{y}}{d\lambda}=\eta \quad \text{y} \quad \frac{d\bar{y}'}{d\lambda}=\eta'$$

Con estos valores, la [Ec. 2.15](#Ec2_15) queda:

<a id='Ec2_16'></a>
\begin{equation*}
\left( \int_{x_1}^{x_2}\left(\frac{\partial F}{\partial \bar{y}}\eta+\frac{\partial F}{\partial \bar{y}'}\eta'\right)dx\right)_{\lambda=0}=\int_{x_1}^{x_2}\left(\frac{\partial F}{\partial y}\eta +\frac{\partial F}{\partial y'}\eta'\right)dx 
\label{eq:Ec2_16} \tag{2.16}
\end{equation*}

Integrando por partes el segundo término de la igualdad en la [Ec. 2.16](#Ec2_16), se llega a:

<a id='Ec2_17'></a>
\begin{equation*}
\left(\frac{dY}{d\lambda}\right)_{\lambda=0}=\int_{x_1}^{x_2}\left[\frac{\partial F}{\partial y}\eta-\frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right)\eta \right]dx+\left.\frac{\partial F}{\partial y'}\eta \right|_{x_1}^{x_2}
\label{eq:Ec2_17} \tag{2.17}
\end{equation*}

El último término en la [Ec. 2.17](#Ec2_17) es nulo ya que $\eta(x)=0$ para $x=x_1$ y $x=x_2$. Con esto:

<a id='Ec2_18'></a>
\begin{equation*}
\left(\frac{dY}{d\lambda}\right)_{\lambda=0}=\int_{x_1}^{x_2}\left[\frac{\partial F}{\partial y}-\frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right) \right]\eta dx=0
\label{eq:Ec2_18} \tag{2.18}
\end{equation*}

Válida para cualquier función admisible $\eta$.

Se puede demostrar que el integrando en la [Ec. 2.18](#Ec2_18) es nulo y determina una condición de extremo del funcional. Para esto, considere que para que el funcional tenga un valor máximo o mínimo, es necesario que tenga el mismo signo para cualquier variación de $\delta y$ en el intervalo dado. Asi, $\delta y$ en cualquier posición $x$ puede ser igual a $\pm k$, donde $k$ es un número pequeño. Para que esto sea posible, la expresión dentro de los corchetes de la [Ec. 2.18](#Ec2_18) tiene qué ser nula.

Introduciendo una notación similar a la usada en el estudio de extremos de una función, la [Ec. 2.13](#Ec2_13) puede ser escrita como:

<a id='Ec2_19'></a>
\begin{equation*}
\bar{Y}=Y+\delta Y + \frac{1}{2!}\delta^2 Y+\ldots
\label{eq:Ec2_19} \tag{2.19}
\end{equation*}

con

<a id='Ec2_20'></a>
\begin{equation*}
\delta Y=\lambda\left(\frac{dY}{d\lambda}\right)_{\lambda=0}; \quad \delta^2Y=\lambda^2\left(\frac{d^2Y}{d\lambda^2}\right)_{\lambda=0}
\label{eq:Ec2_20} \tag{2.20}
\end{equation*}

Con esto, la [Ec. 2.18](#Ec2_18), que es una primera variación del funcional queda:

<a id='Ec2_21'></a>
\begin{equation*}
\delta Y=\lambda \int_{x_1}^{x_2}\left[\frac{\partial F}{\partial y} -\frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right)\eta dx \right]=\int_{x_1}^{x_2}\left[\frac{\partial F}{\partial y} -\frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right)\right]\delta y dx =0
\label{eq:Ec2_21} \tag{2.21}
\end{equation*}

Recordando que $\delta y=\lambda \eta$ es la variación de la función $y$. 

Con esto, queda demostrado que el valor estacionario de un funcional es obtenido anulando su primera variación.

La notación empleada en la [Ec. 2.19](#Ec2_19) es denominada *notación variacional*, y es útil porque puede trabajarse de la misma forma con que se trabajan los diferenciales.

La ecuación:

<a id='Ec2_22'></a>
\begin{equation*}
\frac{\partial F}{\partial y}-\frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right)=0
\label{eq:Ec2_22} \tag{2.22}
\end{equation*}

se conoce como ecuación de [*Euler-Lagrange*](https://en.wikipedia.org/wiki/Euler%E2%80%93Lagrange_equation). Es una condición necesaria para que $y(x)$ sea una función extrema.

Existen funcionales más complejos que dependen de diversas funciones, variables y derivadas de mayor orden. Como ejemplo, se puede citar el funcional que representa la energía interna de una placa para el estado plano de tensiones:

$$W=\frac{Eh}{2(1-\nu^2)}\int\int\left\{ \left( \frac{\partial u}{\partial x}\right)^2+\left( \frac{\partial v}{\partial y}\right)^2+2\nu \frac{\partial u}{\partial x} \frac{\partial u}{\partial y}+\frac{1-\nu}{2}\left[\left( \frac{\partial u}{\partial y}\right)^2+\left( \frac{\partial v}{\partial x}\right)^2+2\frac{\partial u}{\partial y}\frac{\partial v}{\partial x}\right]\right\}dxdy$$

donde $h$ es el espesor de la placa, $E$, el [módulo de Young](https://en.wikipedia.org/wiki/Young%27s_modulus), $\nu$, el [coeficiente de Poisson](https://en.wikipedia.org/wiki/Poisson%27s_ratio), y $u$ y $v$, las componentes de los desplazamientos en la dirección de los ejes cartesianos $x$ y $y$, respectivamente, que definen el plano medio de la placa. 

Para funcionales del tipo:

<a id='Ec2_23'></a>
\begin{equation*}
Y=\int_{x_1}^{x_2} F\left(x,y,y',y'', \ldots, y^{(n)}\right)dx
\label{eq:Ec2_23} \tag{2.23}
\end{equation*}

La ecuación de *Euler-Lagrange* es de la forma:

<a id='Ec2_24'></a>
\begin{equation*}
\frac{\partial F}{\partial y}-\frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right)+\frac{d^2}{dx^2}\left(\frac{\partial F}{\partial y''}\right)-\ldots (-1)^n \frac{d^n}{dx^n}\left(\frac{\partial F}{\partial y^{(n)}}\right)=0
\label{eq:Ec2_24} \tag{2.24}
\end{equation*}

### Ejemplo: Viga prismática

<a id='Fig_2'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C02_F03_VigaPrismatica.PNG?raw=true" width="250" />
</p>

<div style="text-align: center"><strong>Figura 3.</strong>Viga prismática. <strong>Fuente:</strong> <a href="https://editoraunicamp.com.br/produto/503/metodo-dos-elementos-finitos-primeiros-passos-ed--3">Assan, Método dos elementos finitos - Primeiros passos. </a></div>

Considerese la viga prismática de la figura. El funcional que corresponde a este problema está dado por:

<a id='Ec2_25'></a>
\begin{equation*}
Y=\frac{1}{2}\int_0^l EI \left[ \frac{d^2v}{dx^2}\right]^2 dx - \int_0^l qvdx
\label{eq:Ec2_25} \tag{2.25}
\end{equation*}

donde $E$ es el módulo de Young del material e $I$ el momento de inercia de la sección transversal de la viga en relación al eje baricéntrico. El funcional definido en la [Ec. 2.25](#Ec2_25) es del tipo presentado en la [Ec. 2.23], donde:

$$F=\frac{1}{2}EI\left(v''\right)^2-qv$$

y la ecuación de *Euler-Lagrange* para el problema analizado, considerando la [c. 2.24](#Ec2_24) es:

<a id='Ec2_26'></a>
\begin{equation*}
\frac{d^2}{dx^2}\left(\frac{\partial F}{\partial v''}\right)-\frac{d}{dx}\left(\frac{\partial F}{\partial v'}\right)+\frac{\partial F}{\partial v}=0
\label{eq:Ec2_26} \tag{2.26}
\end{equation*}

Sustituyendo las derivadas de $F$ en la [Ec. 2.26](#Ec2_26) la ecuación de *Euer-Lagrange* asume la forma de:

$$\frac{d^2}{dx^2}\left(EIv''\right)-\frac{d}{dx}(0)-q=0$$

o

<a id='Ec2_27'></a>
\begin{equation*}
EI\frac{d^4v}{dx^4}-q=0
\label{eq:Ec2_27} \tag{2.27}
\end{equation*}

Esta es la [ecuación de equilibrio de la viga](https://en.wikipedia.org/wiki/Euler%E2%80%93Bernoulli_beam_theory). La solución de esta ecuación diferencial, con condiciones de contorno, permite encontrar la función $v(x)$ que representa las deformaciones del eje baricéntrico de la viga.

### Forma fuerte y débil

Cuando el problema es representado por ecuaciones diferenciales, como la [Ec. 2.27](#Ec2_27), se dice que está en la *forma fuerte*, y cuando es representada por una ecuación integral, como el funcional dado por la [Ec. 2.25](#Ec2_25), que contiene la ecuación diferencial, es llamada de [*forma débil*](https://en.wikipedia.org/wiki/Weak_formulation).

Esta distinción es realizada para resaltar el hecho de que, en la *forma fuerte*, las condiciones establecidas en la caracterización del problema analisado son satisfechas en todos los puntos del dominio (*solución continua*), mientras que, en la *forma débil*, solo en algunos puntos determinados (*solución discreta*).

El funcional representado por la [Ec. 2.25](#Ec2_25) es igual a la energía potencial total para una viga con carga uniformemente distribuida $q$ a lo largo de su longitud, $l$. Es formado por dos partes:

- La primera corresponde a la [energía de deformación](https://es.wikipedia.org/wiki/Energ%C3%ADa_de_deformaci%C3%B3n) $U$,

$$U=\frac{1}{2}\int_0^l\left(\frac{d^2v}{dx^2}\right)^2dx$$

- La segunda representa la energía potencial de las cargas $q$, 

$$\Omega = -\int_0^l qvdx$$

La energía de deformación $U$ puede escribirse en función de la energía de deformación específica volumétrica $u$

$$U=\int_V udV$$

donde $V$ es un folumen infinitesimal del cuerpo en estudio.

La energía de deformación especifica volumétrica depende del estado de deformaciones, y de tensiones, en cada punto del sólido. Por lo tanto, la energía potencial total también depende del estado de deformaciones que se designa como $\epsilon$.

Para demostrar que la energía de deformación total pasa por un mínimo para la configuración de equilibrio sobre la carga $q$, se considera la diferencia entra la energía potencial total correspondiente a la configuración de equilibrio cuyo campo de desplazamientos es $v$, y de deformaciones $\epsilon$ y de una configuración admisible cerca a la anterior $v+\delta v$ y $\epsilon + \delta \epsilon$:

<a id='Ec2_28'></a>
\begin{equation*}
Y_{\epsilon+\delta \epsilon} - Y_{\epsilon} = \int_V \left( u_{\epsilon+\delta \epsilon} - u_{\epsilon}\right)dV-\int_0^l q \delta v dx
\label{eq:Ec2_28} \tag{2.28}
\end{equation*}

la expansión en series de Taylor del término $u_{\epsilon+\delta \epsilon}$ queda:

<a id='Ec2_29'></a>
\begin{equation*}
u_{\epsilon+\delta \epsilon}= u_{\epsilon} + \frac{\partial u}{\partial \epsilon} \delta \epsilon+\frac{1}{2!} \frac{\partial^2 u}{\partial \epsilon^2} \left(\delta \epsilon\right)^2 + \ldots
\label{eq:Ec2_29} \tag{2.29}
\end{equation*}

sustituyendo la [Ec. 2.29](#Ec2_29) en la [Ec. 2.28](#Ec2_28), se llega a 

<a id='Ec2_30'></a>
\begin{equation*}
Y_{\epsilon+\delta \epsilon} - Y_{\epsilon} = \int_V \frac{\partial u}{\partial \epsilon} \delta \epsilon dV - \int_0^l q \delta v dx + \frac{1}{2!}\int_V \frac{\partial^2 u}{\partial \epsilon^2} \left(\delta \epsilon\right)^2 dV + \ldots
\label{eq:Ec2_30} \tag{2.30}
\end{equation*}

Las dos primeras integrales representan la variación de la energía potencial total que, para toda configuración admisible, es nula. La última igualdad se resume a la derivada segunda de la energía de deformación que será representada por la segunda variación de $Y$, como se hizo anteriormente en la [Ec. 2.13](#Ec2_13).

<a id='Ec2_31'></a>
\begin{equation*}
\delta ^2 Y = \frac{1}{2!}\int_V \frac{\partial^2 u}{\partial \epsilon^2} \left(\delta \epsilon\right)^2 dV
\label{eq:Ec2_31} \tag{2.31}
\end{equation*}

Considerando la serie de Taylor hasta el término de segundo orden en la [Ec. 2.29](#Ec2_29) y asumiendo que $\epsilon = 0$, lo que corresponde al estado de equilibrio, se tiene en dicha ecuación que el primer término del lado derecho es nulo ya que no se almacena energía si no hay deformaciones. La derivada que aparece en el segundo término es igual al estado de tensiones en un punto del sólido, que, para $\epsilon=0$, también es nulo. 

Queda solamente el término de la segunda derivada de la energía específica volumétrica que será dada por:

$$u_{\delta \epsilon}=\frac{1}{2!}\frac{\partial^2 u}{\partial \epsilon^2} \left(\delta \epsilon\right)^2$$

y para $\epsilon=0$ se tiene

$$\delta^2Y=\int_V u_{\delta \epsilon}dV$$

Como la energía de deformación específica $u$ es una función definida positiva de la deformación, la variación del funcional energía potencial total es positiva. Siendo la segunda variación de $Y$ positiva, la energía potencial total es un mínimo para el estado de equilibrio con $\epsilon = 0$.

De aquí en adelante se considerará que la condición de estacionariedad de la energía potencial total es mínima para cualquier configuración de equilibrio, con cualquier estado de deformaciones.
