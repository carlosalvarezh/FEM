## Aplicación del MEF a la ecuación de Poisson en 3D

### Introducción

En este capítulo se tratará la resolución por el MEF de la ecuación estacionaria de Poisson en tres dimensiones. De este ìnodo se concluirá con el estudio de la ecuación de Poisson y simultáneamente con ello, de problemas en los que interviene sólo una variable por nodo.


### Ecuación estacionaria de Poisson en 3D

La generalización a 3D de la ecuación de Poisson estacionaria se expresa como

<a id='Ec8_1'></a>
\begin{equation*}
A(\phi):=\frac{\partial}{\partial x}\left(k_{x} \frac{\partial \phi}{\partial x}\right)+\frac{\partial}{\partial y}\left(k_{y} \frac{\partial \phi}{\partial y}\right)+\frac{\partial}{\partial z}\left(k_{z} \frac{\partial \phi}{\partial z}\right)+Q=0 \quad \text { en } \Omega
\label{eq:Ec8_1} \tag{8.1}
\end{equation*}

siendo $\phi(x, y, z)$ la función incógnita, $k_{x}, k_{y}$ y $k_{z}$ las conductividades en las direcciones $x, y, z$ y $Q$ el flujo generado en el dominio por unidad de volumen.
Esta ecuación surge, como en los casos de 1D y 2D, del balance del flujo en un elemento diferencial (Figura 8.1).

<a id='Fig8_1'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C08_F01_Problema3D.PNG?raw=true" width="450" />
</p>
<div style="text-align: center"><strong>Figura 8.1.</strong>Definición del problema de Poisson en 3D, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

Las condiciones de contorno se definen por

<a id='Ec8_2'></a>
\begin{equation*}
B(\phi):\left\{\begin{array}{l}
\begin{array}{c}
\text { Condición de } \phi \text { prescrita (Dirichlet): } \\
\phi-\bar{\phi}=0 \quad \text { en } \Gamma_{\phi}
\end{array} \\
\begin{array}{l}
\text { Condición de flujo prescrito (Neumann): } \\
k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}+k_{z} \frac{\partial \phi}{\partial z} n_{z}+\bar{q}_{n}+\alpha\left(\phi-\bar{\phi}_{a}\right)=0 \quad \text { en } \Gamma_{q}
\end{array}
\end{array}\right.
\label{eq:Ec8_2} \tag{8.2}
\end{equation*}

Como en el caso bidimensional $\bar{q}_{n}$ es el flujo prescrito que sale en dirección normal al contorno $\Gamma_{q}, \phi_{a}$ la temperatura ambiente, $\alpha$ el coeficiente de convección y $\mathbf{n}=\left[n_{x}, n_{y}, n_{z}\right]^{T}$ la normal unitaria a $\Gamma_{q}$. Asimismo $\Gamma=\Gamma_{\phi} \cup \Gamma_{q}$.


### Resolución por el método de los elementos finitos

#### Forma integral débil

La forma integral equivalente a las ecuaciones [8.1](#Ec8_1) y [8.2](#Ec8_2) se obtiene por aplicación del $MRP$

<a id='Ec8_3'></a>
\begin{equation*}
\begin{aligned}
&\iiint_{\Omega} W\left[\frac{\partial}{\partial x}\left(k_{x} \frac{\partial \phi}{\partial x}\right)+\frac{\partial}{\partial y}\left(k_{y} \frac{\partial \phi}{\partial y}\right)+\frac{\partial}{\partial z}\left(k_{z} \frac{\partial \phi}{\partial z}\right)+Q\right] \mathrm{d}  \Omega+ \\
&+\iint_{\Gamma_{q}} \bar{W}\left[k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}+k_{z} \frac{\partial \phi}{\partial z} n_{z}+\bar{q}_{n}+\alpha\left(\phi-\phi_{a}\right)\right] \mathrm{d}  \Gamma=0
\end{aligned}
\label{eq:Ec8_3} \tag{8.3}
\end{equation*}

siendo $W$ y $\bar{W}$ funciones de peso arbitrarias.
Integrando por partes los términos en derivadas segundas de $\phi$ en la primera integral y escogiendo $\bar{W}=-W$ se obtiene

<a id='Ec8_4'></a>
\begin{equation*}
\begin{aligned}
&-\iiint_{\Omega}\left(\frac{\partial W}{\partial x} k_{x} \frac{\partial \phi}{\partial x}+\frac{\partial W}{\partial y} k_{y} \frac{\partial \phi}{\partial y}+\frac{\partial W}{\partial z} k_{x} \frac{\partial \phi}{\partial z}\right) \mathrm{d} \Omega+ \\
&+\iiint_{\Omega} W Q \mathrm{d} \Omega+\iint_{\Gamma_{e}+\Gamma_{t}} W\left(k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}+k_{x} \frac{\partial \phi}{\partial z} n_{x}\right) \mathrm{d} \Gamma \\
&-\iint_{\Gamma_{q}} W\left[k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}+k_{x} \frac{\partial \phi}{\partial z} n_{x}+\bar{q}_{n}+\alpha\left(\phi-\phi_{a}\right)\right] \mathrm{d} \Gamma=0
\end{aligned}
\label{eq:Ec8_4} \tag{8.4}
\end{equation*}

donde los términos marcarlos se cancelan mutuamente.

Reordenando términos se obtiene la forma débil

<a id='Ec8_5'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\iiint_{\Omega}\left[\frac{\partial W}{\partial x} k_{x} \frac{\partial \phi}{\partial x}+\frac{\partial W}{\partial y} k_{y} \frac{\partial \phi}{\partial y}+\frac{\partial W}{\partial z} k_{x} \frac{\partial \phi}{\partial z}\right] d \Omega+\iint_{\Gamma_{p}} W \alpha \phi \mathrm{d} \Gamma= \\
=\iiint_{\Omega} W Q \mathrm{d} \Omega-\iint_{\Gamma_{q}} W\left[\bar{q}_{n}-\alpha \phi_{a}\right] \mathrm{d} \Gamma-\iint_{\Gamma_{\phi}} W q_{n} \mathrm{d} \Gamma
\end{gathered}
\end{aligned}
\label{eq:Ec8_5} \tag{8.5}
\end{equation*}

En la ecuación anterior se ha hecho uso de la definición de flujo normal

<a id='Ec8_6'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
q_{n}=\mathbf{q}^{T} \mathbf{n}=q_{x} n_{x}+q_{y} n_{y}+q_{x} n_{x}=-k_{x} \frac{\partial \phi}{\partial x} n_{x}-k_{y} \frac{\partial \phi}{\partial y} n_{y}-k_{z} \frac{\partial \phi}{\partial z} n_{z}
\end{gathered}
\end{aligned}
\label{eq:Ec8_6} \tag{8.6}
\end{equation*}

La forma débil es el punto de partida para la solución por el MEF.


#### Discretización en elementos finitos

<a id='Fig8_2'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C08_F02_Solidos3D.PNG?raw=true" width="450" />
</p>
<div style="text-align: center"><strong>Figura 8.2.</strong>Sólidos 3D discretizados, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

Realicemos una discretización del dominio tridimensional en elementos finitos de $n$ nodos (Figura 8.2). La aproximación en el interior de cada elemento se escribe en la forma usual como

<a id='Ec8_7'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\phi \cong \hat{\phi}=\sum_{j=1}^{n} N_{j}^{(e)}(x, y, z) \phi_{j}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec8_7} \tag{8.7}
\end{equation*}

donde $N_{j}^{(e)}$ son las funciones de forma del elemento tridimensional y $\phi_{j}^{(e)}$ los valores nodales de la función aproximada.

Sustituyendo la aproximación anterior en la forma débil y escogiendo un número de funciones de peso igual al de incógnitas nodales, se obtiene la expresión aproximada de la forma débil por

<a id='Ec8_8'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
&\iiint_{\Omega}\left[\frac{\partial W_{i}}{\partial x} k_{x} \frac{\partial \hat{\phi}}{\partial x}+\frac{\partial W_{i}}{\partial y} k_{y} \frac{\partial \hat{\phi}}{\partial y}+\frac{\partial W_{i}}{\partial z} k_{z} \frac{\partial \hat{\phi}}{\partial z}\right] \mathrm{d} \Omega+\iint_{\Gamma_{q}} W_{i} \alpha \hat{\phi} \mathrm{d} \Gamma= \\
&\quad=\iiint_{\Omega} W_{i} \alpha  \hat{\phi} \mathrm{d} \Gamma-\iint_{\Gamma_{q}} W_{i}\left[\bar{q}_{n}-\alpha \phi_{a}\right] \mathrm{d} \Gamma-\iint_{\Gamma_{\phi}} W_{i} q_{n} \mathrm{d} \Gamma \\
&i=1,2, \ldots, N
\end{gathered}
\end{aligned}
\label{eq:Ec8_8} \tag{8.8}
\end{equation*}

donde $N$ es el número total de nodos en la malla.

La forma simétrica de $Galerkin$ se obtiene haciendo $W_{i}=N_{\mathrm{i}}$. Suponiendo que todas las funciones que intervienen en la expresión anterior son integrables y haciendo uso de la propiedad aditiva de la integral, puede expresarse la forma débil como suma de las contribuciones elementales por

<a id='Ec8_9'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
&\sum_{e}\left\{\int \int \int _ { \Omega ^ { ( e ) } } \left[\frac{\partial N_{i}^{(e)}}{\partial x} k_{x}\left(\sum_{j=1}^{n} \frac{\partial N_{j}^{(e)}}{\partial x} \phi_{j}^{(e)}\right)+\frac{\partial N_{i}^{(e)}}{\partial y} k_{y}\left(\sum_{j=1}^{n} \frac{\partial N_{j}^{(e)}}{\partial y} \phi_{j}^{(e)}\right)+\right.\right. \\
&\left.\left.+\frac{\partial N_{i}^{(e)}}{\partial z} k_{z}\left(\sum_{j=1}^{n} \frac{\partial N_{j}^{(e)}}{\partial z} \phi_{j}^{(e)}\right)\right] \mathrm{d} \Omega+\iint_{\Gamma_{q}^{(e)}} N_{i}^{(e)} \alpha\left(\sum_{j=1}^{n} N_{j}^{(e)} \phi_{j}^{(e)}\right) \mathrm{d} \Gamma\right\}= \\
&\quad=\sum_{e}\left\{\iiint_{\Omega^{(e)}} N_{i}^{(e)} Q \mathrm{d} \Omega-\iint_{\Gamma_{q}^{(e)}} N_{i}^{(e)}\left(\bar{q}-\alpha \phi_{a}\right) \mathrm{d} \Gamma-\iint_{\Gamma_{\phi}^{(e)}} N_{i}^{(e)} q \mathrm{d} \Gamma\right\} . \\
&i=1,2, \ldots, N
\end{gathered}
\end{aligned}
\label{eq:Ec8_9} \tag{8.9}
\end{equation*}

donde el sumatorio se extiende sobre todos los elementos de la malla.

La expresión integral anterior puede interpretarse como la fila $i$-ésima del sistema de ecuaciones algebraicas resultante de la discretización. Es decir

<a id='Ec8_10'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
K_{i 1} \phi_{1}+K_{i 2} \phi_{2}+\ldots+K_{i N} \phi_{N}=f_{i}, \quad i=1,2, \ldots, N
\end{gathered}
\end{aligned}
\label{eq:Ec8_10} \tag{8.10}
\end{equation*}

El sistema de ecuaciones global se escribe en la forma usual

$$\mathbf{Ka}=\mathbf{f}$$

o bien

<a id='Fig8_3'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C08_F03_SEL3D.PNG?raw=true" width="350" />
</p>

Puede comprobarse que, como en casos anteriores, la matriz $\mathbf{K}$ y el vector $\mathbf{f}$ se obtienen ensamblando las contribuciones de las matrices de rigidez elementales $\mathbf{K}^{(e)}$ y los vectores de flujos nodales elementales $\mathbf{f}^{(e)}$ dados por

<a id='Ec8_12'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{K}^{(e)}=\left[\begin{array}{ccc}
K_{11}^{(e)} & \ldots & K_{1 n}^{(e)} \\
\vdots & \ddots & \vdots \\
K_{n 1}^{(e)} & \ldots & K_{n n}^{(e)}
\end{array}\right]
\end{gathered}
\end{aligned}
\label{eq:Ec8_12} \tag{8.12}
\end{equation*}

<a id='Ec8_13'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\mathbf{f}_{\mathbf{1}}^{(e)}=\left\{\begin{array}{c}
f_{1}^{(e)} \\
\vdots \\
f_{n}^{(e)}
\end{array}\right\}
\end{gathered}
\end{aligned}
\label{eq:Ec8_13} \tag{8.13}
\end{equation*}

siendo $n$ el número de nodos del elementó $\mathrm{y}$

<a id='Ec8_14'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
K_{i j}^{(e)}=& \overbrace{\iint_{\Omega^{(e)}}\left(\frac{\partial N_{i}^{(e)}}{\partial x} k_{x} \frac{\partial N_{j}^{(e)}}{\partial x}+\frac{\partial N_{i}^{(e)}}{\partial y} k_{y} \frac{\partial N_{j}^{(e)}}{\partial y}+\frac{\partial N_{i}^{(e)}}{\partial z} k_{z} \frac{\partial N_{j}^{(e)}}{\partial z}\right) \mathrm{d} \Omega+}^{K_{d_{ij}}^{(e)}} \\
&+\underbrace{\iint_{\Gamma_{q}^{(e)}} \alpha N_{i}^{(e)} N_{j}^{(e)} \mathrm{d} \Gamma}_{K_{c_{i j}}^{(e)}}=k_{d_{i j}}^{(e)}+K_{c_{i j}}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec8_14} \tag{8.14}
\end{equation*}

siendo $K_{d_{i j}}^{(e)}$ el término de rigidez de conducción y $K_{c i j}^{(e)}$ el término de rigidez de convección. Este último término sólo aparecerá en aquellos contornos donde se prescriba la condición de contorno de convección.

Por otra parte

<a id='Ec8_15'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
f_{i}^{(e)}=\underbrace{\iiint_{\Omega^{(e)}} N_{i}^{(e)} Q \mathrm{d} \Omega}_{f_{Q_{i}}^{(e)}} \underbrace{-\iint_{\Gamma_{q}^{(e)}} N_{i}^{(e)}\left(\bar{q}-\alpha \phi_{a}\right) \mathrm{d} \Gamma}_{f_{c_{i}}^{(e)}}-q_{n_{i}}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec8_15} \tag{8.15}
\end{equation*}

con $f_{Q_{i}}^{(e)}$ es el término de flujo nodal asociado a $Q, f_{c_{i}}^{(c)}$ el término de flujo nodal de convección y $q_{n_{i}}^{(e)}=\iint_{\Gamma_{q}^{(e)}} N_{i}^{(e)} q_{n} \mathrm{d} \Gamma$ el flujo reacción en el nodo donde el valor de $\phi$ esté prescrito. De nuevo el término $f_{c_{i}}^{(e)}$ sólo interviene en nodos pertenecientes a los contornos elementales coincidentes con contornos exteriores de *Neumann*.


### Elemento tetraédrico de cuatro nodos

<a id='Fig8_3'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C08_F04_Tetraedro4nodos.PNG?raw=true" width="350" />
</p>
<div style="text-align: center"><strong>Figura 8.3.</strong>Tetraedro lineal de 4 nodos, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

El tetraedro lineal es el elemento tridimensional más sencillo, tal como el triángulo de tres nodos lo es entre los elementos bidimensionales. Asimismo, la familia tetraédrica presenta similares propiedades a las de la familia triangular.
$\mathrm{La}$ aproximación se escribe dentro de cada tetraedro por el polinomio lineal siguiente

<a id='Ec8_16'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\phi^{(e)}=\alpha_{0}+\alpha_{1} x+\alpha_{2} y+\alpha_{3} z
\end{gathered}
\end{aligned}
\label{eq:Ec8_16} \tag{8.16}
\end{equation*}

De la expresión anterior se deduce

<a id='Ec8_17'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\phi_{1}^{(e)} &=\alpha_{0}+\alpha_{1} x_{2}^{(e)}+\alpha_{2} y_{1}^{(e)}+\alpha_{3} z_{1}^{(e)} \\
\phi_{2}^{(e)} &=\alpha_{0}+\alpha_{1} x_{2}^{(e)}+\alpha_{2} y_{2}^{(e)}+\alpha_{3} z_{2}^{(e)} \\
\phi_{3}^{(e)} &=\alpha_{0}+\alpha_{1} x_{3}^{(e)}+\alpha_{2} y_{3}^{(e)}+\alpha_{3} z_{3}^{(e)} \\
\phi_{4}^{(e)} &=\alpha_{0}+\alpha_{1} x_{4}^{(e)}+\alpha_{2} y_{4}^{(e)}+\alpha_{3} z_{4}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec8_17} \tag{8.17}
\end{equation*}

Resolviendo el sistema anterior y sustituyendo los $\alpha_{i}$ en la [Ec. 8.16](#Ec8_16) se obtiene, tras agrupar términos

<a id='Ec8_18'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\phi=\sum_{i=1}^{4} N_{i}^{(e)}(x, y, z) \phi_{i}^{(e)}=N_{1}^{(e)}(x, y, z) \phi_{1}^{(e)}+\ldots+N_{4}^{(e)}(x, y, z) \phi_{4}^{(e)}
\end{gathered}
\end{aligned}
\label{eq:Ec8_18} \tag{8.18}
\end{equation*}

siendo

<a id='Ec8_19'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
N_{i}^{(e)}(x, y, z)=\frac{1}{6 \Omega^{(e)}}\left(a_{i}^{(e)}+b_{i}^{(e)}(x)+c_{i}^{(e)} y+d_{i}^{(e)} z\right)
\end{gathered}
\end{aligned}
\label{eq:Ec8_19} \tag{8.19}
\end{equation*}

con

<a id='Ec8_20'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
a_{i}^{(e)}=
\begin{bmatrix}
x_{j}^{(e)} & y_{j}^{(e)} & z_{j}^{(e)} \\
x_{k}^{(e)} & y_{k}^{(e)} & z_{k}^{(e)} \\
x_{l}^{(e)} & y_{l}^{(e)} & z_{l}^{(e)}
\end{bmatrix}
; \quad \quad
b_{i}^{(e)}=
-\begin{bmatrix}
1 & y_{j}^{(e)} & z_{j}^{(e)} \\
1 & y_{k}^{(e)} & z_{k}^{(e)} \\
1 & y_{l}^{(e)} & z_{l}^{(e)}
\end{bmatrix}
\\
\\
c_{i}^{(e)}=
\begin{bmatrix}
x_{j}^{(e)} & 1 & z_{j}^{(e)} \\
x_{k}^{(e)} & 1 & z_{k}^{(e)} \\
x_{l}^{(e)} & 1 & z_{l}^{(e)}
\end{bmatrix}
; \quad \quad
d_{i}^{(e)}=
-\begin{bmatrix}
x_{j}^{(e)} & y_{j}^{(e)} & 1 \\
x_{k}^{(e)} & y_{k}^{(e)} & 1 \\
x_{l}^{(e)} & y_{l}^{(e)} & 1
\end{bmatrix}
\end{gathered}
\end{aligned}
\label{eq:Ec8_20} \tag{8.20}
\end{equation*}

el resto de las constantes se obtiene rotando cíclicamente los subíndices $(i, j, k, l)$ aplicando la regla de la mano derecha.

El volumen del elemento $\Omega^{(e)}$ se halla a partir del determinante

<a id='Ec8_21'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
6 \Omega^{(e)}=\left|\begin{array}{llll}
1 & x_{i}^{(e)} & y_{i}^{(e)} & z_{i}^{(e)} \\
1 & x_{j}^{(e)} & y_{j}^{(e)} & z_{j}^{(e)} \\
1 & x_{k}^{(e)} & y_{k}^{(e)} & z_{k}^{(e)} \\
1 & x_{l}^{(e)} & y_{i}^{(e)} & z_{l}^{(e)}
\end{array}\right|
\end{gathered}
\end{aligned}
\label{eq:Ec8_21} \tag{8.21}
\end{equation*}

Ya no es posible representar gráficamente las $N_{i}^{(e)}$ como se hacía en los problemas planos. No obstante, se sigue cumpliendo que la función de forma de un nodo toma un valor igual a la unidad en el nodo y cero en los restantes.

Otra dificultad que suele presentarse en cuerpos volumétricos es la de visualizar su subdivisión en tetraedros, lo que no pocas veces conduce a errores en la numeración. Hay metodologias que tienden a evitar tales errores, tal como la subdivisión previa en elementos paralelepipédicos, visualmente más sencillos y, "a posteriori", descomponer cada uno de ellos en varios tetraedros de modo automático.

De la expresión de las funciones de forma se tiene

<a id='Ec8_22'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\frac{\partial N_{i}^{(e)}}{\partial x}=\frac{b_{\mathrm{i}}}{6 \Omega^{(e)}} ; \quad \frac{\partial N_{i}}{\partial y} = \frac{c_{i}}{6 \Omega^{(e)}} ; \quad \frac{\partial N_{i}}{\partial z}=\frac{d_{i}}{6 \Omega^{(e)}}
\end{gathered}
\end{aligned}
\label{eq:Ec8_22} \tag{8.22}
\end{equation*}

Sustituyendo en la expresión de $\mathbf{K}_{i j}^{(e)}$, se obtiene

<a id='Ec8_23'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
K_{d_{i j}}^{(e)}=\frac{1}{36 \Omega^{(e)}}\left(k_{x} b_{i} b_{j}+k_{y} c_{i} c_{j}+k_{z} d_{i} d_{j}\right)
\end{gathered}
\end{aligned}

\label{eq:Ec8_26} \tag{8.26}
