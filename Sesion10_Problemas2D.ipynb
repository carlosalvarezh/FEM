## Aplicación del MEF a la ecuación de Poisson en 2D

### Introducción

En este capítulo se describirá la aplicación del MEF a la solución de la ecuación de Poisson en dominios bidimensionales.

Tras la presentación de las ideas fundamentales se presenta la solución por el MEF utilizando el sencillo elemento triangular de tres nodos.


### Ecuación estacionaria de Poisson en 2D

La ecuación bidimensional de Pöisson en régimen estacionario es

<a id='Ec7_1'></a>
\begin{equation*}
A(\phi):=\frac{\partial}{\partial x}\left(k_{x} \frac{\partial \phi}{\partial x}\right)+\frac{\partial}{\partial y}\left(k_{y} \frac{\partial \phi}{\partial y}\right)+Q=0 \quad \text { en } \boldsymbol{\Omega}
\label{eq:Ec7_1} \tag{7.1}
\end{equation*}

siendo $\phi(x, y)$ la función incógnita. $k_{x}$ y $k_{y}$ son las conductividades en direcciones $x$ e $y$ respectivamente y $Q(x, y)$ es una fuente distribuida sobre el dominio $\Omega$.

Dicha ecuación se puede obtener planteando el balance de flujos para un elemento deferencial (Figura 7.1).

<a id='Fig7_1'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C07_F01_DefinicionProblemaPoisson2D.PNG?raw=true" width="450" />
</p>
<div style="text-align: center"><strong>Figura 7.1.</strong>Definición del problema de Poisson en 2D, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

Las condiciones de contorno pueden ser de dos tipos

<a id='Ec7_2a'></a>
\begin{equation*}
B(\phi):\left\{\begin{array}{c}
\text { Condición de variable prescrita (Dirichlet) } \\
\phi-\bar{\phi}=0 \quad \text { en } \Gamma_{\phi} \\
\text { Condición de flujo normal prescrito (Neumann) } \\
q_{n}=\bar{q}_{n}+\alpha\left(\phi-\phi_{a}^{\prime}\right)
\end{array}\right.
\label{eq:Ec7_2a} \tag{7.2a}
\end{equation*}

El flujo normal $q_{n}$ se obtiene proyectando el flujo en el contorno sobre la normal. Así

$$
q_{n}=\mathbf{n}^{T} \mathbf{q}=n_{x} q_{x}+n_{y} q_{y}
$$

Sustituyendo $q_{x}=-k_{x} \frac{\partial \phi}{\partial x}$ y $q_{y}=-k_{y} \frac{\partial \phi}{\partial y}$ en las ecuaciones anteriores se obtiene la condición de flujo prescrito en el contorno (condición de Neumann) en la forma usual siguiente

<a id='Ec7_2b'></a>
\begin{equation*}
k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}+\bar{q}_{n}+\alpha\left(\phi-\bar{\phi}_{a}\right)=0 \quad \text { en } \Gamma_{q}
\label{eq:Ec7_2b} \tag{7.2b}
\end{equation*}

En lo anterior $\bar{q}_{n}$ es el flujo prescrito en dirección normal al contorno $\Gamma_{q}$ ( $\bar{q}_{n}$ es positivo, si el flujo es en dirección de la normal à contorno $\left.\mathbf{n}\right)$. El último término de la [Ec. 7.2b](#Ec7_2b) expresa el flujo de calor que sale por el contorno debido a una diferencia de temperatura entre los puntos del contorno y el ambiente a temperatura $\phi_{a} .$ Asimismo, $\alpha$ es el denominado coeficiente de convección.

Adviértase que

<a id='Ec7_3'></a>
\begin{equation*}
\Gamma=\Gamma_{\phi} \cup \Gamma_{q}
\label{eq:Ec7_3} \tag{7.3}
\end{equation*}

La división de $\Gamma$ en dos zonas $\Gamma_{\phi}$ y $\Gamma_{q}$ es meramente conceptual, puesto que puede darse la situación de que estas condiciones se vayan alternando en los sucesivos puntos del contorno.

### Resolución por el método de los elementos finitos

Resolveremos el problema definido por las [Ecs. 7.1 y 7.2](#Ec7_1) en modo similar al caso unidimensional, aplicando el método de *residuos ponderados tipo Galerkin*.

#### Forma integral del MRP

La forma integral equivalente a las ecuaciones diferenciales se obtiene por el método de residuos ponderados (MRP). Así, multiplicando las ecs. [7.1](#Ec7_1) y [7.2](#Ec7_2) por sendas funciones de peso $W$ y $\bar{W}$ e integrando sobre el dominio y el contorno $\Gamma_{q}$, respectivamente, se obtiene

<a id='Ec7_4'></a>
\begin{equation*}
\begin{aligned}
&\iint_{\Omega} W\left[\frac{\partial}{\partial x}\left(k_{x} \frac{\partial \phi}{\partial x}\right)+\frac{\partial}{\partial y}\left(k_{y} \frac{\partial \phi}{\partial y}\right)+Q\right] \mathrm{d} x \mathrm{d} y+ \\
&+\int_{\Gamma_{q}} W\left[k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}+\bar{q}_{n}+\alpha\left(\phi-\phi_{a}\right)\right] \mathrm{d} T=0
\end{aligned}
\label{eq:Ec7_4} \tag{7.4}
\end{equation*}

La forma débil se obtiene integrando por partes los términos en derivadas segundas en [Ec. 7.4](#Ec7_4). Así

<a id='Ec7_5'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\iint_{\Omega} W \frac{\partial}{\partial x}\left(k_{x} \frac{\partial \phi}{\partial x}\right) \mathrm{d} x \mathrm{d} y=\int \mathrm{d} y \int W \frac{\partial}{\partial x}\left(k_{x} \frac{\partial \phi}{\partial x}\right) \mathrm{d} x= \\
=-\iint_{\Omega} \frac{\partial W}{\partial x} k_{x} \frac{\partial \phi}{\partial x} \mathrm{d} x \mathrm{d} y+\int_{\Gamma=\Gamma_{q}+\Gamma} W k_{x} \frac{\partial \phi}{\partial x} \mathrm{d} y
\end{gathered}
\end{aligned}
\label{eq:Ec7_5} \tag{7.5}
\end{equation*}

Realizando una operación similar para la integral de $\frac{\partial}{\partial y}\left(k_{x} \frac{\partial \phi}{\partial y}\right)$, se obtiene tras sustituir en [Ec. 7.4](#Ec7_4)

<a id='Ec7_6'></a>
\begin{equation*}
\begin{aligned}
&\iint_{\Omega} W\left[\frac{\partial}{\partial x}\left(k_{x} \frac{\partial \phi}{\partial x}\right)+\frac{\partial}{\partial y}\left(k_{y} \frac{\partial \phi}{\partial y}\right)+Q\right] \mathrm{d} x \mathrm{d} y= \\
&\quad=-\iint_{\Omega}\left[\frac{\partial W}{\partial x} k_{x} \frac{\partial \phi}{\partial x}+\frac{\partial W}{\partial y} k_{y} \frac{\partial \phi}{\partial y}\right] \mathrm{d} x \mathrm{d} y+\iint_{\Omega} W Q \mathrm{d} x \mathrm{d} y+ \\
&\quad+\int_{\Gamma_{q}+\Gamma_{\phi}} W\left[k_{x} \frac{\partial \phi}{\partial x} \mathrm{d} y+k_{y} \frac{\partial \phi}{\partial y} \mathrm{d} x\right]
\end{aligned}
\label{eq:Ec7_6} \tag{7.6}
\end{equation*}

<a id='Fig7_2'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C07_F02_VectorNormal.PNG?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 7.2.</strong>Definición de la normal al contorno, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

Teniendo en cuenta que (de la Figura 7.2)

<a id='Ec7_7'></a>
\begin{equation*}
\begin{aligned}
&\mathrm{d} y=\mathrm{d} \Gamma \cos \alpha=d \Gamma n_{x} \\
&\mathrm{d} x=\mathrm{d} \Gamma \cos \beta=d \Gamma n_{y}
\end{aligned}
\label{eq:Ec7_7} \tag{7.7}
\end{equation*}

y adoptando $W=-W$, tras sustituir en [Ec. 7.6](#Ec7_6), se llega a 

<a id='Ec7_8'></a>
\begin{equation*}
\begin{aligned}
&-\iint_{\Omega}\left[\frac{\partial W}{\partial x} k_{x} \frac{\partial \phi}{\partial x}+\frac{\partial W}{\partial y} k_{y} \frac{\partial \phi}{\partial y}\right] \mathrm{d} x \mathrm{d} y+\iint_{\Omega} W Q \mathrm{d} x \mathrm{d} y+ \\
&+\int_{\Gamma_{q}+\Gamma_{\phi}} W\left[k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}\right] \mathrm{d} \Gamma- \\
&-\int_{\Gamma_{q}} W\left[k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}+\bar{q}_{n}+\alpha\left(\phi-\phi_{a}\right)\right] \mathrm{d} \Gamma=0
\end{aligned}
\label{eq:Ec7_8} \tag{7.8}
\end{equation*}

en la que los términos subrayados se cancelan entre sí mutuamente.

La forma integral resultante es

<a id='Ec7_9'></a>
\begin{equation*}
\begin{aligned}
&-\iint_{\Omega}\left[\frac{\partial W}{\partial x} k_{x} \frac{\partial \phi}{\partial x}+\frac{\partial W}{\partial y} k_{y} \frac{\partial \phi}{\partial y}\right] \mathrm{d} \Omega+\iint_{\Omega} W Q \mathrm{d} \Omega+ \\
&\left.+\int_{\Gamma_{\phi}} W\left[k_{x} \frac{\partial \phi}{\partial x} n_{x}+k_{y} \frac{\partial \phi}{\partial y} n_{y}\right)\right] \mathrm{d} \Gamma-\int_{\Gamma_{\phi}}\left[\bar{q}_{n}+\alpha\left(\phi-\phi_{a}\right)\right] \mathrm{d} \Gamma=0
\end{aligned}
\label{eq:Ec7_9} \tag{7.9}
\end{equation*}

Teniendo en cuenta que

<a id='Ec7_10'></a>
\begin{equation*}
\begin{aligned}
q_{n}=q_{x} n_{x}+q_{y} n_{y}=-k_{x} \frac{\partial \phi}{\partial x} n_{x}-k_{y} \frac{\partial \phi}{\partial y} n_{y}
\end{aligned}
\label{eq:Ec7_10} \tag{7.10}
\end{equation*}

Agrupando los términos en los que interviene $\phi$ en el primer miembro y teniendo en cuenta la expresión de $q_{n}$ anterior, se obtiene finalmente

<a id='Ec7_11'></a>
\begin{equation*}
\begin{aligned}
&\iint_{\Omega}\left[\frac{\partial W}{\partial x} k_{x} \frac{\partial \phi}{\partial x}+\frac{\partial W}{\partial y} k_{y} \frac{\partial \phi}{\partial y}\right] \mathrm{d} \Omega+\int_{\Gamma_{q}} \alpha W \phi \mathrm{d} \Gamma= \\
&=\iint_{\Omega} W Q \mathrm{d} \Omega-\int_{\Gamma_{q}} W\left[\vec{q}_{n}-\alpha \phi_{a}\right] \mathrm{d} \Gamma-\int_{\Gamma_{\phi}} W q_{n} \mathrm{d} \Gamma
\end{aligned}
\label{eq:Ec7_11} \tag{7.11}
\end{equation*}

La forma débil [(Ec. 7.11)](#Ec7_11) es el punto de partida para obtener las ecuaciones de la aproximación por el MEF.

#### Discretización por el MEF

La última integral en [Ec. 7.11](#Ec7_11) se debe al flujo $q_{n}$ que sale por el contorno $\Gamma_{\phi}$ donde el valor de $\phi$ es conccido. Adviértase que el signo que precede la integral es negativo, lo que indica que el flujo normal $q_{n}$ sale del contorno. Como en el problema unidimensional, dicho flujo puede interpretarse como una "reacción" que se calcula una vez conocido el valor de $\phi$ sobre $\Gamma_{\phi}$.

<a id='Fig7_3'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C07_F03_DiscretizacionDominio2D.PNG?raw=true" width="350" />
</p>
<div style="text-align: center"><strong>Figura 7.3.</strong>Discretización de  un dominio 2D, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

Supongamos ahora una discretización del dominio $\Omega$ en elementos finitos bidimensionales de $n$ nodos (Figura 4.3). La aproximación de $\phi$ en el interior de cada elemento se define en la forma usual por

<a id='Ec7_12'></a>
\begin{equation*}
\begin{aligned}
\phi \cong \hat{\phi}=\sum_{i=1}^{n} N_{i}^{(e)}(x, y) \phi_{i}^{(e)}
\end{aligned}
\label{eq:Ec7_12} \tag{7.12}
\end{equation*}

donde $N_{i}^{(e)}(x, y)$ son las runciones de forma bidimensional y $\phi^{(e)}$ los valores nodales de la solución aproximada.

El sistema de ecuaciones de la discretización se obtiene sustituyendo la aproximación [Ec. 7.12](#Ec7_12) en la forma débil [Ec. 7.11](#Ec7_11) y escogiendo $N$ funciones de peso $W_{i} .$ Así

<a id='Ec7_13'></a>
\begin{equation*}
\begin{aligned}
&\iint_{\Omega}\left[\frac{\partial W_{i}}{\partial x} k_{x} \frac{\partial \hat{\phi}}{\partial x}+\frac{\partial W_{i}}{\partial y} k_{y} \frac{\partial \hat{\phi}}{\partial y}\right] \mathrm{d} \Omega+\int_{\Gamma_{q}} \alpha W_{i} \hat{\phi} \mathrm{d} \Omega= \\
&\quad=\iint_{\Omega} W_{i} Q \mathrm{d} \Omega-\int_{\Gamma_{q}} W_{i}\left[\bar{q}_{n}-\alpha \phi_{a}\right] \mathrm{d} \Gamma-\int_{\Gamma_{\phi}} W_{i} q_{n} \mathrm{d} \Gamma \quad i=1, \ldots, N
\end{aligned}
\label{eq:Ec7_13} \tag{7.13}
\end{equation*}

siendo $N$ el número total de nodos en la malla.

La forma de Galerkin se obtiene haciendo $W_{i}=N_{i}$. Suponiendo que todas las funciones que aparecen en [Ec. 7.13](#Ec7_13) son integrables y teniendo en cuenta la propiedad aditiva de la integral, puede expresarse la forma de Galerkin en función de las contribuciones elementales por

<a id='Ec7_14'></a>
\begin{equation*}
\begin{aligned}
&\sum_{e} \iint_{\Omega^{(e)}}\left[\frac{\partial N_{i}^{(e)}}{\partial x} k_{x}\left(\sum_{j=1}^{n} \frac{\partial N_{j}^{(e)}}{\partial x} \phi_{j}^{(e)}\right)+\frac{\partial N_{i}^{(e)}}{\partial y} k_{y}\left(\sum_{j=1}^{n} \frac{\partial N_{j}^{(e)}}{\partial y} \phi_{j}^{(e)}\right)\right] \mathrm{d} \Omega+ \\
&\sum_{e} \int_{\Gamma_{q}^{(e)}} \alpha N_{i}^{(e)}\left(\sum_{j=1}^{n} N_{j}^{(e)} \phi_{j}^{(e)}\right) \mathrm{d} \Gamma= \\
&\quad=\sum_{e} \iint_{\Omega^{(e)}} N_{i}^{(e)} Q \mathrm{d} \Omega-\sum_{e} \int_{\Gamma_{q}^{(e)}} N_{i}^{(e)}\left[\bar{q}_{n}-\alpha \phi_{a}\right] \mathrm{d} \Gamma-\sum_{e} \int_{\Gamma_{\phi}^{(e)}} N_{i}^{(e)} q_{n} \mathrm{d} \Gamma
\end{aligned}
\label{eq:Ec7_14} \tag{7.14}
\end{equation*}

donde el sumatorio se extiende sobre todos los elementos del dominio.

La expresión [Ec. 7.14](#Ec7_14) puede escribirse en la forma matricial estándar

<a id='Ec7_15'></a>
\begin{equation*}
\begin{aligned}
\mathbf{K a}=\mathbf{f}
\end{aligned}
\label{eq:Ec7_15} \tag{7.15}
\end{equation*}

Los términos de $\mathbf{K}$ y $\mathbf{f}$ se obtienen ensamblando las contribuciones elementales en la forma usual.

Asi, advirtiendo que las funciones de forma $N_{i}^{(e)}$ toman un valor nulo fuera de cada elemento, puede deducirse que

<a id='Ec7_16'></a>
\begin{equation*}
\begin{aligned}
K_{i j}^{(e)}=& \iint_{\Omega^{(e)}}\left(\frac{\partial N_{i}^{(e)}}{\partial x} k_{x} \frac{\partial N_{j}^{(\epsilon)}}{\partial x}+\frac{\partial N_{i}^{(e)}}{\partial y} k_{y} \frac{\partial N_{j}^{(e)}}{\partial y}\right) \mathrm{d} \Omega+\\
&+\int_{\Gamma_{q}^{(e)}} \alpha N_{i}^{(e)} N_{j}^{(e)} d \Gamma=K_{d_{i j}}+K_{c_{i j}}
\end{aligned}
\label{eq:Ec7_16} \tag{7.16}
\end{equation*}

<a id='Ec7_17'></a>
\begin{equation*}
\begin{aligned}
f_{i}^{(e)}=\iint_{\Omega^{(e)}} N_{i}^{(e)} Q \mathrm{d} \Omega-\int_{\Gamma_{q}^{(e)}} N_{i}^{(e)}\left[\bar{q}_{n}-\alpha \phi_{a}\right] \mathrm{d} \Gamma-\int_{\Gamma_{\phi}^{(e)}} N_{i}^{(e)} q_{n} d \Gamma
\end{aligned}
\label{eq:Ec7_17} \tag{7.17}
\end{equation*}

Es importante advertir que las contribuciones de cada elemento provenientes de las integrales sobre los contornos elementales se cancelan entre sí, cuando el elemento pertenece al interior del dominio. Los términos $K_{d_{k j}}^{(e)}$ y $K_{c_{i j}}^{(c)}$ en [Ec. 7.16](#Ec7_16) represeutan las contribuciones de la conductividad y la convección en la matriz de rigidez elemental. Dichos términos se obtienen por

<a id='Ec7_18'></a>
\begin{equation*}
\begin{aligned}
K_{d_{i j}}^{(e)}=\iint_{\Omega^{(\epsilon)}}\left[\frac{\partial N_{i}^{(e)}}{\partial x} k_{x} \frac{\partial N_{j}^{(\epsilon)}}{\partial x}+\frac{\partial N_{i}^{(())}}{\partial y} k_{y} \frac{\partial N_{j}^{(\epsilon)}}{\partial y}\right] \mathrm{d} \Omega
\end{aligned}
\label{eq:Ec7_18} \tag{7.18}
\end{equation*}

<a id='Ec7_19'></a>
\begin{equation*}
\begin{aligned}
K_{c_{i j}}^{(e)}=\int_{\Gamma_{q}^{(\epsilon)}} \alpha N_{i}^{(e)} N_{j}^{(e)} \mathrm{d} \Gamma
\end{aligned}
\label{eq:Ec7_19} \tag{7.19}
\end{equation*}

El cálculo por separado de $K_{d_{i j}}^{(e)}$ y $K_{c_{i j}}^{(e)}$ facilita la obtención de los coeficientes de la matriz de rigidez elemental.

De la misma manera, es útil separar los componentes del vector $f_{i}^{(e)}$ como

<a id='Ec7_20'></a>
\begin{equation*}
\begin{aligned}
f_{i}^{(e)}=f_{Q_{i}}^{(e)}+f_{c_{i}}^{(e)}+q_{n_{i}}
\end{aligned}
\label{eq:Ec7_20} \tag{7.20}
\end{equation*}

donde

<a id='Ec7_21'></a>
\begin{equation*}
\begin{aligned}
f_{Q_{i}}^{(e)}=\iint_{\Omega(e)} N_{i}^{(e)} Q \mathrm{d} \Omega
\end{aligned}
\label{eq:Ec7_21} \tag{7.21}
\end{equation*}

es la contribución del término de fuente de calor sobre el dominio

<a id='Ec7_22'></a>
\begin{equation*}
\begin{aligned}
f_{i}^{(e)}=-\int_{\Gamma_{\phi}^{(e)}} N_{i}^{(e)}\left[\bar{q}_{n}-\alpha \phi_{a}\right] \mathrm{d} \Omega
\end{aligned}
\label{eq:Ec7_22} \tag{7.22}
\end{equation*}

es la contribución del tic rnino de flujo de calor a través del contorno de Neumann donde se prescribe el flujo saliente y

<a id='Ec7_23'></a>
\begin{equation*}
\begin{aligned}
q_{n_{1}}^{(e)}=-\int_{\Gamma_{\phi}^{(e)}} N_{i}^{(e)} q_{n} \mathrm{d} \Gamma
\end{aligned}
\label{eq:Ec7_23} \tag{7.23}
\end{equation*}

es el término de flujo "reacción" sobre el contorno de $\Gamma_{\phi}$ donde se prescribe la variable de $\phi$. Dicho flujo se calcula "a posteriori", una vez obtenidos los valores nodales de $\phi .$

Se recuerda de nuevo que las integrales sobre los contornos elementales en [Ec. 7.17](#Ec7_17) sólo contribuyen al vector de fuerzas globales en el caso de que el elemento en cuestión tenga uno o más de sus lados sobre los contornos externos $\Gamma_{q} \circ \Gamma_{\phi}$

Particularizaremos ahora el proceso de obtención de las matrices y vectores anteriores para el sencillo elemento triangular de tres nodos.


### Elemento triangular de tres nodos

Supongamos una discretización del dominio de análisis en elementos triangulares de tres nodos ([Figura 7.3](#Fig7_3)).

Cada nodo del elemento tiene un grado de libertad, lo cual corresponde al valor de la incógnita $\phi$ en el nodo.

La variación de la incógnita se expresa por el polinomio lineal siguiente

<a id='Ec7_24'></a>
\begin{equation*}
\begin{aligned}
\phi(x, y)=\alpha_{0}+\alpha_{1} x+\alpha_{2} y
\end{aligned}
\label{eq:Ec7_24} \tag{7.24}
\end{equation*}

Para hallar los coeficientes $\alpha_{0}, \alpha_{1}$, y $\alpha_{2}$, basta sustituir las coordenadas nodales en [Ec. 7.24](#Ec7_24), se obtiene

<a id='Ec7_25'></a>
\begin{equation*}
\begin{aligned}

&\phi_{2}^{(e)}=\alpha_{0}+\alpha_{1} x_{2}^{(e)}+\alpha_{2} y_{2}^{(e)} \\
&\phi_{3}^{(e)}=\alpha_{0}+\alpha_{1} x_{3}^{(e)}+\alpha_{3} y_{3}^{(e)}
\end{aligned}
\label{eq:Ec7_25} \tag{7.25}
\end{equation*}

Resolviendo el sistema y sustituyendo los valores obtenidos de $\alpha_{i}$ en [Ec. 7.24](#Ec7_24) se obtiene, tras operar

<a id='Ec7_26'></a>

\begin{aligned}

\left.+\left(a_{3}^{(e)}+b_{3}^{(e)} x+c_{3}^{(e)} y\right) \phi_{3}^{(e)}\right]
\end{aligned}
\label{eq:Ec7_26} \tag{7.26}
\end{equation*}

en donde

<a id='Ec7_27'></a>
\begin{equation*}
\begin{aligned}
&a_{i}^{(e)}=x_{j}^{(e)} y_{k}^{(e)}-x_{k}^{(e)} y_{j}^{(e)} \\
&b_{i}^{(e)}=y_{j}^{(e)}-y_{k}^{(e)} \\
&c_{i}^{(e)}=x_{k}^{(e)}-x_{j}^{(c)}
\end{aligned}
\label{eq:Ec7_27} \tag{7.27}
\end{equation*}

Los subíndices $i, j, k$ varían ve 1 a 3 ([Figura 7.3](#Fig7_3)). Permutando $i, j, k$ cíclicamente se pueden hallar las expresiones de cada uno de los parámetros.

La aproximación entre el flujo total generado y el flujo total saliente por el contorno puede considerarse aceptable, dentro de los límites de exactitud de la sencilla malla empleada.
