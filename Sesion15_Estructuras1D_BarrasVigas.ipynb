## Estructuras unidimensionales - Barras

### Introducción

En este capítulo se aplica el método de los elementos finitos a estructuras que se modelan en elementos finitos unidimensionales, como son las estructuras de barras que trabajan a tracción y compresión, entre las que se encuentran las estructuras articuladas.

### Ecuación diferencial de gobierno

La solución del problema elástico en cualquier estructura hace intervenir las ecuaciones de equilibrio entre fuerzas externas e internas, las relaciones entre esfuerzos y deformaciones y las de compatibilidad de las deformaciones. Al expresar la ecuación de equilibrio de una barra sometida a tracción o compresión en función de los desplazamientos se obtiene su ecuación diferencial de gobierno.

<a id='Fig12_1a'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F01a_Barra1D.png?raw=true" width="250" />
</p>
<a id='Fig12_1b'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F01b_Barra1D_Transversal.png?raw=true" width="150" />
</p>
<div style="text-align: center"><strong>Figura 12.1 a y b.</strong>Barra prismática y sección transversal, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Sea, por ejemplo, una barra de sección transversal de área $A(x)$ y módulo de elasticidad $E$ sometida a tracción o compresión, debido a una carga longitudinal distribuida por unidad de longitud $q(x)$ [(Fig. 12.1a)](#Fig12_1a). Del equilibrio longitudinal de la rebanada [(Fig. 12.1b)](#Fig12_1b), se deduce

\begin{equation*}
\begin{aligned}
\sum X=0, \quad -N+(N+\mathrm{d} N)+q(x) \mathrm{d} x=0
\end{aligned}
\end{equation*}

de donde

<a id='Ec12_1'></a>
\begin{equation*}
\begin{aligned}
\frac{\mathrm{d} N}{\mathrm{d} x}+q(x)=0
\end{aligned}
\label{eq:Ec12_1} \tag{12.1}
\end{equation*}

Al ser un caso unidimensional, los desplazamientos longitudinales de los puntos de la barra están definidos por el campo de desplazamientos $u(x)$. Si se satisface la [hipótesis de Bernouilli-Navier](https://en.wikipedia.org/wiki/Bending) de que las secciones planas permanecen planas durante la deformación, la relación

<a id='Ec12_2'></a>
\begin{equation*}
\begin{aligned}
\varepsilon_{x}=\frac{\mathrm{d} u}{\mathrm{d} x}
\end{aligned}
\label{eq:Ec12_2} \tag{12.2}
\end{equation*}

determina el campo de deformaciones de la barra. Y si la barra es linealmente elástica, el campo de esfuerzos está definido por

<a id='Ec12_3'></a>
\begin{equation*}
\begin{aligned}
\sigma_{x}=E \varepsilon_{x}=E \frac{\mathrm{d} u}{\mathrm{d} x}
\end{aligned}
\label{eq:Ec12_3} \tag{12.3}
\end{equation*}

De la relación entre solicitaciones y esfuerzos

$$
N=\int_{A} \sigma_{x} \mathrm{d} A=\int_{A}\left(E \frac{\mathrm{d} u}{\mathrm{d} x}\right) \mathrm{d} A
$$

se deduce la ley de fuerzas normales de la barra

<a id='Ec12_4'></a>
\begin{equation*}
\begin{aligned}
N=E A \frac{\mathrm{d} u}{\mathrm{d} x}
\end{aligned}
\label{eq:Ec12_4} \tag{12.4}
\end{equation*}

Sustituyendo este valor de $N$ en la ecuación de equilibrio (3.1), se obtiene la ecuación diferencial de gobierno de una barra que trabaja a tracción o compresión

<a id='Ec12_5'></a>
\begin{equation*}
\begin{aligned}
\frac{\mathrm{d}}{\mathrm{d} x}\left(E A \frac{\mathrm{d} u}{\mathrm{d} x}\right)+q(x)=0
\end{aligned}
\label{eq:Ec12_5} \tag{12.5}
\end{equation*}

En el caso de que $E$ y $A$ sean constantes, la ecuación diferencial de gobierno se reduce a

<a id='Ec12_6'></a>
\begin{equation*}
\begin{aligned}
E A \frac{\mathrm{d}^{2} u}{\mathrm{d} x^{2}}+q(x)=0
\end{aligned}
\label{eq:Ec12_6} \tag{12.6}
\end{equation*}


### Planteamiento del MEF

#### Elemento lineal

Consideremos un elemento finito de barra de longitud $L$ con nodos en sus extremos sometido a tracción/compresión [(Fig. 12.2)](#Fig12_2). Al ser de segúndo orden la ecuación diferencial de gobierno de una barra sometida a tracción/compresión [(12.5)](#Ec12_5), su solución exige 2 parámetros.

<a id='Fig12_2'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F02_FEMBarra1D.png?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 12.2.</strong>Elemento finito de una barra, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Si se utiliza como función de desplazamientos aproximada $u(x)$ un polinomio de primer grado

$$
u(x)=\alpha_{1}+\alpha_{2} x
$$

su representación es una recta [(Fig. 12.3a)](#Fig12_3a) y el elemento considerado es un elemento lineal. Siendo los desplazamientos $u_{1}$ y $u_{2}$ de sus nodos $1$ y $2$ los únicos parámetros nodales, al particularizar los valores de la función de desplazamientos $u(x)$ en los nodos del elemento, de acuerdo con [(12.2)](#Ec12_2), se obtiene

<a id='Fig12_3b'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F03a_FuncionInterpolacion.png?raw=true" width="150" />
</p>
<a id='Fig12_3c'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F03b_FuncionInterpolacion.png?raw=true" width="150" />
</p>
<div style="text-align: center"><strong>Figura 12.3.</strong>Funciones de interpolación, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

<a id='Ec12_7'></a>
\begin{equation*}
\begin{aligned}
u(x)=N_{1} u_{1}+N_{2} u_{2}=\left|\begin{array}{ll}
N_{1} & N_{2}
\end{array}\right|\left|\begin{array}{l}
u_{1} \\
u_{2}
\end{array}\right|
\end{aligned}
\label{eq:Ec12_7} \tag{12.7}
\end{equation*}

o bien 

<a id='Ec12_8'></a>
\begin{equation*}
\begin{aligned}
\left\{\mathbf{u}_{e}\right\}=\left[\mathbf{N}_{e}\right]\left\{\boldsymbol{\delta}_{e}\right\}
\end{aligned}
\label{eq:Ec12_8} \tag{12.8}
\end{equation*}

donde $\left[\mathbf{N}_{e}\right]$ es la matriz de interpolación o matriz de forma del elemento 

<a id='Ec12_9'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{N}_{e}\right]=\left[\mathbf{N}_{1} \mathbf{N}_{2}\right]=\left|N_{1} \quad N_{2}\right|
\end{aligned}
\label{eq:Ec12_9} \tag{12.9}
\end{equation*}

Las funciones $N_{1} \mathrm{y} N_{2}$, halladas en [(12.4)](#Ec12_4), 

<a id='Ec12_10'></a>
\begin{equation*}
\begin{aligned}
N_{1}=\frac{x_{2}-x}{L}, N_{2}=\frac{-x_{1}+x}{L}
\end{aligned}
\label{eq:Ec12_10} \tag{12.10}
\end{equation*}

son las funciones de interpolación o de forma del elemento lineal y son polinomios que tienen valor unidad en el nodo correspondiente y valor cero en el otro nodo [(Fig. 12.3c)](#Fig12_3c), por lo que son polinomios lagrangianos y el elemento será lagrangiano. Cada función de interpolación $N_{i}$ determina los desplazamientos de los puntos del elemento cuando se le da un valor unidad al desplazamiento del nodo $i$ manteniendo nulo el desplazamiento del otro nodo. Por esta razón, también se les llama funciones piramidales.

Si se adopta como origen de coordenadas el $nodo 1$ , entonces $x_{1}=0, x_{2}=L, y$ las expresiones de las funciones de interpolación se reducen a

<a id='Ec12_11'></a>
\begin{equation*}
\begin{aligned}
N_{1}=1-\frac{x}{L}, \quad N_{2}=\frac{x}{L}
\end{aligned}
\label{eq:Ec12_11} \tag{12.11}
\end{equation*}

y la matriz de interpolación correspondiente es

<a id='Ec12_12'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{N}_{e}\right]=\left[1-\frac{x}{L} \quad \frac{x}{L}\right]
\end{aligned}
\label{eq:Ec12_12} \tag{12.12}
\end{equation*}

Teniendo en cuenta [(12.4)](#Ec12_4), la ley de fuerzas normales de un elemento lineal es

$$
N=E A\left[\frac{\mathrm{d} \mathbf{N}_{e}}{\mathrm{d} x}\right]\{\delta\}=E A\left[\begin{array}{ll}
-\frac{1}{L} & \frac{1}{L}
\end{array}\right]\left|\begin{array}{l}
u_{1} \\
u_{2}
\end{array}\right|
$$

es decir

<a id='Ec12_13'></a>
\begin{equation*}
\begin{aligned}
N=E A \frac{u_{2}-u_{1}}{L}
\end{aligned}
\label{eq:Ec12_13} \tag{12.13}
\end{equation*}

Si se utiliza el sistema de coordenadas de longitud $\xi_{1} \mathrm{y} \xi_{2}$ [(Fig. 3.4a)](#Fig12_4a), 

<a id='Fig12_4'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F04_Coordenadas.png?raw=true" width="350" />
</p>
<div style="text-align: center"><strong>Figura 12.4.</strong>Coordenadas de longitud y natural, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

$$
\xi_{1}=\frac{L_{1}}{L}=1-\frac{x}{L}, \quad \xi_{2}=\frac{L_{2}}{L}=\frac{x}{L}
$$

y, por tanto, teniendo en cuenta [(12.11)](#Ec12_11), resulta

<a id='Ec12_14a'></a>
\begin{equation*}
\begin{aligned}
N_{1}=\xi_{1}, \quad N_{2}=\xi_{2},
\end{aligned}
\label{eq:Ec12_14a} \tag{12.14a}
\end{equation*}

lo que permite establecer que las funciones de forma $N_{1} y N_{2}$ del elemento unidimensional de dos nodos coinciden con las coordenadas longitudinales $\xi_{1}$ y $\xi_{2}$.

Finalmente, si se utiliza la coordenada natural $\xi$ [(Fig. 12.4b)](#Fig12_4),

$$
\xi=\frac{x-x_{C}}{L / 2}=2 \frac{x}{L}-1
$$

las funciones de forma son

<a id='Ec12_14b'></a>
\begin{equation*}
\begin{aligned}
N_{1}=\frac{1}{2}(1-\xi), \quad N_{2}=\frac{1}{2}(1+\xi)
\end{aligned}
\label{eq:Ec12_14b} \tag{12.14b}
\end{equation*}



#### Discretización en un elemento finito

<a id='Fig12_5'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F05_BarraTraccion.png?raw=true" width="200" />
</p>
<div style="text-align: center"><strong>Figura 12.5.</strong>Barra a tracción, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Consideremos una barra sometida a tracción/compresión [(Fig. 12.5a)](#Fig12_5) que se ha discretizado utilizando un elemento lineal cuyos desplazamientos nodales son $u_{1}$, $u_{2}$ [(Fig. 12.5b)](#Fig12_5). Según (2.32) la matriz de rigidez del elemento finito es

<a id='Ec12_15'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{k}_{e}\right]=\int_{V_{e}}\left[\mathbf{B}_{e}\right]^{T}[\mathbf{D}]\left[\mathbf{B}_{e}\right] \mathrm{d} V_{e}
\end{aligned}
\label{eq:Ec12_15} \tag{12.15}
\end{equation*}

siendo $[\mathbf{D}]$ la matriz constitutiva $\mathrm{y}\left[\mathbf{B}_{e}\right]$ la matriz de deformación del elemento, definida por la relación

<a id='Ec12_16'></a>
\begin{equation*}
\begin{aligned}
\{\varepsilon\}=\left[\mathbf{B}_{e}\right]\left\{\delta_{e}\right\}
\end{aligned}
\label{eq:Ec12_16} \tag{12.16}
\end{equation*}

Un punto cualquiera situado en una sección transversal del elemento está sometido a un estado uniaxial de esfuerzos $\sigma_{x}$ al que corresponde una matriz constitutiva.

<a id='Ec12_17'></a>
\begin{equation*}
\begin{aligned}
[\mathbf{D}]=E
\end{aligned}
\label{eq:Ec12_17} \tag{12.17}
\end{equation*}

siendo $E$ el módulo de elasticidad del material. Además, el vector de deformaciones se reduce a

<a id='Ec12_18'></a>
\begin{equation*}
\begin{aligned}
\{\varepsilon\}=\varepsilon_{x}
\end{aligned}
\label{eq:Ec12_18} \tag{12.18}
\end{equation*}

Sustituyendo en [(12.15)](#Ec12_15), $[\mathbf{D}]=E$ y $d V_{e}=A(x) d x$, por ser constante el estado de deformaciones en todos los puntos de una sección transversal, se determina la matriz de rigidez de un elemento sometido a tracción/compresión

<a id='Ec12_19'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{k}_{e}\right]=\int_{L}\left[\mathbf{B}_{e}\right]^{T} E A(x)\left[\mathbf{B}_{e}\right] \mathrm{d} x.
\end{aligned}
\label{eq:Ec12_19} \tag{12.19}
\end{equation*}

Según [(12.2)](#Ec12_2), la deformación lineal en un punto es

$$
\varepsilon_{x}=\frac{\mathrm{d} u}{\mathrm{d} x}
$$

y, teniendo en cuenta [(12.8)](#Ec12_8)

$$
u(x)=\left[\mathbf{N}_{e}\right]\left\{\delta_{e}\right\}
$$

por lo que

<a id='Ec12_20'></a>
\begin{equation*}
\begin{aligned}
\{\varepsilon\}=\varepsilon_{x}=\frac{d u}{d x}=\left[\frac{d \mathbf{N}_{e}}{d x}\right]\left\{\boldsymbol{\delta}_{e}\right\}=\left[\mathbf{B}_{e}\right]\left\{\boldsymbol{\delta}_{e}\right\}
\end{aligned}
\label{eq:Ec12_20} \tag{12.20}
\end{equation*}

y, por tanto, la matriz de deformación de un elemento finito sometido a tracción/compresión es

<a id='Ec12_21'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{B}_{e}\right]=\left[\frac{\mathrm{d} \mathbf{N}_{e}}{\mathrm{d} x}\right]
\end{aligned}
\label{eq:Ec12_21} \tag{12.21}
\end{equation*}

con

<a id='Ec12_22'></a>
\begin{equation*}
\begin{aligned}
\left[\partial\right]=\frac{\mathrm{d}}{\mathrm{d} x}
\end{aligned}
\label{eq:Ec12_22} \tag{12.22}
\end{equation*}

Por consiguiente

$$
\left[\mathbf{B}_{e}\right]=\frac{\mathrm{d}}{\mathrm{d} x}\left[1-\frac{x}{L} \quad \frac{x}{L}\right]
$$

o sea

<a id='Ec12_23'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{B}_{e}\right]=\left[\begin{array}{ll}
-\frac{1}{L} & \frac{1}{L}
\end{array}\right]
\end{aligned}
\label{eq:Ec12_23} \tag{12.23}
\end{equation*}

Sustituyendo esta expresión en [(12.19)](#Ec12_19) se obtiene la expresión de la matriz de rigidez del elemento que, por ser un sólo elemento, es también la matriz completa de rigidez de la barra

<a id='Ec12_24'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{K}_{0}\right]=\left[\mathbf{k}_{e}\right]=\int_{L}\left|\begin{array}{c}
-\frac{1}{L} \\
\frac{1}{L}
\end{array}\right| E\left[-\frac{1}{L} \quad \frac{1}{L}\right] A(x) \mathrm{d} x
\end{aligned}
\label{eq:Ec12_24} \tag{12.24}
\end{equation*}

es decir

<a id='Ec12_25'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{K}_{0}\right]=\left[\mathbf{k}_{e}\right]=\frac{E \int_{L} A(x) \mathrm{d} x}{L^{2}}\left|\begin{array}{rr}
1 & -1 \\
-1 & 1
\end{array}\right|
\end{aligned}
\label{eq:Ec12_25} \tag{12.25}
\end{equation*}

Cuando $E A(x)$ es constante

<a id='Ec12_26'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{K}_{0}\right]=\left[\mathbf{k}_{e}\right]=\frac{E A}{L}\left|\begin{array}{rr}
1 & -1 \\
-1 & 1
\end{array}\right|
\end{aligned}
\label{eq:Ec12_26} \tag{12.26}
\end{equation*}

que coincide con la expresión del método de los desplazamientos. La razón de esta coincidencia es el haber elegido como aproximación para los desplazamientos funciones de forma piramidales que establecen la linealidad de los desplazamientos, como sucede cuando las fuerzas normales actúan únicamente en las secciones extremas de una barra de sección constante.

#### Discretización en dos elementos finitos 

<a id='Fig12_6'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F06_BarraSeccionVariable.png?raw=true" width="200" />
</p>
<div style="text-align: center"><strong>Figura 12.6.</strong>Barra a tracción, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Supongamos ahora que una barra de sección variable, continua [(Fig. 12.6a)](#Fig12_6) o discontinua [(Fig. 12.6b)](#Fig12_6), sometida a tracción/compresión, se ha discretizado en dos elementos finitos lineales de longitudes $L_{1}$ y $L_{2}$, secciones transversales de áreas $A_{1}$ y $A_{2}$ y desplazamientos nodales $u_{1}, u_{2}$ y $u_{3}$ [(Fig. 12.6c)](#Fig12_6).

La matriz completa de rigidez $\left[\mathbf{K}_{0}\right]$ de la barra discretizada es una matriz de orden $n=3$ igual al número de grados de libertad. Según (2.48), la expresión de las submatrices de $\left[\mathbf{K}_{0}\right]$ es

$$
\left[\mathbf{K}_{i j}\right]=\int_{V_{e}}\left[\mathbf{B}_{i}\right]^{T}[\mathbf{D}]\left[\mathbf{B}_{j}\right]  \mathrm{d} V
$$

que en este caso unidimensional se reduce a

$$
\left[\mathbf{K}_{i j}\right]=\int_{L}\left[\mathbf{B}_{i}\right]^{T} E A(x)\left[\mathbf{B}_{j}\right]  \mathrm{d} x
$$

y al ser cada elemento finito de sección constante

<a id='Ec12_27'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{K}_{i j}\right]=\int_{L}\left[\mathbf{B}_{i}\right] E A\left[\mathbf{B}_{j}\right] \mathrm{d} x .
\end{aligned}
\label{eq:Ec12_27} \tag{12.27}
\end{equation*}

<a id='Fig12_6d'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F06d_FuncionesForma.png?raw=true" width="500" />
</p>
<div style="text-align: center"><strong>Figura 12.6d.</strong>Funciones de forma para dos elementos lineales, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Al adoptar las funciones piramidales (Fig. $3.6 d$ ) como funciones de interpolación, las submatrices de deformación correspondientes a cada elemento son

- Elemento 1

$$
\left[\mathbf{B}_{1}^{(1)}\right]=\frac{\mathrm{d} N_{1}}{\mathrm{d} x}=-\frac{1}{L_{1}},\left[\mathbf{B}_{2}^{(1)}\right]=\frac{\mathrm{d} N_{2}}{\mathrm{d} x}=\frac{1}{L_{1}},\left[\mathbf{B}_{3}^{(1)}\right]=\frac{\mathrm{d} N_{3}}{\mathrm{d} x}=0
$$

- Elemento 2

$$
\left[\mathbf{B}_{1}^{(2)}\right]=\frac{\mathrm{d} N_{1}}{\mathrm{d} x}=0, \quad\left[\mathbf{B}_{2}^{(2)}\right]=\frac{\mathrm{d} N_{2}}{\mathrm{d} x}=-\frac{1}{L_{2}}, \quad\left[\mathbf{B}_{3}^{(2)}\right]=\frac{\mathrm{d} N_{3}}{\mathrm{d} x}=\frac{1}{L_{2}} .
$$

Sustituyendo estas submatrices de deformación en (3.27), se obtienen las submatrices de rigidez de la barra discretizada en dos elementos finitos

$$
\begin{aligned}
&{\left[\mathbf{K}_{11}\right]=E A_{1} \int_{L_{1}}\left(-\frac{1}{L_{1}}\right)\left(-\frac{1}{L_{1}}\right) \mathrm{d} x+0=\left[\frac{E A_{1}}{L_{1}}\right]} \\
&{\left[\mathbf{K}_{12}\right]=\left[\mathbf{K}_{21}\right]=E A_{1} \int_{L_{1}}\left(-\frac{1}{L_{1}}\right)\left(\frac{1}{L_{1}}\right) \mathrm{d} x+0=\left[-\frac{E A_{1}}{L_{1}}\right]} \\
&{\left[\mathbf{K}_{13}\right]=\left[\mathbf{K}_{31}\right]=0+0=0} \\
&{\left[\mathbf{K}_{22}\right]=E A_{1} \int_{L_{1}}\left(\frac{1}{L_{1}}\right)\left(\frac{1}{L_{1}}\right) \mathrm{d} x+E A_{2} \int_{L_{2}}\left(-\frac{1}{L_{2}}\right)\left(-\frac{1}{L_{2}}\right) \mathrm{d} x=\left[\frac{E A_{1}}{L_{1}}+\frac{E A_{2}}{L_{2}}\right]} \\
&{\left[\mathbf{K}_{23}\right]=\left[\mathbf{K}_{32}\right]=0+E A_{2} \int_{L_{2}}\left(-\frac{1}{L_{2}}\right)\left(\frac{1}{L_{2}}\right) \mathrm{d} x=\left[-\frac{E A_{2}}{L_{2}}\right]} \\
&{\left[\mathbf{K}_{33}\right]=0+E A_{2} \int_{L_{2}}\left(\frac{1}{L_{2}}\right)\left(\frac{1}{L_{2}}\right) \mathrm{d} x=\left[\frac{E A_{2}}{L_{2}}\right]}
\end{aligned}
$$

Por consiguiente, la matriz completa de rigidez de la barra es

<a id='Ec12_28'></a>
\begin{equation*}
\begin{aligned}
\left[\mathbf{K}_{0}\right]=E\left|\begin{array}{ccc}
\frac{A_{1}}{L_{1}} & -\frac{A_{1}}{L_{1}} & 0 \\
-\frac{A_{1}}{L_{1}} & \frac{A_{1}}{L_{1}}+\frac{A_{2}}{L_{2}} & -\frac{A_{2}}{L_{2}} \\
0 & -\frac{A_{2}}{L_{2}} & \frac{A_{2}}{L_{2}}
\end{array}\right|
\end{aligned}
\label{eq:Ec12_28} \tag{12.28}
\end{equation*}

La matriz completa de rigidez $\left[\mathbf{K}_{0}\right]$ se podía haber obtenido directamente ensamblando las submatrices de los dos elementos

$$
\left[\mathbf{K}_{0}\right]=\left|\begin{array}{ccc}
\mathbf{k}_{11}^{(1)} & \mathbf{k}_{12}^{(1)} & \mathbf{0} \\
\mathbf{k}_{21}^{(1)} & \mathbf{k}_{22}^{(1)}+\mathbf{k}_{22}^{(2)} & \mathbf{k}_{23}^{(2)} \\
\mathbf{0} & \mathbf{k}_{32}^{(2)} & \mathbf{k}_{33}^{(2)}
\end{array}\right|
$$

De forma análoga se determina la matriz completa de rigidez de barras discretizadas en mas de dos elementos finitos.



### Vector de fuerzas nodales equivalentes

#### Carga distribuida

<a id='Fig12_7'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F07_VigaCargaDistribuida.png?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 12.7.</strong>Viga con carga distribuida, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Consideremos que sobre el elemento lineal actúa una fuerza o carga longitudinal cualquiera $q(x)$ distribuida por unidad de longitud [(Fig. 12.7a)](#Fig12_7). El vector de fuerzas nodales equivalente a la carga distribuida es

$$
\left\{\mathbf{F}_{q}\right\}=\int_{V_{e}}\left[\mathbf{N}_{e}\right]^{T}\left\{\mathbf{q}_{e}\right\} \mathrm{d} V_{e} .
$$

En el caso de un elemento finito unidimensional, $\left\{q_{e}\right\} \mathrm{d} V=q(x) \mathrm{d} x$ y, por tanto

$$
\left\{\mathbf{F}_{q}\right\}=\int_{L}\left[\mathbf{N}_{e}\right]^{T} q(x) \mathrm{d} x
$$

y, teniendo en cuenta [(12.12)](#Ec12_12)

$$
\left[\mathbf{N}_{e}\right]^{T}=\left|\begin{array}{c}
N_{1} \\
N_{2}
\end{array}\right|=\left|\begin{array}{c}
1-\frac{x}{L} \\
\frac{x}{L}
\end{array}\right|,
$$

por lo que el vector de fuerzas nodales equivalente [(Fig. 12.7b)](#Fig12_7) es

<a id='Ec12_29'></a>
\begin{equation*}
\begin{aligned}
\left\{\mathbf{F}_{q}\right\}=\left|\begin{array}{c}
F_{1} \\
F_{2}
\end{array}\right|=\left|\begin{array}{l}
\int_{L} N_{1} q(x) \mathrm{d} x \\
\int_{L} N_{2} q(x) \mathrm{d} x
\end{array}\right|
\end{aligned}
\label{eq:Ec12_29} \tag{12.29}
\end{equation*}

o sea

<a id='Ec12_30'></a>
\begin{equation*}
\begin{aligned}
\left\{\mathbf{F}_{q}\right\}=\left|\begin{array}{c}
F_{1} \\
F_{2}
\end{array}\right|=\left|\begin{array}{c}
\int_{L} q(x)\left(1-\frac{x}{L}\right) \mathrm{d} x \\
\int_{L} q(x) \frac{x}{L} \mathrm{d} x
\end{array}\right|
\end{aligned}
\label{eq:Ec12_30} \tag{12.30}
\end{equation*}

Cuando la carga está uniformemente distribuida, $q(x)=q$, el vector de fuerzas nodales equivalente se reduce a

<a id='Ec12_31'></a>
\begin{equation*}
\begin{aligned}
\left\{\mathbf{F}_{q}\right\}=\left|\begin{array}{c}
F_{1} \\
F_{2}
\end{array}\right|=\left|\begin{array}{c}
\frac{q L}{2} \\
\frac{q L}{2}
\end{array}\right|\end{aligned}
\label{eq:Ec12_31} \tag{12.31}
\end{equation*}


#### Carga concentrada

<a id='Fig12_8'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F08_VigaCargaConcentrada.png?raw=true" width="250" />
</p>
<div style="text-align: center"><strong>Figura 12.8.</strong>Viga con carga concentrada, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Supongamos ahora que el elemento finito está sometido a una carga concentrada $Q$ [(Fig. 12.8a)](#Fig12_8). Considerando esta carga como la resultante de una carga distribuida $q$ sobre una longitud de elemento $d x$, tal que $q \mathrm{d} x=Q$ [(Fig. 12.8b)](#Fig12_8), al hacer las integraciones [(12.30)](#Ec12_30) a lo largo de todo el elemento, el producto $q \mathrm{d} x$ es igual a $Q$ únicamente para el punto $x=a$, siendo nulo en los restantes puntos del elemento, lo que significa que

$$
\int_{0}^{L} N q \mathrm{d} x=(N)_{x=\alpha} q \mathrm{d} x=Q(N)_{x=a}
$$

y, por tanto, teniendo en cuenta (3.29), se obtienen las fuerzas nodales equivalentes $F_{1}$ y $F_{2}$ (Fig. 3.10c)

$$
\begin{aligned}
&F_{1}=\int_{0}^{L} N_{1} q \mathrm{d} x=Q\left(N_{1}\right)_{x=a} \\
&F_{2}=\int_{0}^{L} N_{2} q \mathrm{d} x=Q\left(N_{2}\right)_{x=a}
\end{aligned}
$$

y, puesto que

$$
\left(N_{1}\right)_{x=a}=\left(1-\frac{a}{L}\right)=\frac{b}{L}, \quad\left(N_{2}\right)_{x=a}=\frac{a}{L}
$$

resulta

<a id='Ec12_32'></a>
\begin{equation*}
\begin{aligned}
&F_{1}=\frac{Q b}{L} \\
&F_{2}=\frac{Q a}{L}
\end{aligned}
\label{eq:Ec12_32} \tag{12.32}
\end{equation*}

Cuando la carga $Q$ está aplicada en el punto medio del elemento, $a=L / 2$, el vector de fuerzas nodales equivalente es

<a id='Ec12_33'></a>
\begin{equation*}
\begin{aligned}
\left\{\mathbf{F}_{Q}\right\}=\left|\begin{array}{c}
\frac{Q}{2} \\
\frac{Q}{2}
\end{array}\right|
\end{aligned}
\label{eq:Ec12_33} \tag{12.33}
\end{equation*}



#### Reacciones sobre los apoyos

En los resultados obtenidos se observa que en un elemento finito sometido a tracción/compresión las fuerzas nodales equivalentes son iguales y de sentido contrario a las reacciones de las articulaciones. Es decir, iguales a las acciones sobre las articulaciones que son las acciones que el elemento, supuestamente articulado en sus extremos, ejerce sobre las articulaciones.

<a id='Fig12_9'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C12_F08_VigaCargaConcentrada.png?raw=true" width="200" />
</p>
<div style="text-align: center"><strong>Figura 12.9.</strong>Reacción sobre los apoyos, por Vasquez y López, <a href="https://www.buscalibre.com.co/libro-el-metodo-de-los-elementos-finitos-aplicado-al-analisis-estructural/9788488012067/p/2885127"> El método de los Elementos Finitos aplicado al análisis estructural.</a></div>

Para demostrar la anterior equivalencia, consideremos un estado real de cargas en el que el elemento finito $1-2$ está sometido a cargas cualesquiera $q(x)$ que producen tracción/compresión en el elemento [(Fig. 12.8a)](#Fig12_8). Este estado de cargas origina un estado $A$ de desplazamientos compatibles con los nodos y, por ello, se incluyen los posibles desplazamientos nodales $u_{1}, u_{2}$ [(Fig. 12.8b)](#Fig12_8). Seguidamente, se considera un estado ficticio $B$ de cargas en el que el elemento sometido a las cargas reales aplicadas tiene articulados sus extremos [(Fig. 12.8c)](#Fig12_8) y, por tanto, estará sometido también a las reacciones de las articulaciones $R_{1}, R_{2}$. Aplicando el teorema de los trabajos virtuales al estado $B$ de cargas suponiéndoles el estado $A$ de desplazamientos virtuales, resulta

$$
\int_{0}^{L}(q(x) \mathrm{d} x) u(x)+R_{1} u_{1}+R_{2} u_{2}=0
$$

y, por tanto

$$
\int_{0}^{L} q(x) u(x) \mathrm{d} x=\left(-R_{1}\right) u_{1}+\left(-R_{2}\right) u_{2} .
$$

Según esta ecuación, el trabajo que realizan las fuerzas aplicadas $q(x)$ es igual al que realizan las acciones sobre las articulaciones suponiendo a las mismas desplazamientos nodales. Debido a ello, las acciones sobre las articulaciones son las fuerzas nodales equivalentes. Esta propiedad se utiliza al considerar el estado de carga 2 del método de los desplazamientos aplicado a estructuras articuladas.
