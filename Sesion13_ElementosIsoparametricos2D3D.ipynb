## Elementos Isoparamétricos 2D y 3D

### Introducción

Se presenta el concepto de "interpolación isoparamétrica", para formular elementos isoparamétricos en dos y tres dimensiones.


### Elementos cuadriláteros isoparamétricos

Los elementos isoparamétricos utilizan las mismas funciones de forma para interpolar la geometria y las incógnitas. Por consiguiente, expresaremos la geometría de un elemento isoparamétrico bidimensional a partir de las coordenadas $x$ e $y$ de sus nodos como

<a id='Ec10_1'></a>
\begin{equation*}
\begin{aligned}
x=\sum_{i=1}^{n} N_{i}(\xi, \eta) x_{i}, \quad y=\sum_{i=1}^{n} N_{i}(\xi, \eta) y_{i}
\end{aligned}
\label{eq:Ec10_1} \tag{10.1}
\end{equation*}

donde $N_{i}(\xi, \eta)$ son precisamente las funciones de forma del elemento. Las [Ecs. 10.1](#Ec10_1) relacionan las coordenadas cartesianas de un punto y las naturales $\xi$ y $\eta$. Dicha relación debe ser biunívoca, para lo cual debe cumplirse que el determinante de la matriz jacobiana de la transformación de coordenadas $x y \rightarrow \xi \eta$ (dicha matriz se define más adelante) sea de signo constante en todo el elemento.

Puede demostrarse que si se utilizan funciones de forma lineales dicha condición exige que ningún ángulo interior entre dos lados del elemento sea mayor que $180^{\circ}$. Si las funciones de forma son cuadráticas es necesario además que los nodos sobre los lados se encuentren en el tercio central de la distancia entre los nodos esquina adyacentes. Para funciones de forma de órdenes superiores no existen reglas prácticas y es necesario comprobar el signo del determinante del jacobiano. No obstante, las funciones de grado superior a dos son poco utilizables en la práctica. En la [Figura 10.1](#Fig10_1) se muestran algunos ejemplos de elementos isoparamétricos en dos dimensiones.

Para mayor claridad y sencillez, prescindiremos en lo que sigue del índice $(e)$ en las funciones de forma del elemento, así como en la matriz de flujos elementules.

La [Ec. 10.1](#Ec10_1) permite obtener la relación entre las derivadas de las funciones de forma con respecto a los sistemas de coordenadas cartesiano y natural. En general, $N_{i}$ vendrá expresada en las coordenadas naturales $\xi$ y $\eta$, por lo que la regla de la derivación en cadena permite escribir

<a id='Ec10_2'></a>
\begin{equation*}
\begin{aligned}
\frac{\partial N_{i}}{\partial \xi} &=\frac{\partial N_{i}}{\partial x} \frac{\partial x}{\partial \xi}+\frac{\partial N_{i}}{\partial y} \frac{\partial y^{\prime}}{\partial \xi} \\
\frac{\partial N_{i}}{\partial \eta} &=\frac{\partial N_{i}}{\partial x} \frac{\partial x}{\partial \eta}+\frac{\partial N_{i}}{\partial y} \frac{\partial y}{\partial \eta}
\end{aligned}
\label{eq:Ec10_2} \tag{10.2}
\end{equation*}

o en forma matricial

<a id='Ec10_3'></a>
\begin{equation*}
\begin{aligned}
\left\{\begin{array}{l}
\frac{\partial N_{i}}{\partial \xi} \\
\frac{\partial N_{i}}{\partial \eta}
\end{array}\right\}=\underbrace{\left[\begin{array}{ll}
\frac{\partial x}{\partial \xi} & \frac{\partial y}{\partial \xi} \\
\frac{\partial x}{\partial \eta} & \frac{\partial y}{\partial \eta}
\end{array}\right]}_{\mathbf{J}^{(e)}}\left\{\begin{array}{l}
\frac{\partial N_{i}}{\partial x} \\
\frac{\partial N_{i}}{\partial y}
\end{array}\right\}=\mathbf{J}^{(e)}\left\{\begin{array}{l}
\frac{\partial N_{i}}{\partial x} \\
\frac{\partial N_{i}}{\partial y}
\end{array}\right\}
\end{aligned}
\label{eq:Ec10_3} \tag{10.3}
\end{equation*}

donde $\mathbf{J}^{(e)}$ es la matriz jacobiana, o simplemente el jacobiano, de la transformación de coordenadas naturales a cartesianas. De [Ec. 10.3](#Ec10_3) se deduce

<a id='Ec10_4'></a>
\begin{equation*}
\begin{aligned}
\left\{\begin{array}{l}
\frac{\partial N_{i}}{\partial x} \\
\frac{\partial N_{i}}{\partial y}
\end{array}\right\}=\left[\mathbf{J}^{(e)}\right]^{-1}\left\{\begin{array}{l}
\frac{\partial N_{i}}{\partial \xi} \\
\frac{\partial N_{i}}{\partial \eta}
\end{array}\right\}=\frac{1}{|\mathbf{J}^{(e)}|}\left[\begin{array}{cc}
\frac{\partial y}{\partial \eta} & -\frac{\partial y}{\partial \xi} \\
-\frac{\partial x}{\partial \eta} & \frac{\partial x}{\partial \xi}
\end{array}\right]\left\{\begin{array}{l}
\frac{\partial N_{i}}{\partial \xi} \\
\frac{\partial N_{i}}{\partial \eta}
\end{array}\right\}
\end{aligned}
\label{eq:Ec10_4} \tag{10.4}
\end{equation*}

donde $\left|\mathbf{J}^{(e)}\right|$ es el determinante del jacobiano.

El determinante del jacobiano permite también expresar el diferencial de área en coordenadas naturales como

<a id='Ec10_5'></a>
\begin{equation*}
\begin{aligned}
\mathrm{d} x \mathrm{d} y=\left|\mathrm{J}^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta
\end{aligned}
\label{eq:Ec10_5} \tag{10.5}
\end{equation*}

Para calcular los términos del jacobiano se utiliza la transformación isoparamétrica [Ec. 10.1](#Ec10_1). Así

\begin{equation*}
\begin{aligned}
\frac{\partial x}{\partial \xi}=\sum_{i=1}^{n} \frac{\partial N_{i}}{\partial \xi} x_{i} \quad ; \quad \frac{\partial x}{\partial \eta}=\sum_{1=1}^{n} \frac{\partial N_{i}}{\partial \eta} x_{i} ; \text { etc. }
\end{aligned}
\end{equation*}

por lo que

<a id='Ec10_6'></a>
\begin{equation*}
\begin{aligned}
\mathbf{J}^{(e)}=\sum_{i=1}^{n}\left[\begin{array}{ll}
\frac{\partial N_{i}}{\partial \xi_{i}} & \frac{\partial N_{i}}{\partial \xi_{i}} y_{i} \\
\frac{\partial N_{i}}{\partial \eta} x_{i} & \frac{\partial N_{i}}{\partial \eta} y_{i}
\end{array}\right]
\end{aligned}
\label{eq:Ec10_6} \tag{10.6}
\end{equation*}

Si el elemento es rectangular es fácil obtener

<a id='Ec10_7'></a>
\begin{equation*}
\begin{aligned}
\mathbf{J}^{(e)}=\left[\begin{array}{ll}
a & 0 \\
0 & b
\end{array}\right] \quad \text { y } \quad\left|\mathbf{J}^{(e)}\right|=a b
\end{aligned}
\label{eq:Ec10_7} \tag{10.7}
\end{equation*}

Volviendo al problema de Poisson, la matriz de gradiente nodal de un elemento isoparamétrico bidimensional se expresa en función de las coordenadas naturales, haciendo uso de [Ec. 10.4](#Ec10_4), por

<a id='Ec10_8a'></a>
\begin{equation*}
\begin{aligned}
\mathbf{B}_{i}(\xi, \eta)=\left\{\begin{array}{l}
\frac{\partial N_{i}}{\partial x} \\
\frac{\partial N_{i}}{\partial y}
\end{array}\right\}=\frac{1}{|\mathbf{J}^{(e)}|}\left\{\begin{array}{c}
\bar{b}_{i} \\
\bar{c}_{i}
\end{array}\right\}
\end{aligned}
\label{eq:Ec10_8a} \tag{10.8a}
\end{equation*}

donde

<a id='Ec10_8b'></a>
\begin{equation*}
\begin{aligned}
\bar{b}_{i}=\frac{\partial y}{\partial \eta} \frac{\partial N_{i}}{\partial \xi}-\frac{\partial y}{\partial \xi} \frac{\partial N_{i}}{\partial \eta} \quad ; \quad \bar{c}_{i}=\frac{\partial x}{\partial \xi} \frac{\partial N_{i}}{\partial \eta}-\frac{\partial x}{\partial \eta} \frac{\partial N_{i}}{\partial \xi}
\end{aligned}
\label{eq:Ec10_8b} \tag{10.8b}
\end{equation*}

Haciendo uso de las expresiones anteriores, la matriz de rigidez del elemento puede escribirse como una integral sobre el dominio normalizado de las coordenadas naturales por 

<a id='Ec10_9a'></a>
\begin{equation*}
\begin{aligned}
K_{i j}^{(e)}=& \iint_{A^{(e)}} \mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j} \mathrm{d} x \mathrm{d} y=\int_{-1}^{+1} \int_{-1}^{+1} \mathbf{B}_{i}^{T}(\xi, \eta) \mathbf{D} \mathbf{B}_{j}(\xi, \eta)\left|\mathbf{J}^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta=\\
=& \int_{-1}^{+1} \int_{-1}^{+1}\left(k_{x} \bar{b}_{i} \bar{b}_{j}+k_{y} \bar{c}_{i} \bar{c}_{j}\right) \frac{1}{|\mathbf{J}(e)|} \mathrm{d} \xi \mathrm{d} \eta=\\
=& \int_{-1}^{+1} \int_{-1}^{+1} G_{i j} \mathrm{d} \xi \mathrm{d} \eta
\end{aligned}
\label{eq:Ec10_9a} \tag{10.9a}
\end{equation*}

con

<a id='Ec10_9b'></a>
\begin{equation*}
\begin{aligned}
G_{i j}=\left(k_{x} \bar{b}_{i} \bar{b}_{j}+k_{y} \bar{c}_{i} \bar{c}_{j}\right) \frac{1}{|J^{(e)}|}
\end{aligned}
\label{eq:Ec10_9b} \tag{10.9b}
\end{equation*}

Se deduce de id expresión anterior que los términos del integrando son funciones racionales en $\xi$ y $\eta$ a menos que el determinante del jacobiano sea constante. Esto sólo ocurre en elementos rectangulares o en elementos triangulares de lados rectos, en cuyo caso las integrales se simplifican notablemente. Sin embargo, en elementos de lados curvos, la integración analítica de los términos de $K_{i j}^{(e)}$ es compleja y es necesario hacer uso de integración numérica.


### Elementos triangulares isoparamétricos

En elementos triangulares la interpolación isoparamétrica se define de forma similar a la [Ec. 10.1](#Ec10_1) por

<a id='Ec10_10'></a>
\begin{equation*}
\begin{aligned}
x=\sum_{i=1}^{n} N_{i}\left(L_{1}, L_{2}, L_{3}\right) x_{i} \quad ; \quad y=\sum_{i=1}^{n} N_{i}\left(L_{1}, L_{2}, L_{3}\right) y_{i}
\end{aligned}
\label{eq:Ec10_10} \tag{10.10}
\end{equation*}

Si el elemento triangular es de lados rectos, el cálculo de las derivadas cartesianas de las funciones de forma es inmediato. En dicho caso es fácil obtener

<a id='Ec10_11'></a>
\begin{equation*}
\begin{aligned}
\mathbf{J}^{(e)}=\left[\begin{array}{ll}
x_{2}-x_{1} & y_{2}-y_{1} \\
x_{3}-x_{1} & y_{3}-y_{1}
\end{array}\right] \quad \mathbf{y} \quad\left|\mathrm{J}^{(e)}\right|=2 A^{(e)}
\end{aligned}
\label{eq:Ec10_11} \tag{10.11}
\end{equation*}

Si el elemento es de lados curvos, es más conveniente operar en función de las coordenadas naturales $\alpha \ddot \beta$ definidas anteriormente, lo que implica sencillamente sustituir $L_{2}$ y $L_{3}$ por $\alpha$ y $\beta$, respectivamente, y $L_{1}$ por $1-\alpha-\beta$. A partir de aquí el cálculo de las derivadas cartesianas de las funciones de forma sigue exactamente los pasos descritos en el apartado anterior, intercambiando simplemente las coordenadas $\xi$ y $\eta$ por $\alpha$ y $\beta$, respectivamente. Así, por ejemplo

<a id='Ec10_12'></a>
\begin{equation*}
\begin{aligned}
\frac{\partial x}{\partial \alpha}=\sum_{i=1}^{n} \frac{\partial N_{i}(\alpha, \beta)}{\partial \alpha} x_{i} \quad ; \quad \frac{\partial x}{\partial \beta}=\sum_{i=1}^{n} \frac{\partial N_{i}\left(\alpha_{1} \beta\right)}{\partial \beta} x_{i} \quad ; \quad \text { etc. }
\end{aligned}
\label{eq:Ec10_12} \tag{10.12}
\end{equation*}

La matriz de rigidez del elemento se obtiene por una expresión análoga a la [Ec. 10.9](#Ec10_9), tal como

<a id='Ec10_13'></a>
\begin{equation*}
\begin{aligned}
K_{i j}^{(e)}=\int_{0}^{1} \int_{0}^{1-\beta} \mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j}\left|\mathbf{J}^{(e)}\right| \mathrm{d} \alpha d \beta=\int_{0}^{1} \int_{0}^{1-\beta} G_{i j}(\alpha, \beta) \mathrm{d} \alpha d \beta
\end{aligned}
\label{eq:Ec10_13} \tag{10.13}
\end{equation*}

donde todos los términos de $\mathbf{B}_{i}, \mathbf{J}^{(e)}$ y $G_{i j}$ se deducen de las [Ecs. 10.6 - 10.9](#Ec10_6) sustituyendo las coordenadas $\xi$ y $\eta$ por $\alpha$ y $\beta$, respectivamente.

En elementos de lados curvos los términos del integrando de [Ec. 10.13](#Ec10_13) son funciones racionales en $\alpha$ y $\beta$. Esta dificultad se complica con la interdependencia de los límites de integración debida a la geometría del elemento. No obstante, el cálculo de las integrales puede efectuarse de manera sencilla y sistemática mediante integración numérica, como veremos en el apartado siguiente.


### Integración numérica en 2D

Hemos visto que mediante la formulación isoparamétrica se pueden transformar todas las integrales sobre el dominio del elemento a otras sobre el espacio de coordenadas naturales. Para el cálculo de dichas integrales, que suelen contener términos racionales, puede hacerse uso de cualquiera de las cuadraturas de integración numérica. Consideraremos aquí de nuevo la cuadratura de $Gauss-Legendre$ sobre dominios bidimensionales.

#### Integración numérica en dominios cuadriláteros isoparamétricos

La integral de un término cualquiera $g(\xi, \eta)$ de la matriz de rigidez $K_{i j}^{(e)}$ sobre el dominio de coordenadas naturales de un elemento cuadrilátero puede evaluarse por una cuadratura de Gauss-Legendre bidimensional como

<a id='Ec10_14'></a>
\begin{equation*}
\begin{aligned}
\int_{-1}^{+1} \int_{-1}^{+1} g(\xi, \eta) \mathrm{d} \xi \mathrm{d} \eta=\int_{-1}^{+1} \mathrm{d} \xi\left[\sum_{q=1}^{n_{q}} g\left(\xi, \eta_{q}\right) W_{q}\right]=\sum_{p=1}^{n_{p}} \sum_{q=1}^{n_{q}} g\left(\xi_{p}, \eta_{q}\right) W_{p} W_{q}
\end{aligned}
\label{eq:Ec10_14} \tag{10.14}
\end{equation*}

donde $n_{p}$ y $n_{q}$ son el número de puntos de integración seleccionados en cada una de las direcciones $\xi$ y $\eta ; \xi_{p}$ y $\eta_{q}$ son las coordenadas naturales del punto de integración $p, q$ y $W_{p}, W_{q}$ los pesos correspondientes a cada dirección en dicho punto.

Las coordenadas y los pesos para cada dirección se deducen directamente de los dados en la [Tabla 5.1](./Sesion08_Elementos1D_avanzados.ipynb#Tabla5_1) del Capítulo 5 para el caso unidimensional. Recordemos que una cuadratura de orden $n$ en cada dirección natural integra exactamente un polinomio de grado $\leq 2 n-1$ en la correspondiente coordenada natural. En la [Figura 10.2](#Fig10_2) se muestran algunas de las cuadraturas bidimensionales más usuales sobre elementos cuadriláteros.

#### Ejemplo 10.1

 Integrar numéricamente la función $f(\xi, \eta)=\xi^{2} \eta^{2}$ para un elemento rectangular plano de dimensiones $2 a \times 2 b$ cuadrado.
 
***Solución***

Por ser el elemento plano $\left|\mathbf{J}^{(e)}\right|=a b$ ([Ec. 10.7](#Ec10_7)).
Por otra parte, por ser el integrado una función cuadrática en $\xi$ y $\eta$ utilizaremos una cuadratura $2 \times 2$ ([Figura 10.2](#Fig10_2)). Así

$$
\begin{aligned}
I &=\iint_{A} \xi^{2} \eta^{2} \mathrm{d}  A=a b \int_{-1}^{+1} \int_{-1}^{+1} \xi^{2} \eta^{2}=a b \sum_{p=1}^{2} \sum_{q=1}^{2}\left(\xi^{2} \eta^{2}\right)_{p, q} w_{p} w_{q}=\\
&=a b\left[\left(-\frac{\sqrt{3}}{3}\right)^{2}\left(-\frac{\sqrt{3}}{3}\right)^{2}+\left(-\frac{\sqrt{3}}{3}\right)^{2}\left(\frac{\sqrt{3}}{3}\right)^{2}+\left(\frac{\sqrt{3}}{3}\right)^{2}\left(-\frac{\sqrt{3}}{3}\right)^{2}+\left(\frac{\sqrt{3}}{3}\right)^{2}\left(\frac{\sqrt{3}}{3}\right)^{2}\right]=\frac{4}{9} a b
\end{aligned}
$$

que es la solución exacta


#### Integración numérica sobre dominios triangulares isoparamétricos

La cuadratura de Gauss para elementos triangulares se escribe como

<a id='Ec10_15'></a>
\begin{equation*}
\begin{aligned}
\int_{0}^{1} \int_{0}^{1-L_{3}} f\left(L_{1}, L_{2}, L_{3}\right) \mathrm{d} L_{2} \mathrm{d} L_{3}=\sum_{p=1}^{n_{p}} f\left(L_{1_{p}}, L_{2_{p}}, L_{3_{p}}\right) W_{p}
\end{aligned}
\label{eq:Ec10_15} \tag{10.15}
\end{equation*}

donde $n_{p}$ es el número de puntos de integración; $L_{1_{p}}, L_{2_{p}}, L_{3_{p}}$ y $W_{p}$ son los valores de las coordenadas de área y del peso en el punto de integración $p$, respectivamente.

En la [Figura 10.3](#Fig10_3) se muestran las coordenadas y los pesos más utilizados en la práctica; la "precisión"en dicha figura es el polinomio de mayor grado que la fórmula integra exactamente. La Figura $10.3$ es también de utilidad inmediata para el cálculo de integrales definidas en función de las coordenadas naturales $\alpha$ y $\beta$ haciendo uso de la relación entre dichas coordenadas y las de área.

Es importante advertir que en dicha figura se han normalizado los pesos $W_{p}$ de manera que su suma sea $\frac{1}{2}$. Es también usual que los pesos se tabulen de modo que sumen la unidad. En este caso debe afectarse el sumatorio en la [Ec. 10.15](#Ec10.15) del coeficiente $1 / 2$ para que el área del elemento se calcule exactamente.

Calcular el área de un elemento triangular de lados rectos por integración numérica.

### Selección del orden de integración

El número de puntos de integración se selecciona de acuerdo con el grado de los polinomios que aparecen en las integrales del elemento. Si el elemento es isoparamétrico dichas integrales contienen funciones racionales y la integración exacta no es posible. En este caso suele escogerse una cuadratura que integre exactamente la matriz (o vector) de un elemento anélogo rectangular o triangular de lados rectos en el que, por ser el jacobiano constante, las integrales sólo contienen funciones polinómicas.

Está comprobado que en este último caso basta con que la cuadratura seleccionada integre exactamente los términos de $K_{i j}^{(e)}$ correspondientes al polinomio completo contenido en las funciones de forma, pues, de hecho, dichos términos son los únicos que contribuyen significativamente a la aproximación y convergencia de la solución.

Este orden de integración recibe el nombre de cuadratura mínima para mantener la convergencia. Vemos como de nuevo una integración "inexacta" de la matriz de rigidez conduce a resultados correctos.

En la práctica la cuadratura mínima es la más $t$ recomendable ya que, obviamente, es la más económica en número de operaciones. Es interesante constatar cómo, en ocasiones, la integración mínima proporciona incluso mejores resultados debido a la mayor flexibilidad que confiere al elemento, que cancela en parte los errores por exceso de rigidez inherentes a la discretización y al campo de desplazamiento supuesto.

Algunos autores asocian el nombre de cuadratura mínima a aquella que garantiza que el elemento puede reproducir en el límite un estado de gradiente constante. Esto implica que la cuadratura escogida debe poder evaluar correctamente el área (o el volumen) del elemento, lo que en coordenadas naturales representa calcular exactamente

$$
\iint_{A^{(e)}}\left|J^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta
$$

En elementos rectangulares y triangulares de lados rectos esta condición es muy débil pues exige únicamente una cuadratura de un solo punto, lo que generalmente viola la exigencia mínima para mantener la convergencia descrita más arriba (salvo en elementos triangulares de tres nodos), y puede dar origen a mecanismos internos asociados a modos de energía nula. Dichos mecanismos se producen cuando el campo de gradientes se anula en los puntos de integración numérica. En ocasiones estos mecanismos son compatibles entre sí y provocan la singularidad de la matriz de rigidez $\mathbf{K}$, con la consiguiente pérdida de la solución. Este es el caso de los mecanismos que se producen en el elemento de cuatro nodos de elasticidad plana (Capítulo siguiente ) con un solo punto de integración (Figura 10.5a). En otros casos los mecanismos del elemento no pueden propagarse en la malla, lo que preserva la solución correcta (Figura 10.5b). De cualquier manera es deseable que el elemento esté libre de mecanismos internos, por lo que es más práctico definir la cuadratura mínima en base a los criterios de integración del polinomio completo antes mencionados, ya que siempre garantizan la integración exacta del área del elemento.

### Integración numérica de matrices y vectores a nivel de elemento

De acuerdo con todo lo anterior, la expresión de la matriz de rigidez de un elemento de Poisson cuadrangular isoparamétrico evaluada por integración numérica sería

<a id='Ec10_16'></a>
\begin{equation*}
\begin{aligned}
K_{i j}^{(e)} &=\iint_{A^{(e)}} \mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j} \mathrm{d} x \mathrm{d} y=\int_{-1}^{+1} \int_{-1}^{+1} \mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j}|\mathbf{J}(e)| \mathrm{d} \xi d_{i}=\\
&=\sum_{p=1}^{n_{p}} \sum_{q=1}^{n_{q}}\left[\mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j}\left|\mathbf{J}^{(e)}\right|\right]_{p, q} W_{\dot{p}} W_{\dot{q}}=\\
&=\sum_{p=1}^{n_{p}} \sum_{q=1}^{n_{q}}\left[G_{i j}\right]_{p, q} W_{p} W_{q}
\end{aligned}
\label{eq:Ec10_16} \tag{10.16}
\end{equation*}

donde $G_{i j}$ está dada en la [Ec. 10.9](#Ec10_9). Para un elemento triangular se deduce de ([10.15](#Ec10_15)) y ([10.16](#Ec10_16))

<a id='Ec10_17'></a>
\begin{equation*}
\begin{aligned}
&\mathbf{K}_{i j}^{(e)}=\int_{0}^{1} \int_{0}^{1-\beta} \mathbf{B}_{i}^{T} \mathbf{D B}_{j}\left|\mathbf{J}^{(e)}\right| \mathrm{d} \alpha \mathrm{d} \beta= \\
&=\sum_{p=1}^{n_{p}}\left[\mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j}\left|\mathbf{J}^{(e)}\right|\right]_{p} W_{p}=\sum_{p=1}^{n_{p}}\left[G_{i j}\right]_{p} W_{p}
\end{aligned}
\label{eq:Ec10_17} \tag{10.17}
\end{equation*}

donde la posición de los puntos de integración y los pesos correspondientes se obtienen de la [Figura 10.3](#Fig10_3).

Vemos, por consiguiente, que el cálculo numérico de la matriz de rigidez exige evaluar el jacobiano $\mathbf{J}^{(e)}$ y su determinante, así como las matrices de deformación $\mathbf{B}_{i}$ y constitutiva $\mathbf{D}$ en cada punto de integración de la cuadratura seleccionada. Similarmente, el cálculo de cualquiera de los vectores de flujos nodales equivalentes que implican integrales sobre el elemento (por ejemplo, el debido a una fuente distribuida por unidad de área $Q$ ) se evaluaría en elementos cuadriláteros por

<a id='Ec10_18'></a>
\begin{equation*}
\begin{aligned}
f_{b_{i}}^{(c)} &=\iint_{A(e)} N_{i} Q \mathrm{d} x \mathrm{d} y=\int_{-1}^{+1} \int_{-1}^{+1} N_{i} Q\left|\mathbf{J}^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta=\\
&=\sum_{p=1}^{n_{p}} \sum_{q=1}^{n_{q}}\left(N_{i} Q\left|\mathbf{J}^{(e)}\right|\right)_{p, q} W_{\dot{p}} W_{\dot{q}}
\end{aligned}
\label{eq:Ec10_18} \tag{10.18}
\end{equation*}

Para elementos triangulares el doble sumatorio se reemplazaría por el sumatorio simple de la [Ec. 10.15](#Ec10_15).

El cálculo del vector de flujos nodales debido a flujos prescritos en el contorno merece un comentario especial. Recordemos de la [Ec. 9.24](./Sesion12_FormulacionMatricial.ipynb#Ec9_24) que las componentes de dicho vector de flujos son

<a id='Ec10_19'></a>
\begin{equation*}
\begin{aligned}
f_{i}^{(e)}=\oint_{l(e)} N_{i}\left(\alpha \phi_{a}-\bar{q}_{n}\right) \mathrm{d} \Gamma
\end{aligned}
\label{eq:Ec10_19} \tag{10.19}
\end{equation*}

donde $l^{(e)}$ es el contorno del elemento en el cual se prescribe el flujo. En general dicho contorno representa en el espacio de coordenadas naturales una recta $\xi=$ cte o $\eta=$ cte (Figura 10.7). 

<a id='Fig10_7'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C10_F07_FlujoContorno2D.png?raw=true" width="350" />
</p>
<div style="text-align: center"><strong>Figura 10.7.</strong>Flujo sobre un contorno de un elemento 2D, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Análisis Estático Lineal. Volumen 1. Sólidos. </a></div>

Por consiguiente, para un contorno de un elemento isoparamétrico cuadrilátero que corresponde, por ejemplo con el lado $\eta=+1$, se tiene

<a id='Ec10_20'></a>
\begin{equation*}
\begin{aligned}
\mathrm{d} \Gamma_{\eta=1} &=\left(\mathrm{d} x^{2}+\mathrm{d} y^{2}\right)_{\eta=1}=\left[\left(\frac{\mathrm{d} x}{\mathrm{d} \xi}\right)^{2}+\left(\frac{\mathrm{d} y}{\mathrm{d} \xi}\right)^{2}\right]_{\eta=1}^{1 / 2} \mathrm{d} \xi=\\
&=\left[\left(\sum_{i=1}^{n} \frac{\mathrm{d} N_{i}}{\mathrm{d} \xi} x_{i}\right)^{2}+\left(\sum_{i=1}^{n} \frac{\mathrm{d} N_{i}}{\mathrm{d} \xi} y_{i}\right)^{2}\right]_{\eta=1}^{1 / 2} \mathrm{d} \xi=c(\xi) \mathrm{d} \xi
\end{aligned}
\label{eq:Ec10_20} \tag{10.20}
\end{equation*}

Sustituyendo esta expresión en [Ec. 10.18](#Ec10_18) se obtiene una integral de líneas que es función únicamente de la coordenada natural $\xi$ y que puede calcularse con una cuadratura unidimensional por

<a id='Ec10_21'></a>
\begin{equation*}
\begin{aligned}
f_{i}^{(e)}=\oint_{l(s)} N_{i}\left(\alpha \phi_{a}-\bar{q}_{n}\right) c(\xi) \mathrm{d} \xi=\int_{-1}^{+1} g(\xi) \mathrm{d} \xi=\sum_{p=1}^{n_{p}} g\left(\xi_{p}\right) W_{p}
\end{aligned}
\label{eq:Ec10_21} \tag{10.21}
\end{equation*}

El número de puntos de integración $n_{p}$ depende de la geometría del contorno y de la variación de los flujos prescritos en el contorno. Se recuerda que en triángulos de tres nodos y cuadriláteros de cuatro nodos, si los flujos prescritos son constantes, basta integrar con un solo punto para obtener

<a id='Ec10_22'></a>
\begin{equation*}
\begin{aligned}
f_{i}^{(e)}=\frac{l^{(e)}}{2}\left(\alpha \phi_{a}-\bar{q}_{n}\right)
\end{aligned}
\label{eq:Ec10_22} \tag{10.22}
\end{equation*}

Es decir, el flujo total sobre el contorno se reparte en partes iguales sobre los dos lados del elemento sobre el contorno.


### Hexaedros isoparamétricos

La formulación isoparamétrica permite utilizar elementos tetraédricos y hexaédricos irregulares y con lados curvos como los que se muestran en la Figura 10.8. La definición del elemento se efectúa a partir de las coordenadas de su geometría real para, con la ayuda de la transformación isoparamétrica, referir el cálculo de todas las integrales a la geometría normalizada del elemento sobre prismas y tetraedros rectos.

<a id='Fig10_8'></a>
<p float="center">
  <img src="https://github.com/carlosalvarezh/FEM/blob/main/images/C10_F08_ElementosIsoparametricos3D.png?raw=true" width="350" />
</p>
<div style="text-align: center"><strong>Figura 10.8.</strong>Elementos isoparamétricos en 3D, por Oñate, <a href="https://books.cimne.com/product/calculo-de-estructuras-por-el-metodo-de-los-elementos-finitos-analisis-estatico-lineal-vol-1/"> Cálculo de Estructuras por el Método de los Elementos Finitos. Consideraremos en primer lugar la formulación isoparamétrica de elementos hexaédricos. Las coordenadas $x, y, z$ de un punto cuaiquiera de un elemento de $n$ nodos se expresan en forma isoparamétrica como

<a id='Ec10_23a'></a>
\begin{equation*}
\begin{aligned}
\mathbf{x}=\left\{\begin{array}{l}
x \\
y \\
z
\end{array}\right\}=\sum_{i=1}^{n} \mathbf{N}_{i}\left\{\begin{array}{l}
x_{i} \\
y_{i} \\
z_{i}
\end{array}\right\}=\mathbf{N} \mathbf{x}^{(e)}
\end{aligned}
\label{eq:Ec10_23a} \tag{10.23a}
\end{equation*}

con

<a id='Ec10_23b'></a>
\begin{equation*}
\begin{aligned}
\mathbf{N}=\left[\mathbf{N}_{1}, \mathbf{N}_{2}, \ldots, \mathbf{N}_{n}\right] ; \quad \mathbf{N}_{i}=\left[\begin{array}{ccc}
N_{i} & & \\
& N_{i} & \\
& & N_{i}
\end{array}\right] ; \quad N_{i}=f(\xi, \eta, \zeta)
\end{aligned}
\label{eq:Ec10_23b} \tag{10.23b}
\end{equation*}

donde $N_{i}$ es la misma función de forma utilizada para interpolar el campo de incógnitas.

La [Ec. 10.23a](#Ec10_23a) expresa una relación entre las coordenadas cartesianas y las naturales. Dicha relación es biunívoca si se cumple que el determinante del jacobiano de la transformación $x, y, z \rightarrow \xi, \eta, \zeta$ es de signo positivo en todos los puntos del elemento, lo que se cumple para los casos más usuales salvo para formas muy distorsionadas de los elementos. En general son aplicables las reglas dadas anteriormente para elementos bidimensionales.

Para el cálculo de las derivadas cartesianas de las funciones de forma se. sigue un procedimiento análogo al explicado para ele mentos bidimensionales. Así, la regla de derivación en cadena permite escribir en forma matricial

<a id='Ec10_24'></a>
\begin{equation*}
\begin{aligned}
\left\{\begin{array}{c}
\frac{\partial N_{i}}{\partial \xi} \\
\frac{\partial N_{i}}{\partial \eta} \\
\frac{\partial N_{i}}{\partial \zeta}
\end{array}\right\}=\left[\begin{array}{lll}
\frac{\partial x}{\partial \xi} & \frac{\partial y}{\partial \xi} & \frac{\partial z}{\partial \xi} \\
\frac{\partial x}{\partial \eta} & \frac{\partial y}{\partial \eta} & \frac{\partial z}{\partial \eta} \\
\frac{\partial x}{\partial \zeta} & \frac{\partial y}{\partial \zeta} & \frac{\partial z}{\partial \zeta}
\end{array}\right]\left\{\begin{array}{c}
\frac{\partial N_{i}}{\partial x} \\
\frac{\partial N_{i}}{\partial y} \\
\frac{\partial N_{i}}{\partial z}
\end{array}\right\}=\mathbf{J}^{(e)} \frac{\partial \mathbf{N}_{i}}{\partial \mathbf{x}}
\end{aligned}
\label{eq:Ec10_24} \tag{10.24}
\end{equation*}

donde $\mathbf{J}^{(e)}$ es la matriz jacobiana cuyos términos se calculan haciendo uso de las relaciones isoparamétricas [Ec. 10.23a](#Ec10_23a) como

<a id='Ec10_25'></a>
\begin{equation*}
J^{(e)}=\sum_{i=1}^{n}\left[\begin{array}{lll}
\frac{\partial N_{i}}{\partial \xi} x_{i} & \frac{\partial N_{i}}{\partial \xi} y_{i} & \frac{\partial N_{i}}{\partial \xi} z_{i} \\
\frac{\partial N_{i}}{\partial \eta} x_{i} & \frac{\partial N_{i}}{\partial \eta} y_{i} & \frac{\partial N_{i}}{\partial \eta} z_{i} \\
\frac{\partial N_{i}}{\partial \zeta} x_{i} & \frac{\partial N_{i}}{\partial \zeta} y_{i} & \frac{\partial N_{i}}{\partial \zeta} z_{i}
\end{array}\right]
\label{eq:Ec10_25} \tag{10.25}
\end{equation*}

Invirtiendo [Ec. 10.24](#Ec10_24) se encuentra la expresión de las derivadas cartesianas de $N_{i}$ como

<a id='Ec10_26'></a>
\begin{equation*}
\begin{aligned}
\boldsymbol{\nabla} N_{i}=\left\{\begin{array}{c}
\frac{\partial N_{i}}{\partial x} \\
\frac{\partial N_{i}}{\partial y} \\
\frac{\partial N_{i}}{\partial z}
\end{array}\right\}=\left[{J}^{(e)}\right]^{-1}\left\{\begin{array}{c}
\frac{\partial N_{i}}{\partial \xi} \\
\frac{\partial N_{i}}{\partial \eta} \\
\frac{\partial N_{i}}{\partial \zeta}
\end{array}\right\}
\end{aligned}
\label{eq:Ec10_26} \tag{10.26}
\end{equation*}

Por tanto, combinando [Ec. 10.25](#Ec10_25) y [Ec. 10.26](#Ec10_26) puede expresarse la matriz de gradientes del nodo $i$ de un elemento isoparamétrico tridimensional en función de las coordenadas naturales como

<a id='Ec10_27a'></a>
\begin{equation*}
\begin{aligned}
\mathbf{B}_{i}(\xi, \eta, \zeta)=\nabla N_{i}=\left\{\begin{array}{l}
\bar{b}_{i} \\
\bar{c}_{i} \\
\bar{d}_{i}
\end{array}\right\}
\end{aligned}
\label{eq:Ec10_27a} \tag{10.27a}
\end{equation*}

donde

<a id='Ec10_27b'></a>
\begin{equation*}
\begin{aligned}
\left\{\begin{array}{c}
\bar{b}_{i} \\
\bar{c}_{i} \\
\bar{d}_{i}
\end{array}\right\}=\sum_{k=1}^{3}\left\{\begin{array}{l}
\bar{J}_{1 k}^{(e)} \\
\bar{J}_{2 k}^{(e)} \\
\bar{J}_{3 k}^{(e)}
\end{array}\right\} \frac{\partial N_{i}}{\partial \xi_{k}}
\end{aligned}
\label{eq:Ec10_27b} \tag{10.27b}
\end{equation*}

donde $\bar{J}_{i j}^{(e)}$ es el elemento $i j$ de la de la matriz jacobiana inversa, $\left[\mathrm{J}^{(e)}\right]^{-1}$, $\xi_{1}=\xi, \xi_{2}=\eta$ y $\xi_{3}=\zeta$.

Por otra parte, un diferencial de volumen se expresa por

<a id='Ec10_28'></a>
\begin{equation*}
\begin{aligned}
\mathrm{d} x \mathrm{d} y \mathrm{d} z=\left|{J}^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta \mathrm{d} \zeta
\end{aligned}
\label{eq:Ec10_28} \tag{10.28}
\end{equation*}

La expresión genérica de la matriz de rigidez de un elemento hexaédrico. tridimensional en coordenadas naturales es, por tanto,

<a id='Ec10_29a'></a>
\begin{equation*}
\begin{aligned}
\begin{aligned}
K_{i j}^{(e)} &=\iiint_{V^{(e)}} \mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j} \mathrm{d} V=\\
&=\int_{-1}^{+1} \int_{-1}^{+1} \int_{-1}^{+1} \mathbf{B}_{i}^{T}(\xi, \eta, \zeta) \mathbf{D} \mathbf{B}_{j}(\xi, \eta, \zeta)|\mathbf{J}^{(e)}| \mathrm{d} \xi \mathrm{d} \eta \mathrm{d} \zeta=\\
&=\int_{-1}^{+1} \int_{-1}^{+1} \int_{-1}^{+1} G_{i j}(\xi, \eta, \zeta) \mathrm{d} \xi \mathrm{d} \eta \mathrm{d} \zeta
\end{aligned}
\end{aligned}
\label{eq:Ec10_29a} \tag{10.29a}
\end{equation*}

con

<a id='Ec10_29b'></a>
\begin{equation*}
\begin{aligned}
\begin{aligned}
G_{i j}=\mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j}\left|\mathbf{J}^{(e)}\right|
\end{aligned}
\end{aligned}
\label{eq:Ec10_29b} \tag{10.29b}
\end{equation*}

Es fácil apreciar que en $G_{i j}$ intervienen expresiones racionales por lo que su integración analítica es sumamente complicada y hay que recurrir a la integración numérica


### Tetraedros isoparamétricos

En elementos tetraédricos que utilicen coordenadas de volumen la interpolación isoparamétrica se define similarmente a la [Ec. 10.23a](#Ec10_23a), siendo ahora $N_{i}=f\left(L_{1}, L_{2}, L_{3}, L_{4}\right) .$ Si los tetraedros son de lados rectos el cálculo de las derivadas cartesianas de las funciones de forma es inmediato. Asimismo, las integrales sobre el elemento pueden calcularse exactamente. Si el elemento es de lados curvos es más conveniente, como ocurría en el caso bidimensional, operar con las coordenadas naturales lo que simplemente implica sustituir $L_{2}, L_{3}$ y $L_{4}$ por $\alpha, \beta$ y $\gamma$, respectivamente, y $L_{1}$ por $1-\alpha-\beta-\gamma$. A partir de aquí el cálculo de las componentes de la matriz $\mathbf{B}_{i}$ sigue idénticos pasos a los explicados entre las ecs. [10.23 - 10.29](#Ec10_23a) para elementos hexaédricos, sin más que sustituir las coordenadas $\xi, \eta, \zeta$ por $\alpha, \beta, \gamma$, respectivamente.

Por consiguiente, la matriz de rigidez del' elemento tetraédrico isoparamétrico de lados curvos tiene una expresión similar a la [Ec. 10.29](#Ec10_29a), tal como

<a id='Ec10_30'></a>
\begin{equation*}
\begin{aligned}
\begin{aligned}
K_{i j}^{(e)}=\int_{0}^{1} \int_{0}^{1-\alpha} \int_{0}^{1-\alpha-\beta} G_{i j}(\alpha, \beta, \gamma) \mathrm{d} \alpha \mathrm{d} \beta \mathrm{d} \gamma
\end{aligned}
\end{aligned}
\label{eq:Ec10_30} \tag{10.30}
\end{equation*}

donde la matriz $G_{i j}(\alpha, \beta, \gamma)$ se deduce de [Ec. 10.29b](#Ec10_29b), teniendo en cuenta el cambio de variables mencionado.


### Integración numérica en 3D

#### Elementos hexaédricos

Sea $f(x, y, z)$ una función cualesquiera definida sobre un elemento hexaédrico isoparamétrico. Para calcular el valor de la integral de dicha función sobre el elemento efectuamos en primer lugar la transformación del dominio real al normalizado hexaédrico recto, es decir

<a id='Ec10_31'></a>
\begin{equation*}
\begin{aligned}
\begin{gathered}
\iiint_{V(e)} f(x, y, z) \mathrm{d} x \mathrm{d} y \mathrm{d} z=\int_{-1}^{1} \int_{-1}^{1} \int_{-1}^{1} f(\xi, \eta, \zeta)\left|\mathbf{J}^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta \mathrm{d} \zeta= \\
=\int_{-1}^{+1} \int_{-1}^{+1} \int_{-1}^{+1} g\left(\xi, r_{1}, \zeta\right) \mathrm{d} \xi \mathrm{d} \eta \mathrm{d} \zeta
\end{gathered}
\end{aligned}
\label{eq:Ec10_31} \tag{10.31}
\end{equation*}

Si utilizamos la cuadratura de Gauss-Legendre, el valor de la integral se calcula por la expresión

<a id='Ec10_32'></a>
\begin{equation*}
\begin{aligned}
&\int_{-1}^{+1} \int_{-1}^{+1} \int_{-1}^{+1} g(\xi, \eta, \zeta) \mathrm{d} \xi \mathrm{d} \eta \mathrm{d} \zeta=\int_{-1}^{+1} \int_{-1}^{+1} \sum_{p=1}^{n_{p}} W_{p} g\left(\xi_{p}, \eta, \zeta\right) \mathrm{d} \eta \mathrm{d} \zeta= \\
&=\int_{-1}^{+1} \sum_{q=1}^{n_{q}} \sum_{p=1}^{n_{p}} W_{p} W_{q} g\left(\xi_{p}, \eta_{q}, \zeta\right) \mathrm{d} \zeta=\sum_{r=1}^{n_{r}} \sum_{q=1}^{n_{q}} \sum_{p=1}^{n_{p}} W_{p} W_{q} W_{r} g\left(\xi_{p}, \eta_{q}, \zeta_{r}\right)
\end{aligned}
\label{eq:Ec10_32} \tag{10.32}
\end{equation*}

donde $n_{p}, n_{q}$ y $n_{r}$ son el número de puntos de integración en cada una de las direcciones $\xi, \eta$ y $\zeta ; \xi_{p}, \eta_{q}, \zeta_{r}$ son las coordenadas del punto de integración $(p, q, r)$ y $W_{p}, W_{q}, W_{r}$ son los pesos correspondientes a cada dirección natural asociados a dicho punto.

Las coordenadas y los pesos para cada dirección se deducen directamente del caso unidimensional para el caso unidimensional. Recordemos que una cuadratura de orden $n$ en cada dirección natural integra exactamente un polinomio de grado $2 n-1$ ó menor en la correspondiente coordenada natural. \end{equation*}

### Integración numérica de las matrices y vectores de elementos en 3D

#### Elementos hexaédricos

De acuerdo con las ecs. [(10.29)](#Ec10_29) y [(10.32)](#Ec10_32) el cálculo de la matriz de rigidez de un elemento hexaédrico isoparamétrico sería

<a id='Ec10_35'></a>
\begin{equation*}
\begin{aligned}
K_{i j}^{(e)}=\iiint_{V^{(e)}} \mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j} \mathrm{d} x \mathrm{d} y \mathrm{d} z=\int_{-1}^{+1} \int_{-1}^{+1} \int_{-1}^{+1} \mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j}\left|\mathbf{J}^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta \mathrm{d} \zeta=\\  =\sum_{p=1}^{n_{p}} \sum_{q=1}^{n_{q}} \sum_{r=1}^{n_{r}}\left[\mathbf{B}_{i}^{T} \mathbf{D} \mathbf{B}_{j}|\mathbf{J}(e)|\right]_{p, q, r} W_{p} W_{q} W_{r}=\sum_{p=1}^{n_{p}} \sum_{q=1}^{n_{q}} \sum_{r=1}^{n_{r}}\left[G_{i j}\right]_{p, q, r} W_{p} W_{q} W_{r}
\end{aligned}
\label{eq:Ec10_35} \tag{10.35}
\end{equation*}

El cálculo de los vectores de flujos nodales equivalentes que implican integrales sobre el volumen del elemento se efectuaría de manera idéntica.

Así, para los flujos nodales debidos a un flujo distribuido $Q$

<a id='Ec10_36'></a>
\begin{equation*}
\begin{aligned}
f_{i}^{(e)}=\iiint_{V^{(e)}} N_{i} Q \mathrm{d} x \mathrm{d} y \mathrm{d} z=\int_{-1}^{+1} \int_{-1}^{+1} \int_{-1}^{+1} N_{i} Q\left|\mathbf{J}^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta \mathrm{d} \zeta=\\
=\sum_{p=1}^{n_{p}} \sum_{q=1}^{n_{q}} \sum_{r=1}^{n_{r}}\left[N_{i} Q\left|\mathbf{J}^{(e)}\right|\right]_{p, q, r} W_{p} W_{q} W_{r}
\end{aligned}
\label{eq:Ec10_36} \tag{10.36}
\end{equation*}

El caso de flujos actuando sobre una de las caras del elemento es algo más complicado. Para explicar el proceso consideremos que actúa un flujo $t_{n}$ ortogonal a la cara situada en $\zeta=+1$ y definida por los nodos 5 a 8 (*Figura 10.11*).

El flujo nodal equivalente viene dado

<a id='Ec10_37'></a>
\begin{equation*}
\begin{aligned}
f_{i}=\iint_{A} N_{i} t_{n} \mathrm{d} A
\end{aligned}
\label{eq:Ec10_37} \tag{10.37}
\end{equation*}

El diferencial de area se obtiene por $\left|\vec{V}_{1} \times \vec{V}_{2}\right|$, donde $\vec{V}_{1}$ y $\vec{V}_{2}$ son dos vectores tangentes a las líneas $\eta=$ cte y $\xi=$ cte, respectivamente, contenidas en la superficie $A$. Así

<a id='Ec10_38'></a>
\begin{equation*}
\begin{aligned}
&\vec{V}_{1}=\left(\frac{\partial x}{\partial \xi} \overrightarrow{\mathbf{i}}+\frac{\partial y}{\partial \xi} \overrightarrow{\mathbf{j}}+\frac{\partial x}{\partial \xi} \overrightarrow{\mathbf{k}}\right)_{\zeta=+1} \mathrm{d} \xi \\
&\vec{V}_{2}=\left(\frac{\partial x}{\partial \eta} \overrightarrow{\mathbf{i}}+\frac{\partial y}{\partial \eta} \overrightarrow{\mathbf{j}}+\frac{\partial z}{\partial \eta} \overrightarrow{\mathbf{k}}\right)_{\zeta=+1} \mathrm{d} \eta
\end{aligned}
\label{eq:Ec10_38} \tag{10.38}
\end{equation*}

Por tanto

<a id='Ec10_39'></a>
\begin{equation*}
\begin{aligned}
\vec{V}_{1} \times \vec{V}_{2} \mid=\left\{\begin{array}{l}
J_{12} J_{23}-J_{22} J_{13} \\
J_{21} J_{13}-J_{11} J_{23} \\
J_{11} J_{32}-J_{21} J_{12}
\end{array}\right\}_{\zeta=+1}^{(e)} \mathrm{d} \xi \mathrm{d} \eta=\frac{1}{\mathrm{d} A} \mathbf{j}^{(e)} \mathrm{d} \xi \mathrm{d} \eta
\end{aligned}
\label{eq:Ec10_39} \tag{10.39}
\end{equation*}

donde las $\mathbf{j}_{i j}^{(e)}$ se obtienen de [Ec. 10.25](#Ec10_25).

Por consiguiente, la expresión final del vector de flujos normales es

<a id='Ec10_40'></a>
\begin{equation*}
\begin{aligned}
f_{t_{i}}^{(e)} &=\iint_{A^{(e)}} N_{i} t_{n} \mathrm{d} A=\int_{-1}^{+1} \int_{-1}^{+1} N_{i} t_{n}\left|\mathbf{j}^{(e)}\right| \mathrm{d} \xi \mathrm{d} \eta=\\
&=\sum_{p=1}^{n_{p}} \sum_{q=1}^{n_{q}}\left[N_{i} t_{n}\left|\mathbf{j}^{(e)}\right|\right]_{p, q} W_{p} W_{q}
\end{aligned}
\label{eq:Ec10_40} \tag{10.40}
\end{equation*}

donde $\mathbf{j}^{(e)}$ se deduce de [Ec. 10.40](#Ec10_40) y $N_{i}=N_{i}(\xi, \eta, \zeta=+1)$.

Las expresiones anteriores son la base para la elaboración de las subrutinas correspondientes de un programa de ordenador para elementos tridimensionales.
